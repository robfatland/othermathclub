# Graphs and NetworkX


### Motivation


Graphs one might say concern the mathematics of abstract relationships. 
Graphs are easy to define in terms of two sets: A 
set of *things* first, and then a second set whose elements are
pairs of things from the first set.
What could be simpler?


> Note on terminology: The *things* are interchangeably called **nodes** 
or **vertices** of a graph. The *pairs of things* are called **edges**.


Graphs occupy an interesting crossroads in the landscape of 
mathematics: Connecting with set theory, with logic, with geometry, 
with combinatorics, with probability, with network theory, and so
on (not to mention graph theory relevance to practical subjects). 
The list of references below are a start on a trail of stories,
problems and clues involving graphs.


This notebook uses the machinery of the Python NetworkX library 
to explore properties of graphs, leading up to 
'spectral theory' in the ***Spectral Graph Theory*** notebook. 
Along the way I will describe constructing and working with graphs 
in the NetworkX formalism. 


> Terminology: `NetworkX` is abbreviated `nx` in code



### Two problems


By way of tactics *senza* code, here is an interesting problem in graph theory: 
Prove that a graph G with $2n$ vertices ($n \; \ge \; 2$) and $n^2 + 1$ edges 
contains a triangle. 


> Spoiler alert: Contradiction works here. That is: Presume the graph 
contains no triangles and count the number of forbidden edges. If the 
total possible number of edges minus the forbidden count is less than
the number of edges that are given...


Another problem in this vein is that of the wood-boring worm who finds
a 12-centimeter cube comprised of 27 identical smaller 4-cm 
cubes. All these cubes are made of wood. The worm chooses an external 
face of one of the small cubes and begins chewing its way
through the interior. It can leave one of the small cubes to enter 
another only via cube faces, not through corners or
edges. Once the worm enters the big cube: Can it visit each of the
small cubes once and arrive finally at the cube in the center?

# Tutte's Embedding Theorem

This was mentioned in a lecture by Daniel Spielman: A 3-vertex-connected planar graph can
be supplied with springs for edges and allowed to relax into a minimal energy configuration
where planarity is directly evident! Well let's see how hard it is to make that happen.


- Begin with G = a ring of vertices
    - Add new edges at random requiring that...
        - ...a new edge maintains the graph as planar
    - Repeat until vertex connectivity reaches 3
    - Given N there is a maximum number of edges, empirically
        - 10: 24
        - 11: 27
        - 12: 30
        - 13: 33
        - ...
        - 40: 114
        - 80: 234
        - N: 3N - 6
        - N = 200 starts to test one's patience
- Identify 3+ vertices making up a face
    - Since the graph begins as a ring: Any vertex connecting to two adjacent vertices defines a face
- Pin the face vertices as a regular polygon
    - this must surround the remaining vertices 
- Assign random positions to remaining vertices
    - Draw the graph
- "Make the remaining edges springs, vertices can move"
- Calculate the potential energy of the graph from the springs
- Adjust positions of the movable vertices
    - `while` a vertex feels a net force:
        - Move that vertex some distance in that direction
        - Shall we use Hooke's law?
- Recalculate the potential energy
- Animate / Draw the end result

````python
# G is a ring of vertices with edges added; still planar

N, G = 12, nx.Graph()
for i in range(N-1): G.add_edge(i, i+1)
G.add_edge(0, N-1)
print("G nodes = ", G.number_of_nodes())

done = False
tries = 0
try_limit = 10000
c_max = 0
add_try_max = 0
while not done:
    tries += 1
    H = G.copy()
    i = randint(0, N-1)
    j = randint(0, N-1)
    if not i == j and not G.has_edge(i, j):
        H.add_edge(i, j)
        if nx.is_planar(H):
            G.add_edge(i, j)
            add_try_max = tries
            c = nx.node_connectivity(G)
            if c > c_max:
                c_max = c
                print("increase connectivity to", c_max, "on try", tries)
    if tries >= try_limit:
        break
        
print("end connectivity:", nx.node_connectivity(G))
print("number of edges:", G.number_of_edges())
# ShowGraph(G, 'cyan')

# Find a triangular face
# face_a = face_b = face_c = -1
# done = False
# for i in range(N-2):
#     for j in range(i+1, N-1):
#         if G.has_edge(i, j) and G.has_edge(i, j+1):
#             face_a = i
#             face_b = j
#             face_c = j+1
#             done = True
#             break
#     if (done): break

# Find a square face
face_a = face_b = face_c = face_d = -1
done = False
nTriesSquare = 0
while not done:
    nTriesSquare += 1
    ca = randint(0, N-1)
    cb = randint(0, N-1)
    cc = randint(0, N-1)
    cd = randint(0, N-1)
    if (not ca == cb and not ca == cc and not ca == cd and
        not cb == cc and not cb == cd and
        not cc == cd
        and G.has_edge(ca, cb)
        and G.has_edge(cb, cc)
        and G.has_edge(cc, cd)
        and G.has_edge(cd, ca)
        and not G.has_edge(ca, cc)
        and not G.has_edge(cb, cd)):
        done = True
        face_a = ca
        face_b = cb
        face_c = cc
        face_d = cd
        
print("Found a square, it took", nTriesSquare, "tries")
print(face_a, face_b, face_c, face_d)

# Now that we have three working vertices let's pin them down
#   on an equilateral triangle centered on (1/2, 1/2).
#   This is the start of the construction of the pos[] list.
#   Then locate the remaining N - 3 vertices on the unit square.

# tri_y = .8
# tri_x = tri_y * sqrt(3)

center_x = 1/2
center_y = 1/2

# triangle
# pos[face_a] = [center_x - tri_x, center_y - tri_y]
# pos[face_b] = [center_x + tri_x, center_y - tri_y]
# pos[face_c] = [center_x,         center_y + 2*tri_y]

# square
hw_sq = 0.6
pos[face_a] = [center_x - hw_sq, center_y + hw_sq]
pos[face_b] = [center_x + hw_sq, center_y + hw_sq]
pos[face_c] = [center_x + hw_sq, center_y - hw_sq]
pos[face_d] = [center_x - hw_sq, center_y - hw_sq]

fixed_vertices = [face_a, face_b, face_c, face_d]

for i in range(N):
    if not i in fixed_vertices:
        x = random()
        y = random()
        pos[i] = [x, y]

# All faces in pos are now uniformly random on the unit square

# Make copies of G and pos for future comparison. Draw G.

def TuttePlot(G, pos, color):
    plt.figure(figsize=(7, 7))
    nx.draw_networkx_edges(G, pos)
    nx.draw_networkx_nodes(G, pos, node_size=10, node_color = color)
    labels = nx.draw_networkx_labels(G, pos, font_size = 26)
    plt.axis('off')
    plt.show()
    
pos_copy = pos.copy()
G_copy = G.copy()
TuttePlot(G_copy, pos_copy, "red")
````

````python
# Generate an adjacency list of lists
adjacent = []
for i in range(N):
    this_a = []
    for j in range(N):
        if G.has_edge(i, j): this_a.append(j)
    adjacent.append(this_a)
    
# Expedite experiments in this section: Recopy G, pos from copies
G = G_copy.copy()
pos = pos_copy.copy()
````

````python
nTimes = 20000
epsilon = .01
pin_factor = 3

for time in range(nTimes):
    for i in range(N):
        if not i in fixed_vertices:
            x_sum = 0.
            y_sum = 0.
            for neighbor in adjacent[i]:
                if neighbor in fixed_vertices:
                    x_sum += pin_factor*(pos[neighbor][0] - pos[i][0])
                    y_sum += pin_factor*(pos[neighbor][1] - pos[i][1])
                else:
                    x_sum += pos[neighbor][0] - pos[i][0]
                    y_sum += pos[neighbor][1] - pos[i][1]
            # if i == 7: print(x_sum, y_sum)
            pos[i][0] += x_sum * epsilon
            pos[i][1] += y_sum * epsilon

plt.figure(figsize=(7, 7))
nx.draw_networkx_edges(G, pos)
nx.draw_networkx_nodes(G, pos, node_size=10, node_color = 'blue')
labels = nx.draw_networkx_labels(G, pos, font_size = 26)

# nx.draw_networkx_edges(G_copy, pos_copy, edge_color = 'red')
# nx.draw_networkx_nodes(G_copy, pos_copy, node_size=10, node_color = 'red')
plt.axis('off')
plt.show()
````

Let's decipher the brute force code above. $n$ is the number of nodes in $G$. Here is a paraphrase of a repl: 

```
F = [(G,[list(G.nodes())[0]])]
print(type(F[0]), len(F[0]), type(F[0][0]), type(F[0][1]), len(F[0][1]), type(F[0][1][0]), F[0][1][0])

'tuple', 2, graph G, 'list', 1, 'int', 0
```

From this we see that F is a list of one element; which is a 2-ple: The passed graph G and (second) a list of length 1, with list[0] = 0. 
This zero is the index of the first element of G's node list, i.e. the start of a path. The second element of the tuple (i.e. the list) 
will prove to be a path of vertices. 


The `while` loop runs until `F` is empty.


`graph, path = F.pop()` breaks out the `F` tuple into `graph` and `path`; leaving `F` an empty list after the pop(). Remember `path` will grow.


`confs` is an empty list that grows as we go. It accumulates (Graph, path-list) tuples.


`neighbors` is a list of nodes that are neighbors to the last node on the working `path`... excluding that node itself. This implies that if 
G has self-loops then `.neighbors(node)` would include 'myself'.  

Incidentally `G.neighbors(node)` returns an enumerable; I found it expedient to use it like `range()`: 

```
for i in G.neighbors(4): print(i)

9 18 11 13 6 8 17 19 1 3 12 5 14 7 16
```

As the while progresses brute force additions are made to generate new paths and graphs that have the newest path vertext removed. 
I'll leave it at that for now but it is an interesting bit of code.

````python
import numpy.linalg

# These imports give us control sliders that we use for selecting depth slices from the dataset
from ipywidgets import *
from traitlets import dlink
 
e = []
v = []

# This creates a 2D color-coded view of oxygen at the surface, attaching a slider to a depth parameter
def DrawSomeGraph(type_choice, n, control, layout):
    
    global e, v
    
    plt.figure(num=None, figsize=(6,6), dpi=80)
    plt.axis('off')
    fig = plt.figure(1)
    
    title_string = 'i was never given a title for this... odd'

    probability = float(control)/100.0

    if type_choice == 0: # G(n,m) chooses one possible graph with n vertices and m edges at random
        m = control # number of edges
        G = nx.gnm_random_graph(n, m)
        title_string = 'Randomly chosen from among all graphs with m vertices and n edges'
    elif type_choice == 1: 
        G = nx.random_graphs.erdos_renyi_graph(n, probability)
        title_string = 'Erdos-Renyi random graph with probability ' + str(probability)
    elif type_choice == 2:
        # Use this graph to win a trophy at an information theory conference
        G=nx.karate_club_graph()
        title_string = 'The Karate club graph; go claim your trophy!'
    elif type_choice == 3: 
        G = nx.heawood_graph()
        title_string = 'Heawood graph, a cubic graph on 14 vertices with girth 6'
    elif type_choice == 4:
        k = min(n, 5)
        G = nx.newman_watts_strogatz_graph(n, k, probability)
        title_string = 'Small world graph'
    elif type_choice == 5: 
        k = min(n, 10)
        G = nx.powerlaw_cluster_graph(n, k, probability)
        title_string = 'Power law cluster graph'
    elif type_choice == 6: 
        k = int(control * n / 100)
        if k < 1: k = 1
        G = nx.barabasi_albert_graph(n, k)
        title_string = 'Preferential attachment graph'
    elif type_choice == 7: 
        G = nx.complete_graph(n)
        title_string = 'complete graph'
    elif type_choice == 8: 
        G = nx.complete_multipartite_graph(n, control)
        title_string = 'complete bipartite n, m'
    elif type_choice == 9: 
        # since the size parameter is a dimension (exponent) we auto-curtail this
        if n > 7: n = 7
        G = nx.hypercube_graph(n)
        title_string = 'Hypercube graph where n is now the dimension'
    elif type_choice == 10: 
        G = nx.random_lobster(n, probability, probability)
        title_string = 'random lobster with both probabilities ' + str(probability)
    else:
        distance_threshold = float(control)/100.
        G = nx.random_geometric_graph(n, distance_threshold)
        title_string = 'random geometric using a threshold of ' + str(distance_threshold)

    
    # Use the layout parameter to choose an arrangement scheme
    if layout == 0:
        pos=nx.circular_layout(G)
        title_string += '    (circular)'
    elif layout == 1:
        pos=nx.random_layout(G) 
        title_string += '    (random)'
    elif layout == 2:
        if type_choice == 9:
            # Multiple shells hang on a hypercube 
            pos=nx.shell_layout(G)
        else: 
            shells = []
            nShells = 1
            if n > 5: nShells += 1
            if n > 11: nShells += 1
            if n > 36: nShells += 1
            if n > 72: nShells += 1
            deltaShell = int(n / nShells)
            # n = 43 gives nShells = 4, deltaShell = 10; 0-10, 10-20, 20-30, 30-40
            for i in range(nShells):
                shells.append(range(int(i*deltaShell), int((i+1)*deltaShell)))
            if n > nShells * deltaShell: 
                shells.append(range(int(nShells*deltaShell), n))
            pos=nx.shell_layout(G, shells) 
        title_string += '    (shell)'
    elif layout == 3:
        pos=nx.spring_layout(G) 
        title_string += '    (spring)'
    elif layout == 4:
        pos=nx.spectral_layout(G)
        title_string += '    (spectral)'
    else: # there is a 'graphviz' layout but not using this 'pos =' mechanism
        pos=nx.circular_layout(G)
        title_string += '    (circular)'
    
    # This does not work properly
    # nx.draw(G)
    cut = 1.1
    xmax = cut * max(xx for xx, yy in pos.values())
    ymax = cut * max(yy for xx, yy in pos.values())
    xmin = cut * min(xx for xx, yy in pos.values())
    ymin = cut * min(yy for xx, yy in pos.values())
    plt.xlim(xmin, xmax)
    plt.ylim(ymin, ymax)
    plt.title(title_string)

    nx.draw_networkx(G,pos,with_labels=False,node_size=50,node_color='blue')
    plt.show()
    
    # print(title_string)
    
    L = nx.normalized_laplacian_matrix(G)
    e, v = numpy.linalg.eig(L.A)
    e.sort()
    for i in range(len(e)):
        if e[i] < 1.0e-14: e[i] = 0
    eivstring = "Eigenvalues:"
    nTerms = min(n, 9)
    for i in range(nTerms):
        eivstring += '%.3f' % e[i] + ', '
    if n > 10: eivstring += ', ..., '
    eivstring += '%.3f' % e[n-1]
    print(eivstring)
    # plt.hist(e, bins=100)  # histogram with 100 bins
    # plt.xlim(0, 2)  # eigenvalues between 0 and 2 
          
# This is the interactive slider
interact(DrawSomeGraph, 
            type_choice={'Gnm random': 0, \
                         'Erdos-Renyi random': 1, \
                         'Karate Club': 2, \
                         'Heawood Graph': 3, \
                         'Small world': 4, \
                         'Power law cluster': 5, \
                         'Preferential attachment': 6, \
                         'Complete': 7, \
                         'Complete Bipartite': 8, \
                         'Hypercube': 9, \
                         'Random Lobster': 10 \
                        }, 
            # type_choice=widgets.IntSlider(min=0,max=20,step=1,value=2, continuous_update=False),
            n=widgets.IntSlider(min=2,max=100,step=1,value=10, continuous_update=False),
            control=widgets.IntSlider(min=0,max=100,step=1,value=50,continuous_update=False),
            layout = {'circular': 0, \
                      'random': 1, \
                      'shell': 2, \
                      'spring': 3, \
                      'spectral': 4
                        }
        )
````
