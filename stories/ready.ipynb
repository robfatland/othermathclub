### Whether tis nobler


Recently I encountered -- second-hand -- a person with an ill-considered opinion. 
Shocking! A friend was present, not I, and as they did not have the time, 
inclination, patience or energy to give the time of day let alone a rebuttal, 
they simply wandered off to quieter more interesting places. Good for them!


Now my question is: Under what circumstances might we, indeed, exchange a word
concerning ill-considered opinions? 


Exchanging a word *without* the necessary time to follow through is a lost cause. 
At some point in the future, though: Suppose there is time for such a word. How 
shall we then presume to engage? To be clear this is not 'How do I enlighten a fool?' 
Rather, 'Given a disagreement: When and under what circumstances shall I attempt 
a conversation?'


I propose to give a practical answer, particularly on the 'what circumstances?' part.
The answer is not about how to slam the door and turn the key in a lock, rejoicing
over another person converted to my idea of enlightenment. The answer is about opening 
doors to new ideas.


The short version of what follows is this: Prepare; and by being prepared: Keep that 
conversation moving along non-threatening axes. Puzzling? Indeed. This is in fact a
very fun puzzle to try and solve. Much of what follows involves how science works.
Keep in mind that science is never done, it is never complete, and the business of
science depends upon knowing for sure that the easiest person to fool is us.

### Necessary conditions

At this time in history: Identity is knotted up with confrontation. 
Pointed remarks, taking offense, escalating, hurling insults, breaking windows and worse: 
The underlying principle is '*question me: question my very existence*' in this age of outrage. 
So let's agree that aggravating another person along the identity axis is an option
but not likely conducive to an exchange of views. We need an orthogonal axis of conversation...


But first a little bit of a side remark: What if the topic is very *important* to me. 
Perhaps in that case I throw caution to the wind and make
some pointed comments. Very well; I have spoken. Now let's make a distinction between 
*engagement* and *emotional transaction*: 
In an emotional transaction I deal with a misguided view the world hands me by responding 
back in the heat of the moment. This might be an emotional outburst; nothing wrong with that. 
There are times when a good vent is called for. In contrast, the idea of engagement: 
Supposing I have the time; and I come in prepared; with a plan; and a calm mind: 
My hope is that here is an opportunity to open some doors, to broaden someone's view 
in addition to broadening my own.


Knowing about the choice between transaction and engagement is 'meta-perspective'.

### Professor Oreskes takes on the Royal Society


Around 1660, a new 'club' calling itself the Royal 
Society, centered in London, came up with the motto *Nullius in verba* 
('Don't take anybody's word for it').
To illustrate the biscuit: Mister So-and-so (back then) was regarded as an expert in
the natural history of swallows; so he publishes a monologue on where swallows
go in winter. It turns out that they swim down to the bottoms of ponds and hibernate until
spring. Another expert comes along with a rebuttal, 'No, no, of course
they hibernate on the moon.' Now the Royal Society can have its say: 
Read our new motto. Or if you like: Read between the lines of the motto: 
If you are unable to produce evidence that can be reliably reproduced: You have 
not carried the day in this important matter of swallow hibernation. 
While that emphasis on reproducibility in scientific discourse 
was pivotal to how science was done, it highlights a perpetual
social element of scientific discourse: How to treat subject matter authorities?


Today we sometimes ironically remark that science advances by death, meaning: 
The death of an intransigent elder scientist makes room for new ideas to flourish.
This is only part of the full picture of how science works, however. 
Scientific opinions are 
developed often through years of concentrated work. Often progress is gradual; 
sometimes progress proceeds in jumps in understanding; but by whatever
degrees progress is made there is always a self-examination process 
in its wake. 'Does this make sense? Does it fit together? What are 
the remaining and/or new questions? Are we missing something?'


Naomi Oreskes is a historian of science. Last I checked she was at Harvard. 
In 2004 she published an analysis of papers on climate change in the scientific literature.
Her paper, 'Beyond the Ivory Tower: The Scientific Consensus on Climate Change' (*Science*)
immediately produced a great wave of relief in the geoscience community because it 
controverted the disinformation campaign that 'the science isn't settled yet'. 
The consensus was overwhelming. Dr. Oreskes later wrote a book on authority in science
called 
[Why trust science?](https://press.princeton.edu/books/hardcover/9780691179001/why-trust-science).

- There is a balanced way of dealing with scientific consensus
    - Start with the premise that a scientist is legitimately looking to publish accurate conclusions.
    - In aggregate the literature represents a kind of consensus backed by time and careful thought
    - To controvert parts or all of a given consensus is possible; that is built into how we conduct science
        - But the circumstances for this to happen are difficult to achieve
            - A new explanation must account for the existing evidence
            - A new explanation must in particular not dispense with fact
            - New evidence is necessary supporting the new explanation
            - There is no utility in this process for unsupported statements of *belief*
        - This *difficult task* can be formulated as a 'smell test' for new explanations
    - Much more often, in fact the norm: New work builds upon and sharpens existing consensus
- This raises the question 'Why is there so much doubt about scientific consensus?'
    - 'Cynical actors undermine the truth.' -Naomi Oreskes
    - Anyone with a microphone and an agenda can use scientific consensus for their own ends
        - The idea here is that attention is all that matters in this forum
        - Graffiti on a boxcar gets very little attention
        - Graffiti on the walls of the Smithsonian or the Louvre gets attention



### Part 3: Rules of engagement


All other things being equal: Engage when prepared; therefore prepare!


#### 1. What the point is **not**

- We see engagement on satirical TV shows like The Daily Show calculated to get a laugh...
- The press amplifies situations where person A belittles person B or idea C
    - Example: James Inhofe belittled climate change on the senate floor, walking in with a snowball
- We are not here for that
    - Men go crazy in congregations but they only get better one by one
    
    

#### 2. Crossing motivations, the Alan Hamlet lesson


Alan Hamlet is a climate scientist. In about 2011 he attended a public forum in Skagit County
in person to explain the impacts of climate change on Skagit Valley. He was met with skepticism
as Skagit residents took turns at the microphone 
reading from disinformation flyers they had printed off the internet. 
They (the incensed readers) were there looking for someone who wanted to condemn them or take
away their freedom of choice or their tractors. Alan Hamlet was there to talk 
about how rising sea level was creating a kind of up-river pressure, a measurable creep of 
salt water up the Skagit River, closer and closer to the freshwater intakes that provide 
water for drinking and irrigation. He was there to explain how lower summer river flow and
higher temperatures were impacting salmon eggs in the upper Skagit.
Alan Hamlet did not drive for two hours to attend a town hall meeting so he could talk to farmers
about surrendering their guns and their tractors. 
He went to that meeting from his interest in promoting public awareness of the physical consequences
of climate change. 

#### 3. The nature of disagreement, the Tad Pfeffer lesson


At the University of Colorado Boulder some time around 2009 a glaciologist named Tad Pfeffer 
hosted a climate change forum. In the discussion that followed the speaker presentations the
discussion turned to 'winning the argument' by virtue of the facts and the truth being the
ultimate arbiters of the disagreement. 


Tad took the microphone and made the following observation for the benefit of the academics
in attendance: The point of science is to get at the truth, to get to the correct explanation. 
That's not how law works. In a court of law the point is to win the case, irrespective of what
the truth is. If you go into a court of law expecting to win because you are correct, you are
in for a sad disappointment.

#### 4. The shall Brock teach himself lesson


A co-worker of mine, one Brock Johnson, was interested in discussing and understanding
climate change. He had read that water vapor was a much more impactful greenhouse
gas than carbon dioxide; and yet all the focus of global warming centered on carbon dioxide. 
It seemed like the scientists were focused on the wrong gas. 
So I made the mistake of looking up the answer and explaining it to him. 
He was unsatisfied with my explanation and asked additional questions. 
After a few iterations of this I realized: Brock was not interested in understanding the 
situation. Brock was interested in having someone else do the work for him. 


I realized that intellectual laziness is not an actionable state of mind.
However: Having a conversation that subtly *addresses* intellectual laziness is possible. 
Of course one must avoid any pointed language. 'Insults rarely strengthen an argument' as
the saying goes. So let us say to Brock, 'Yeah, water vapor *is* a greenhouse gas, and you 
are right that it is *not* the critical factor in human-driven warming of the earth. Rather:
CO2 *is* responsible. The explanation is pretty simple but it isn't like five words long.
So let's do this: Suppose you do the reading as if you are going to explain it to me; 
and I'll do the reading to be prepared. Whenever you are done let's pick up this 
conversation. I'm dying to see how you will explain it!'


The purpose of the 'let's suppose' is to focus on the topic of Brock teaching himself 
about a scientific consensus. Maybe it works; maybe he just gives up on it. Point is
we have turned it into a non-loaded process and the door is open. 

#### 5. Normative and positive statements


We have logical fallacies and rhetorical speech; various devices intended to sway opinion.
Time for a crash course in these matters: Detecting them in the moment can help reframe a 
conversation. 

As an example of the way rhetorical speaking can insinuate itself into a conversation, consider
the task of proving there are no extraterrestrials currently visiting earth. Very difficult task; 
you can't disprove a negative. In the climate debate there is a tendency to mix positive and
normative statements; so let's take a moment on that. 


A normative statement is about the way things should be. 
A positive statement is about the way things are. 
'There are no swallows hibernating at the bottom of this lake' sounds like a negative; but as a 
factual remark it is a *positive statement*. 


How things should be is really a matter of understanding the way things are and then
prioritizing future outcomes; and making a plan for getting to those outcomes. So 
positive statements precede normative statements. 


Be on the lookout for a pivot from positive to normative. It is ok to call out 
normatives as not positives. As a starting point: Don't argue normatives. At least,
not until the positives are sorted out. 

#### 6. The rhetorical fulcrum


A fulcrum supports a board that can tip to one side or the other. The metaphor is meant to 
invoke how small changes in language -- particularly in agreeing on meaning -- can have big consequences. 


Let's suppose that language matters. Terminology matters. Some terms are pretty innocent looking... 
and yet they can have very inflammatory connotations. A big part of preparation is knowing what 
that lit fuse looks like. In a chat about climate change the term *climate* is frequently 
misconstrued. Climate refers to a running average view taken over several decades. 
Year-to-year we find that environmental observations such as temperature are noisy; but
a 35 year running average reduces this noise and shows the actual trend.


Practically speaking we have to be listening for signs of a misunderstanding. 
The moment we hear "the past ten years have shown
a marked downward trend in global temperature" we have the classic 
misunderstanding of climate. What to do? Simply ask for the microphone and
point out climate is an average over 35 years. Now celebrate that the conversation 
course-correction, talking about the same thing!


Let's look at two additional examples of oft-misunderstood language: 


- climate change consensus is based on observation
    - the consensus on human-driven climate change is based on physics
        - ...and not on computer models
        - the moment we hear "climate models disagree with one another"...
            - ...we have arrived at a classic misunderstanding
            - Climate models are predictive *only* in a rudimentary sense
            - Consider the earth as a complicated system (and don't forget to include the sun)
                - Imagine modeling the room one is in on a scale of one-centimeter cubes
                - Now ask 'How big is the computer needed to model the earth?'
            - Make a computer model: Is it correct? Answering this is a starting point for modeling
                - From here we go to a very exploratory, speculative process of improving models
                - Modeling is not the basis of *understanding* the nature of climate change
                - Modeling is a hopeful attempt to understand the potential severity of climate change in coming decades
            
- evolution is both a fact and a theory
    - evolution theory is supported by multiple independent lines of factual evidence
    - commonly objections to evolution invoke *belief*: 'I don't *believe* in evolution.'
    - Scientific theory is a matter of explanation supported by reason, observation and experiment
    - An alternative approach to understanding based in belief begs absurd questions:
        - Are other scientific theories subject to a person's state of belief?
            - If I choose to stop believing in...
                - ...electromagnetism: does my car fail to start?
                - ...quantum mechanics: does my television burst into flames?
                - ...gravity: can I fly?
                - ...biology: can I cure cancer?
        - Ok no... but evolution is *different* because my TV does not use it
            - ...or really it is different as a threat to my sense of identity
            - Very well, let us set up and ask a helpful question about *intent*
                - Science tells us that evolution is...
                    - ...simple at a high level, based on two simple observations
                    - ...complicated in how it gets the job done (molecular biology)
                - Consider the tens of thousands of individuals responsible for evolution theory
                    - They devoted years of their lives on their particular analysis...
                    - They did experiments, they built new equipment, they wrote papers, ...
                    - They persevered and gradually this increasingly complicated picture emerged...
                    - ...that explains the fundamental nature of life on earth
                - So speculate: Which explanation of this process seems more likely?
                    - Those scientists solved a vast, intricate, utterly fascinating puzzle
                    - Those scientists concocted a story so they could threaten your self-esteem
