# Infectious Disease Spread Model 


## ...with Vaccination


This description is from a CA...


This model simulates how an infectious disease spreads through a population over time, incorporating vaccination rates and proximity factors.


### Population Structure


The population is divided into compartments: 

- Susceptible (uninfected)
- Infected (contagious)
- Recovered (immune)
- Vaccinated

Individuals move between these states based on disease dynamics and 
interventions.


### Transmission Mechanism


Disease spreads when Susceptible individuals contact Infected individuals. 


Transmission rate depends on the basic reproduction number: 


- How many people one infected person typically infects
- Contact frequency
- Proximity factors
    - Closer physical proximity increases transmission probability.


### Vaccination Parameters


A fractional vaccination rate determines what portion of the Susceptible 
population receives vaccines over time. Vaccination can occur before exposure (preventive) or during the outbreak (reactive). 


Vaccine effectiveness reduces both infection probability and transmission 
rates for breakthrough cases.


### Proximity Modeling


Population density and mobility patterns affect contact rates. 


High-density areas (urban centers, workplaces, schools) have elevated 
transmission risks. 


Social distancing measures reduce effective proximity and contact frequency.


### Temporal Dynamics


The model tracks disease progression over time, showing infection peaks, 
vaccination impact, and population immunity buildup. It accounts for 
waning immunity, seasonal effects, and potential variant emergence 
that may affect vaccine effectiveness.


### Outcomes


Results show total infections, hospitalizations, deaths, and the effectiveness of different vaccination strategies and proximity interventions in controlling disease spread.
