```python
# Calculate pi (3.1415...)
#   If you read this code and it makes sense: You are in the right place. 
#   If it does not make sense: see the coaches to discuss.

pi_sum = 0.
this_sign = 1. 
nFractions = 17

for i in range(nFractions):
    pi_sum, this_sign = pi_sum + 4.0 * this_sign / float(2*i + 1), -this_sign
    
print '\npi ~=', pi_sum, '(', nFractions, 'fractions ); compare this with 3.141592653589...\n'   # end result
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import warnings; warnings.simplefilter('ignore')
from ipywidgets import *
from traitlets import dlink
import random as r

def PlotPolynomial(factor_choice, yAxisControl):
    db = ['Boom', 'Put Bunkin', 'Frump Frumpkin', 'Kid Pipkin', 'Allebasi', 'CWTSIHE']
    n = r.randint(0,5)
    description = 'Ok ' + db[n] + ':\n  y = '
    if factor_choice == 0: description += '1'

    incA = False
    incB = False
    incC = False
    incD = False
    if factor_choice & 1: 
        incA = True
        description += '(x - 4)'
        if factor_choice != 1: description += '*'
    if factor_choice & 2: 
        incB = True
        description += '(x - 3)'
        if factor_choice > 3: description += '*'
    if factor_choice & 4: 
        incC = True
        description += '(x - 2)'
        if factor_choice > 7: description += '*'
    if factor_choice & 8: 
        incD = True
        description += '(x - 1)'
        
    yPos = float(yAxisControl)/2.
    yNeg = -yPos
    
    x = np.arange(-1.0, 8.0, 0.01)
    y = []
    for xx in x:
        value = 1.0
        if incA: value *= (xx - 4.)
        if incB: value *= (xx - 3.)
        if incC: value *= (xx - 2.)
        if incD: value *= (xx - 1.)
        y.append(value)

    figdim = 9
    
    fig = plt.figure(1,figsize=(figdim,figdim))
    plt.xlim([-1., 8.])
    plt.ylim([yNeg, yPos])
    plt.scatter(x, y, s=8, color='r')
    plt.plot([0.,0.],[yNeg,yPos],color='k')
    plt.plot([-1.,8.],[0.,0.],color='k')
    for x in range(-1, 9):
        xtic = float(x) 
        ytic = float(yPos)/32.
        plt.plot([xtic, xtic], [-ytic, ytic], color='k')
    for y in range(int(yNeg), int(yPos+1)):
        xtic = 9./50. 
        ytic = float(y)
        plt.plot([-xtic, xtic], [ytic, ytic], color='k')
    plt.text(4.2, 0.5*yNeg, description, fontsize = '12')
    plt.show()

        
    
interact(PlotPolynomial, 
            factor_choice={'nothing': 0, \
                           'A': 1, \
                           'B': 2, \
                           'C': 4, \
                           'D': 8, \
                           'AB': 3, \
                           'AC': 5, \
                           'AD': 9, \
                           'BC': 6, \
                           'BD': 10, \
                           'CD': 12, \
                           'ABC': 7, \
                           'ABD': 11, \
                           'ACD': 13, \
                           'BCD': 14, \
                           'ABCD': 15 \
                        }, 
            yAxisControl=widgets.IntSlider(min=1,max=50,step=1,value=5, continuous_update=False),
        )
```
