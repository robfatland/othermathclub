# the other math club / python bytes


Question: Have you ever read a book about a *seemingly* ordinary world
when suddenly the rules change? Perhaps it was a book about Harry Potter...
for me it was books that featured animals who could speak English.
In the first of these I teetered on the edge of a rabbit hole and then tumbled over
to fall downward in slow motion past cupboards containing jars of marmalade. 
How peculiar! But I knew the author was on my side because we fell in slow
motion.


Consider finding oneself in a new world within a book. 
This new world is built mostly from echoes of the 
ordinary world together with bits of the extraordinary and amazing...
what we might call New Rules. *This material* is inspired by *that book*.


Now please note: When you set down *that book* you find yourself
back in the regular ordinary world: Rabbits do not check their watches and 
there are no trolls to turn to stone with the sunrise. 
In contrast: Here nothing resets to ordinary when
you set this book down. This book is about *extraordinary* things
but all of it is *real*. The first extraordinary thing is you, by 
the way. You are extraordinary for so many reasons. 
But the good news is there is more of the extraordinary to discover.


Let us begin with the story of a metal pole.


## The Fishing Kid must morph into a more solid example, literally... loh


Once upon a time there was a kid who wished to go fishing... but 
she did not have a fishing pole. So she took the bus to a stop near 
a bridge over a river and climbed down the bank to see if she could 
at least spot some fish.



Along the bank of the river she happened upon a five foot steel pipe and as
she did not have any better options she decided to assemble a makeshift fishing
pole using this metal pipe. 


She gathered up some string and made a hook from a 
paper clip and dug into a bag of cheese puffs for some bait. She had five dollars
in her pocket ... fishing ... no luck ... catch the bus home... walk to the 
bus stop... pass a post office. A sign out front 
advertised cubical moving boxes, so many feet on a side. A small one foot cubical 
moving box cost $\$1$, a two foot box cost $\$2$, and so on up to five feet. 


D'enigmes River


Never a bite from a fish, not even a nibble... and the cheese puffs would become
soggy and fall off the hook and drift away. 


The bus pulled ... driver ...'No no no, I'm sorry but you may not bring anything
on the bus longer than three feet. That metal pole is certainly five feet long.' 


She does not want to abandon her fishing pole... but the 
bus is the best way way home. The driver has a 15 minute layover at this stop.
What can she do?
