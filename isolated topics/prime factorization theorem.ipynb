# **Prime Factorization Theorem** 


> All positive integers > 1 can be factored into a **unique** set of prime numbers.


### Proof


We prove this by contradiction: We assume the theorem is *false* and then arrive at a contradiction.


If certain numbers can be factored as primes in at least two different ways then there is a smallest such number $N$.


The first factorization of $N$ will be the product of some primes designated $\alpha$ (with subscripts). The second 
factorization will be written in terms of some $\beta$ prime factors.


If the $\alpha$ set and the $\beta$ set have common elements then they can be canceled out to produce a *smaller*
number with two different factorizations. But $N$ is the smallest. So the $\alpha$ and $\beta$ factor sets share
no common elements. 


That's enough of a set-up. Now we follow through some arithmetic to arrive at a contradiction.
On the way we find new numbers less than $N$; and since $N$ is the smallest number with two unique prime 
factorizations these smaller numbers can have only one prime factorization.


$$N = \alpha_{0} \; \cdot \; \alpha_{1} \; \cdot \; \alpha_{2} \; ... \; \alpha_{j} = \alpha_{0}\; \cdot \; A$$
$$N = \beta_{0} \; \cdot \; \beta_{1} \; \cdot \; \beta_{2} \; ... \; \beta_{k} = \beta_{0}\; \cdot \; B$$


where we have already said that


$$
\lbrace \alpha \rbrace \;\;\; \bigcap \;\;\; \lbrace \beta \rbrace = \emptyset
$$


$N$ can't be even for if it *was* it would have $2$ as a factor and $2$ would show up in both 
the $\alpha$ and $\beta$ sets. (*Odd* times *odd* is *odd*.)


Therefore $N,\;\; \lbrace \alpha \rbrace, \;\; \lbrace \beta \rbrace, \;\; A$ 
and $B$ are all odd.


Let's assume $\;\alpha_{0} < \beta_{0}$. (If not: Swap the $\alpha$'s and $\beta$'s.)
Therefore $A > B$. Therefore for some $q > 0$ and some $r > 0$ we have $\beta_{0} = \alpha_{0} + 2q$ and $A = B + 2r$. 


This is worth re-stating because it is crucial! There are positive integers $q$ and $r$ which account
for the differences between $\alpha_0$ and $\beta_0$ and between $A$ and $B$. Those two differences
are even numbers: Everything is odd and the difference between two odds is even. And so the differences
(being even) contain a factor of $2$. In fact: $2\cdot q$ and $2 \cdot r$.


Now we continue:  


$$N = \alpha_{0} \;\cdot\;A = \alpha_{0}\; \cdot \; (B + 2 \cdot r) = \beta_{0} \; \cdot \; B = (\alpha_{0} + 2 \cdot q) \; \cdot \; B$$. 


Just focus on the third and fifth expressions in this equation sequence:


$$\alpha_{0} \cdot B + \alpha_{0} \cdot 2 \cdot r = \alpha_{0} \cdot B + 2 \cdot q \cdot B$$


We can eliminate the $\alpha_{0} \cdot B$ term from each side and divide out a $2$ to arrive at


$$r \; \cdot \; \alpha_{0} = q \; \cdot \; B$$


Both of these are less than $N$. Therefore there is only one way to factor both the left and the right
sides of this equation: The prime factor sets are identical.


As $r \cdot \alpha_{0}$ has but one prime factorization (it being less than $N$) 
and as $B$ does not include $\alpha_{0}$ in its factors (it is the product of $\beta$s): 
$q$ must be a multiple of $\alpha_{0}$. We write


$$\;\;q = s \cdot \alpha_{0}$$ 


for some number $s > 0$. 


Recall we began with $N = \alpha_{0} \cdot A = \beta_{0} \cdot B$ where $A$ and $B$ have no common factors. 
We chose the $\alpha$'s and $\beta$'s so that $A \; < \; B$ and this gives us the sequence: 


$$\alpha_0 A = (\alpha_0 + 2q)B$$


substituting for $q$ from


$$q = \alpha_0 s$$


to arrive at


$$\alpha_0 A = \alpha_0(1 + 2s)B$$

<br>

$$A = (1+2s)B$$


Now $A < N$ means that there is but one prime factorization for $A$, one that apparently includes $B$, but $A$
and $B$ share no common factors; so this is a contradiction. Our original assumption that some $N$ exists with
two *different* prime factorizations must be wrong. Therefore prime factorizations are unique. $\boxdot$
