# Sodium Project


## On the Recurring Theme of Distributed Information


> See the sad shire horses walking home beneath the sodium lights... <br>
> ...Men go crazy in congregations, they only get better one by one. <br>


The idea here is to ask 'What is a useful generalization from the utility of the Fourier transform?'
The first answer is that the Fourier transform is its own useful generalization.
In `Godel Escher Bach` the theme is (in part) self-reference and recursion... but in a throwaway
reference to Prufrock I claim Hofstadter felt his book was misinterpreted: 'That is 
not what I meant at all.' To avoid this pitfall let me suppose a generalizing concept is
distributed information, specifically information not distilled to a point in
the way that human cognition conceives of a dictionary. In the dictionary there is one and only one
entry for a hippopotamus; and then there is the language itself that gives us comparatives 
and context. Whereas if I overlay or stack transparent greyscale images of 10 animals:
The resulting blur has in each of its pixels superimposed information about each animal. 
Given just the composite image with no information about its construction...
we might infer some peculiar creatures. More commonly we see distributions of information under
very well-controlled conditions; so let's contrive a list of these as fodder for this 
exploratory project. In keeping with my theme, there may be overlap between these topics.
Also as a gesture towards appropriation: I am also going to include situations where
point-like or simplistic information proves to have unforseen complex behavior. In this way
we can include the logistics curve and other manifestations of chaos.

## Topic notes


### The 2 Prisoner Gold Key Problem

* Parity sets, exclusive-OR: addition is the same as subtraction
* There are many approaches to encoding; the clever one uses a calculated binary address
    * For the chessboard a natural address scheme is bitwise
        * Suppose tails is 1, heads is 0
        * The scheme uses six parity bits to produce a location pointer
        * A good name for this is a 'parity address'
        * Now we need to establish that 6 bits can be used to give a unique address on the board
        * 3 bits for horizontal / file (low-octal), 3 more for vertical / rank
        * Each parity bit corresponds to the parity count of a mask
        * Each mask corresponds to 32 squares
        * A given mask parity bit is only affected by coins inside the mask; the other 32 coins/squares are irrelevant
        * Define a mask product such that the masks are mutually orthogonal
        * Find how many 'makes no difference' squares there are
        * There is degeneracy between coin configurations: Many of them produce an identical parity address
            * There are 2^64 board configurations but only 2^6 parity addresses
        * The key's address xor the initial parity address yields a new address F
        * The solution to the problem is to show that a coin flip at F sets the board's parity address to the key address
* The notion of orthogonal functions
    * ...extend multiplication to be its own inverse (no carry)
