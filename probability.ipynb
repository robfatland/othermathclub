### Factorials


Zero factorial is written as $0!$ and it is equal to one: $0! = 1$.

One factorial is also equal to one: $1! = 1$.

Two factorial is equal to two: $2! = 2$. 

This is easy so far but now it becomes more complicated. 

Three factorial is six and four factorial is $24$. Here we go: 

$$3! \;=\;3\;\cdot\;2\;\cdot\;1\;\cdot\;=\;6$$

$$4! \;=\;4\;\cdot\;3\;\cdot\;2\;\cdot\;1\;\cdot\;=\;24$$

So there you have it. Just remember that $0!$ is $1$ and for all the other positive integers you just multiply 
all the numbers together from one up to that number. As you can imagine these results start getting pretty large.

### Combinations

We use the word *combination* to mean counting how many ways there are of choosing $b$ objects from a set 
of $a$ objects. Being familiar with this idea is surprisingly useful.
By the way: We do not care about what *order* these objects are placed in.
In the previous case with *factorial* we worried about order: First place, second place and so on. 
Now we are just looking at subsets. 
Let's say the objects in the subset are just placed
in a pile on the floor, with no order. 


The rule is presented first. If it makes sense: Don't bother with the examples. Just carry on 
to the next cell. If you would like to see some examples of how the rule works: Keep reading. 
That is what this part is for.


By the way (in case you skipped the section above): 
The expression $a!$ does not mean that the $a$ is really excited. It is a function called *factorial*
and it means *multiply all the numbers from $1$ to $a$ together*. So $4!$ is actually $4\cdot3\cdot2\cdot1=24$.


Counting combinations is often spoken as 'a choose b'.
What does this mean, a-choose-b?  a-choose-b means 'the number of ways 
of selecting $b$ objects from a set of $a$ objects'.  It is written $\binom{a}{b}$ and it is calculated using factorials
in a fraction like this:


$$\binom{a}{b}=\frac{a!}{b!\cdot (a-b)!}$$


a-choose-b is also written ${}^{a}C_{b}$ but we won't use that here.


Let's work an example:


$$\binom{9}{4}=\frac{9!}{4!\cdot (9-4)!}=\frac{9\cdot8\cdot7\cdot6\cdot5\cdot4\cdot3\cdot2\cdot1}{(4\cdot3\cdot2\cdot1)\cdot(5\cdot4\cdot3\cdot2\cdot1)}=126$$


That is, there are 126 ways of pulling 4 objects from a larger set of 9 objects. After I interrupt myself
we will look at some examples of how this works.

Again (write in notebook): *a-choose-b* is actually *the number of ways of dividing a set of

$$\binom{a}{b}=\binom{a}{a-b}$$


Then the long version... which is a proof.


$$\binom{a}{b}={}^aC_b=\frac{a!}{b!\cdot (a-b)!}=\frac{a!}{(a-(a-b))! \cdot (a-b)!}={}^aC_{a-b}=\binom{a}{a-b}.\;\;\boxdot$$

#### Counting ways

Suppose you have a pair of dice and you roll them and add the numbers. You will get 2 at a minimum and at
most 12. What does it mean to say that the probability of rolling a $7$ is greater than the probability 
of rolling a $5$? 


First solve this problem in your notebook; or write down your ideas. Then continue to read the solution below.

**Solution** Let's count the ways of getting $7$ and $5$: 

```
First die    Second die   Total
    1            6          7
    2            5          7
    3            4          7
    4            3          7
    5            2          7
    6            1          7
    
First die    Second die   Total
    1            4          5
    2            3          5
    3            2          5
    4            1          5
```

By counting possible ways we see that there are more ways to get a $7$ than there are to get a $5$.

<br>${\Huge \infty}$


In your notebook: Add up and write down all the ways of getting the numbers from $2$ to $12$.
The total number of outcomes should be $36$; and notice this is $6^2$. That is not a coincidence.
For every outcome of the first dice roll there are $6$ possible outcomes of the second dice
roll. The total number of outcomes is six sets of six.

#### Ways over total ways


If you wonder 'what is the probability of rolling a seven' you will need a consistent definition of 
probability. You might think about this for a moment if the idea is new to you. 

<br>${\Huge \infty}$


How would you define probability? (As always 'I'm not sure' is a perfectly reasonable answer.)


There are six ways of rolling a $7$ with two dice and $36$ total ways of rolling two dice. 
The probability of rolling a $7$ is defined to be the ratio of these two numbers or $7/36$.
The probability of rolling a $2$ is $1/36$ since there is only one way of rolling a $2$. 


In this way probability is always a number between zero and one, inclusive. The probability
of rolling a 17 on two dice is zero. Not a chance. The probability of rolling a number
between 2 and 12 inclusive is one. It's a sure thing. (We don't count when the dice go 
under the couch.) 

### The oneness of it all


Suppose you have an experiment where there are only two possible outcomes, A and B. You determine
that the probability of A is 0.7. What is the probability of B? After a little reflection you
will probably decide that this must be 0.3. This is correct *because* B is the only other possible
outcome. The sum of all the probabilities for all the possible outcomes equals $1$. This allows
you to ask 'What is the probability that one of those things happens?' and get the answer $1$.


This can be useful if you are having a difficult time figuring out the probability that something
happens. You can try asking 'What is the probability that it *doesn't* happen?' If you can figure
that out... and let's suppose it is $x$... then immediately you know the probability that it 
*does* happen must be $1-x$.
