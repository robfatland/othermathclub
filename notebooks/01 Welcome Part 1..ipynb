$\large{\;\;\infty\;\;}$ **Write a computer program that calculates $\pi$ using the formula given above.**


```python
# Calculate pi (3.1415...)
#   If you read this code and it makes sense: You are in the right place. 
#   If it does not make sense: see the coaches to discuss.

pi_sum = 0.
this_sign = 1. 
nFractions = 17

for i in range(nFractions):
    pi_sum, this_sign = pi_sum + 4.0 * this_sign / float(2*i + 1), -this_sign
    
print '\npi ~=', pi_sum, '(', nFractions, 'fractions ); compare this with 3.141592653589...\n'   # end result
```
