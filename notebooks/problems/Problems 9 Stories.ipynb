### Friendly introduction

This page features problems in the form of stories. The first collection are a type of problem that we could
call *pursuit* problems because they involve something moving in relation to something else. The first of these
is the famous *duck and wolf* problem which does not ask you the important question. So I will ask you the

### $\star\star$ Dawn to Dusk Climb


A girl stands at the bottom of a mountain (not Mount Meru) where there is just one trail 
to the summit. She begins at dawn and walks up the trail, arriving at the very top at dusk. She is tired
so she has a little food and goes to sleep. Notice we have no information about how fast or how slow she
goes, when she takes breaks, and so on. We just know she got to the top at dusk. The next morning she 
rises at dawn and takes the same trail back down the mountain, arriving where she started the day before 
at dusk. Dawn and dusk are at the same time of day on these two days by the way. 

Prove that at some point during her hike up and at some point during her hike down the following day she
was at the exact same spot on the trail at the same time of day.

**Note**: This problem is akin to many of the problems in the *Problem* pages in the following way:
The challenge of the problem is not so much about calculating something as it is about figuring out how 
to view the problem. It is like walking around a sculpture and finding the angle to look at it where it 
suddenly makes sense.


###  $\star\star\star\;$The duck and wolf problem

A certain duck can take wing and fly away to safety instantly provided she is standing on dry land; 
but she can not take off from water. Unfortunately she finds herself swimming in a circular pond where a wolf 
(who does not swim) is walking about the perimeter, hungry.
The wolf can and does run briskly about the pond four times faster than the 
swimming duck. That is: If the duck can swim one meter per second then the wolf can run four meters per second.
The exact numbers are not important; only the proportion of their speeds.


What strategy can the duck adopt to escape the wolf and fly away? 

### $\star\star\;$ Sue Snail indecisive

Sue Snail was feeling very indecisive one day... in this problem she will stop half way through any task and do something
else. In this way she reminds me of Egon who only travels half way to a paint can before painting a dot. (If you
are wondering who Egon is please see the subject track about **Monte Carlo**.) 

One morning Sue decides to go to the store which is one hundred meters away. Half way there 
(at 50 meters from her start location) she has a change of heart and
decides to turn around and go back home. But alas half way home she again stops. No more of this! She decides now
to return to the point where she last turned around, which you recall was the half-way point. 

Sue is currently 25 meters from where she began; and so she sets off for the 50 meter point, back towards
the store. Half way there (she is now 37.5 meters from where she started) she decides to turn around and go back 
to the 25 meter point. She continues in this fashion, each time walking half the distance that she walked on 
the previous stage. We will call these half-way excursions the *legs* of her journey. Notice the each leg
is half as long as the previous one... so eventually as they get smaller and smaller they become very close to zero
and she stops moving entirely. Then she takes a nap.

Where does Sue take her nap?

### Four travellers

Four travellers $A$, $B$, $C$, and $D$ move at constant (non-zero) speed in straight lines across an infinite flat plane. 
Their speeds, directions, and locations at any moment are arbitrary (they could be *any* speed and *any* direction).
Let's assume that all four of their paths are different. 
Now one more detail must be noted: The four travelers experience certain *meetings*. A *meeting*
is where two travelers arrive at the same point at the same time. Each continues on his or her way from there. 
Here are the meetings: $A$ meets $B$, $C$ and $D$. $B$ meets $A$ of course; and $B$ also meets $C$ and $D$. 

Problem: Does $C$ meet $D$? 

This problem is also stated at [a math website](http://cut-the-knot.org) like this: 

> Four roads on a plane, each a straight line, are in general position. 
> Along each road walks a traveler at a constant speed. Their speeds, however, 
> may not be the same. It's known that traveler #1 met with Travelers #2, #3, 
> and #4. #2, in turn, met #3 and #4 and, of course, #1. 
> Show that #3 and #4 have also met. 

### Four Bugs Puzzle

Four bugs named Alpher, Bethe, Gamow and Stan begin at four corners of a square of side $s$. They walk directly 
towards one another: Alpher towards Bethe, Bethe towards Gamow, Gamow towards Stan and Stan towards Alpher. 
They move at the same speed and they each turn as they walk so as to keep facing directly at the next bug. 
Finally they meet in the center of the square. How far did each one of them walk? How many times -- if any -- 
did they cross the diagonals of the original square?

### $\star\star\star\;$ The boring worm problem


Imagine a cube made from 27 smaller, identical wooden sub-cubes as in the figure above. 
These 27 small cubes are 3 x 3 x 3, resembling a Rubik's cube including a single cube at the center,
inside surrounded by the other 26.
A certain worm likes to chew or bore a path through wood of this sort; and so he begins a journey through the 
big cube by entering one of the small cubes from the outside. The worm goes from one sub-cube to another through
touching faces. It never goes through edges or corners of the sub-cubes. He continues 
until he has visited all 27 sub-cubes just once. He never emerges from the interior of the large cube. 
Is it possible for the worm to arrive at the center small cube last?


**Note** Ross Honsberger wrote that guessing is 'the heart and soul of mathematics' and indeed after some 
thought you may find yourself with a good guess as to the answer. However the real challenge of this problem
is to *prove* your answer is correct.



### $\star\star\star\;$ Triangular picture frames

Imagine you and your business partner (who happens to own a katana (if you do not know what a katana
is you should look it up or ask a friend of yours who likes Samurai comics))... where was I? Oh yes:
You and your partner decide to start a photography business. You specialize in pictures of
people printed out and placed in simple wooden frames. 


There is a problem however. Every time you leave your business partner alone with the photo printer he
throws the photo prints into the air and slices them with his katana. He claims he is practicing to become
a Samurai. While you are tempted to be angry you quickly realize that this is a tremendous 
business opportunity. The sliced up photos are triangular in shape and are still often recognizable... 
so you decide to advertise 'triangular pictures' as your specialty. Business starts booming as triangular 
pictures become all the rage.


Then you realize another problem: While pictures are cheap to print, picture *frames* are expensive.
However a local carpenter named Bradley throws thin wooden strips out of his window
(scrap wood that he does not need) and these strips are perfect for making picture frames. 
Bradley is fine with you collecting these wood strips provided you do not take more than you need. 
The strips vary in size from 50 centimeters at the longest all the way down to little slivers of wood
that have length zero. 


If you pick up three wooden sticks at random either they *may* make a triangle. 
For example if you pick up two very short sticks and one very long one you will not be able to 
construct a triangle. What is the probability that three random sticks form a triangle 
if each has a random length between sliver size (very very small) and 50 centimeters?


Please notice this is an idealization; and any triangle you make does not have extra 
wood sticking out: It must use the entire length of all three sticks.


This problem is discussed in both the **Geometry** and the **Probability** subject tracks.

### Hilbert Hotel Question


We have established that Professor Hilbert's hotel, with an infinite number of rooms, can host an
infinite number of guests. Using the intercom we can even assign them to new rooms so as to make
an infinite number of rooms available when new guests arrive. This question concerns one possible
strategy for room changes.


Suppose you list all the counting numbers 1, 2, 3, 4, 5, ... and you circle the primes, beginning 
with 2, then 3, then 5, then 7, then 11 and so on. Certainly all the even numbers larger than 2
are not circled; so there is still an infinite number of rooms available. 


Now consider all of the rooms that are powers of two: Rooms 2, 4, 8, 16, 32 and so on. 
Let's call these rooms 'power-of-two' rooms. How many such rooms are there? 

If the hotel is full at the moment (there are people in rooms 1, 2, 3, 4, ...)
what would you say on the intercom to assign all of these people to just the power-of-two rooms? 
In other words: What is the recipe for my new room number if my current room is room $n$?

Suppose that rather than assigning everyone to power-of-two rooms you decide to assign them to
power-of-seven rooms: Rooms 7, 49, 343 and so on. What is the new room recipe here? Will any of the 
power-of-seven rooms also happen to be power-of-two rooms?


Suppose that rather than assigning everyone to power-of-seven rooms you pick some other prime
number $p$ and assign everyone in rooms 1, 2, 3, 4, 5, 6, ... to the power-of-$p$ rooms. 
If $p$ is not 7 and not 2: Are there any power-of-$p$ rooms in common with power-of-two rooms?
Or with power-of-7 rooms? 


How many prime numbers are there?  


The last question is this: Suppose that the hotel starts out empty. If you were informed that 
a never-ending stream of busses, each with an infinite number of passengers was about to start
arriving: What sort of strategy could you use to assign everyone a room number so that you
would never have to use the intercom to move people around?

### The Modified Clock

A mischievous student named Adam has cut short the minute hand of a working clock so that it looks identical to the hour hand. (This is a standard 12-hour clock with hands; not a digital clock.) Emily notices that nevertheless when she looks at the clock she can still generally figure out what time it is. For example at 2:15 she looks at the clock and sees that one of the identical hands is a bit past the 2 and the other hand is right at the 3. She thinks as follows: 

> **These two hand positions are consistent with 2:15 so the hand pointing to 3 must be the minute hand. It is not 
> possible for the other interpretation: If the hand pointing to 3 was the hour hand then the minute hand would be 
> pointing straight up to 12, not where it currently is a little past the 2.**

Along comes Meghna; and after thinking for a moment she says to Emily 'Yes you are correct, it must be 2:15... but I wonder if there are other times where it is impossible to tell what o'clock it is.'

Indeed there are such times; so this problem is: Determine *how many* such times there are in a 12 hour span where 
you can not be certain of the time. This is a moderately difficult problem. Of course the solution does not involve 
things like looking at other clocks!
