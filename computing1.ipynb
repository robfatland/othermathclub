### The Progression of Numbers


Remember when you learned to count? One day they showed you a picture of a sheep 
and they asked 'How many cows are there?' You looked up and said 'That's a *sheep*
not a *cow*.' They said 'Surprise! It is Zero cows!'


So they introduced a new number 'zero' by leveraging your idea of 
'all gone'; and over time this pattern of introducing new sorts of 
numbers repeated. 
One day it was fractions, another day it was negative numbers... and 
decimals came in at some point... Perhaps *irrational* numbers? 
*Algebraic* numbers? *Transcendental* numbers? 
It seems never to end.


Now here is a curious thing: Numbers can also be defined in terms of 
following rules. For this it helps to have a notion of sets: A set is
a named collection of distinct elements, like $S = \{ block, \ sphere, \ prism \}$. 
An operation called addition is defined that produces an element of $S$ from any
two elements of $S$. Let's call this the **ADD** operation:


- *block* **ADD** *block* is *block*
- *block* **ADD** *sphere* is *sphere*
- *block* **ADD** *prism* is *prism*
- *sphere* **ADD** *block* is *block*
- *sphere* **ADD** *sphere* is *block*
- *sphere* **ADD** *prism* is *block*
- *prism* **ADD** *block* is *prism*
- *prism* **ADD** *sphere* is *block*
- *prism* **ADD** *prism* is *sphere*



In this case you might point out that the elements in $S$ are not actually
what their name suggests (three-dimensional objects). Rather they are behaving
like $0$, $1$, and $2$ and the **ADD** procedure is just to add those two 
numbers and then reduce the result using $\mod{3}$. This kind of blows the 
cover off the whole charade... but the idea is still valid: We might come up 
with other '*entities*' that we can place in sets that behave like numbers; so 
the ideas we have about numbers are transferrable to these new sets. Not just
adding (what we called **ADD**) but also subtracting and multiplying and dividing
all come into play. 


Another example: Define a *transporter* as something that transports you from Point **A**
to Point **B**. Now both a *car* and a *bike* are *transporters*. As is a 
*camel* and a *magic carpet*. I imagine you are thinking about what distinguishes them.


Defining a number as 'something we can do arithmetic with' is a way of turning the 
idea of a number upside down. 


Suppose we invent a new type of number that follows the rules of arithmetic. 
Aside from being interesting (finding out how it works) this path leads directly
to curious structures that look like themselves, called fractals.

### How Computing Works

Let's start by baking a cake. We know we will need some flour and some sugar... is that it? We better look up 
a recipe. Then note down what we don't have. Then go to the store and buy those ingredients. Then come back.
Then tidy up the kitchen. Then get out the bowls and the mixer. We know we will need those. And then we will 
return to the recipe. 

Oh look: The recipe calls for three eggs. We better buy some eggs because we do not have any on hand. 

(some time passes... we do all these things... now we are going through the recipe)

Ok now it is time to add the eggs... well this is very curious. The directions are as follows: 


- write the number 1 on a little post-it note
- place the little post-it note just below this line

<br><br>


- break open an egg and add it to the cake batter
- mentally add one to the number on the post-it note above; call this the New Number
- if the New Number is 4 skip the next instruction
- go back up to the post-it note, scratch out the old number, write in the New Number and continue with the instructions from there
- measure out 1/3 cup vegetable oil and add that to the batter


Now let's do the same thing using a computer program. Just like with the cake there is a bunch of stuff 
we have to do before we can start making the program. And then this program just *pretends* to make the cake. 
Well... you could make a real cake and bring it to a club meeting and then we would be in business.

But for now here is the program. 
Think of it as a recipe that proceeds from the top instruction to the bottom but -- just 
like with the post-it-note -- it can jump backwards to do something (add an egg) over and over.

```python
how_many_eggs = 3

# This is a comment. When the program runs it does not do anything. It is for people to read.

print ''

post_it_note = 1

# Now I have a bucket labeled 'post_it_note'. Inside the bucket is the number 1
# That is: I need to add egg number 1; and I am tracking this in my post_it_note bucket.

print 'I just wrote down the number 1 on my egg-counter post-it note'
print ''

adding_eggs_now = True

# 'while' will do everything indented below as long as 'adding_eggs_now' is True

while adding_eggs_now:
    print 'I just added egg number', post_it_note, 'to the batter'
    if post_it_note == how_many_eggs:
        adding_eggs_now = False
        print '  ... ok I added enough eggs, better stop!'
        print ''
    post_it_note = post_it_note + 1

# And now we continue onward
print 'I just added 1/3 cup vegetable oil to the batter'
print ''
```

### Why this is remarkable

The program above pretends to make a cake... but we are no closer to having any cake to eat. 


The program does an action over and over again without getting tired.


The first line of the program sets the number of eggs. Change this to 7 eggs 
to be sure it works as you expect. Try 1,000 eggs.


People who use programs are Users.


People who code programs are Coders.

```python
import numpy as np, matplotlib.pyplot as plt
%matplotlib inline

def prime(n):
    if n < 2: return False
    floor_sqrt = int(np.floor(np.sqrt(float(n))))
    for i in range(2, floor_sqrt + 1):
        if n % i == 0: return False
    return True

start_value, square_side, dotsize = 41, 41, 4
n_numbers                         = square_side * square_side
numberRange                       = range(start_value, start_value + n_numbers)
(scale_x, scale_y)                = (1.0/float(square_side), 1.0/float(square_side))
(x0, y0), (x_current, y_current)  = (0., 0.), (x0, y0)
x, y                              = [], []
dxdy                              = [(scale_x,0),(0,scale_y),(-scale_x,0),(0,-scale_y)]
dxdyIndex, dxdyModulus, dxdyRun   = 0, 4, 1
dxdyCounter, dxdyBit              = 0, 0

# return coordinates based on a = cell (x) and b = row (y)
fig, ax = plt.subplots(figsize=(12,12))
ax.axis('equal')
ax.set(xlim=(-0.54, 0.54), ylim=(-0.54,0.54))

for i in numberRange:
    if prime(i):
        x.append(x_current)
        y.append(y_current)
        plt.text(x_current, y_current, str(i), fontsize='8')
    x_current += dxdy[dxdyIndex][0]
    y_current += dxdy[dxdyIndex][1]
    dxdyCounter += 1
    if dxdyCounter >= dxdyRun:
        dxdyCounter = 0
        dxdyIndex = (dxdyIndex + 1) % dxdyModulus
        dxdyBit += 1
        if dxdyBit > 1:
            dxdyBit = 0
            dxdyRun += 1

plt.scatter(x, y, s=dotsize, marker='.', color='r')
plt.show()
```
