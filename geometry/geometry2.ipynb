The following program shows roughly what Egon's painting looks like; with some sliders for you to
change. 

```python
import numpy as np
from ipywidgets import *
from traitlets import dlink
import matplotlib.pyplot as plt
%matplotlib inline

# iA = incident angle, dA = branch angle
def IterateTree(ax, vx, vy, callerDist, branch, iA, dA, callerDepth, maxDepth, fractalScale = 2.0):
    mycolors = ['r','b','g','k','magenta','cyan']
    colMod = len(mycolors)
    # establish how deep we are in the recursion
    myDepth = callerDepth + 1
    myDist = callerDist / fractalScale
    
    # establish the first angle a0
    a0Factor = float(branch-1)/2.
    a0 = iA - a0Factor*dA

    for thisBranch in range(branch):
        angle = a0 + float(thisBranch)*dA
        vxP = vx + myDist * np.cos(angle)
        vyP = vy + myDist * np.sin(angle)
        ax.plot([vx, vxP],[vy, vyP], color=mycolors[callerDepth%colMod], linewidth=1)
        if myDepth < maxDepth:
            IterateTree(ax, vxP, vyP, myDist, branch, angle, dA, myDepth, maxDepth, fractalScale)
    return

def DrawSymmetryTree(angle, depth, branch):
    fig, ax = plt.subplots(figsize=(12,10))
    ax.axis('equal')
    ax.set(xlim=(-0.1, 1.4), ylim=(-0.1,1.4))
    # ax.get_xaxis().set_visible(False)
    # ax.get_yaxis().set_visible(False)
    if (depth > 4 and branch > 6) or (depth > 6 and branch > 4):
        plt.text(0.2, 0.2, 'this plot will take too long; try depth = 4', color='k')
    else:
        vx = 0.65
        vy = 0.5
        startDistance = 0.5*np.sqrt(2.)
        incidentAngle = np.pi/2.
        deltaAngle = float(angle)*np.pi/180.
        myDepth = 0
        IterateTree(ax, vx, vy, startDistance, branch, incidentAngle, deltaAngle, myDepth, depth, 2.0)
        plt.show()
    
interact(DrawSymmetryTree, angle=widgets.IntSlider(min=10,max=120,step=1,value=25,continuous_update=False),\
                           depth=widgets.IntSlider(min=1,max=8,step=1,value=6,continuous_update=False),\
                           branch=widgets.IntSlider(min=1,max=8,step=1,value=3,continuous_update=False))
```

### Let's have some $\pi$

You have heard about $\pi$, $3.14...$, the mysterious ratio of the circumference to the diameter of a
circle.  The digits of $\pi$ are an infinite, never repeating... Here are a few more digits:


$$3.14159265358979323846264338327950288419716939937510582097494459230781640628620899862803482534211...$$


Over centuries different methods of calculating $\pi$ have been discovered.  We can compare the results 
from each new method to verify they agree. This way uses multiplication of fractions:


$\frac{\pi}{2}=\frac{2}{1}\cdot\frac{2}{3}\cdot\frac{4}{3}\cdot\frac{4}{5}\cdot\frac{6}{5}\cdot\frac{6}{7}\cdot\frac{8}{7}\cdot\; \cdots$

Another uses addition of fractions:


$\frac{\;\pi^2}{6}=\frac{1}{1}+\frac{1}{4}+\frac{1}{9}+\frac{1}{16}+\frac{1}{25}+\frac{1}{36}+\;\cdots$


$\pi$ is approximated in track C (Monte Carlo) using random numbers; and there are many other approaches
besides. Before continuing on try calculating $\frac{\pi}{2}$ or $\frac{\;\pi^2}{6}$ by hand and see how close this gets
you to $\pi$. 

### Calculate $\pi$ as a product of fractions

The computer will use that many fractions for its calculation. If you make **how_many_fractions** larger 
and re-run the cell: Does the estimate of $\pi$ get better?  You do this by double-clicking on the cell
and changing the '100' to a larger number. Then re-run the cell.


Again we are calculating $\pi$ from $\frac{\pi}{2}=\frac{2}{1}\cdot\frac{2}{3}\cdot\frac{4}{3}\cdot\frac{4}{5}\cdot\frac{6}{5}\cdot\frac{6}{7}\cdot\frac{8}{7}\cdot\frac{8}{9}\cdot\cdot\cdot$.

```python
import numpy as np
from ipywidgets import *
from traitlets import dlink

def piEstimate(how_many_fractions):
    my_pi = 2.0
    for i in range(how_many_fractions): my_pi *= (2. + 2.*(i/2)) / (((i+1)/2)*2.+1.)
    print('With %d fractions pi is about %10.9f, off by %10.9f' % (how_many_fractions, my_pi, np.pi - my_pi))

interact(piEstimate, how_many_fractions=widgets.IntSlider(min=1,max=10000,step=1,value=1))
```

As you can see this method is not very effective. With 517 fractions we are accurate only to
about one part in one thousand. That's a lot of arithmetic for three digits of an infinite-digit number. 


### Suppose we calculate $\pi$ as an infinite sum


Here is another method for approximating $\pi$


$$\frac{\;\; \pi^2}{6}=\frac{1}{1^2} + \frac{1}{2^2} + \frac{1}{3^2} + \frac{1}{4^2} + \frac{1}{5^2} + \; \cdots$$


This is from Leonhard Euler's solution to a mathematics problem called the **Basel Problem**.
Euler is a mathematician and Basel is a city in Switzerland. Euler solved the Basel Problem at age 28.
The problem had gone unsolved for 90 years prior to this.  


As before we have a formula involving $\pi$ on one side of the equals sign and fractions that continue forever. 
This time the fractions are added, not multiplied. The denominators will get bigger and bigger: 
1, 4, 9, 16, 25, and so on. An infinite number of increasingly small pieces add together; we call this our
sum. 


Once we add up those fractions we multiply that by six and take the square root to approximate $\pi$. 
Do you see why? Will this method be faster at approximating $\pi$ than the one above?  Make a guess, 
then try it out.


```python
import numpy as np
from ipywidgets import *
from traitlets import dlink

def piEstimateBasel(how_many_fractions):
    theSum = 0.0
    for i in range(how_many_fractions): theSum += 1.0 / np.power(float(i+1), 2.0)
    my_estimate = np.sqrt(theSum*6.0)
    print('Basel sum pi with %d fractions gives us pi as about %13.12f, off by %13.12f' \
          % (how_many_fractions, my_estimate, np.pi - my_estimate))
    
interact(piEstimateBasel, how_many_fractions=widgets.IntSlider(min=1,max=10000,step=1,value=1))
```

Well... this is just a bit better. For 517 fractions the error is about half as big as the previous method.


On the positive side: Both methods get closer and closer to $\pi$ as we include more fractions... so they *seem*
to have the right idea.

### Extra: A complicated construction


If you enjoy working with straight edge and compass to create figures: Here are instructions for a new figure to try.
Please share your results at OMC if at all possible.


First terms:


- **regular hexagon** has all sides the same length
- **equilateral triangle** has all sides the same length


Second a simple construction: Creating an equilateral triangle from two points.


- Draw a line connecting the two points
- Set the compass equal to the distance between the two points
- Place the pivot pin on one point and draw an arc
- Move the pivot pin to the second point and draw an arc that overlaps
- Draw segments from each of the two points to the point where the arcs overlap
- This should be an equilateral triangle


Third: Divide a segment in half


- Place the pivot pin on one end of the segment
- Draw a large arc that crosses the segment
- Place the pivot pin on the other end of the segment
- Draw a second large arc that crosses the first arc in two places
- Draw a line between the two crossing points of the two arcs
- This line should cut the segment into two equal halves


Now for the complicated construction

- At the center of a large sheet of paper draw a medium-large triangle of any shape
  - The idea is that this construction of a *regular* hexagon comes from *any* shape of starting triangle
- Mark the center points of the sides of your triangle (see above about dividing a segment in half)
- Draw a long line that passes through the center point of one side and the opposite vertex
- Do likewise for the remaining two sides
- Let's call these the crossing lines
- Where these three points intersect is called the *centroid* of the triangle
  - If your three lines do not meet at one point you may want to try again... they should
- Set your compass width to the distance from centroid to one vertex (corner) of the triangle
- Place your pivot pin on the centroid and draw a circle
- Repeat these two steps for the other two vertices of the triangle
  - You should now have three circles centered upon the centroid
  - Let's call these circles the vertex circles
  - Each vertex circle corresponds to a crossing line that goes through that vertex
  - Therefore each crossing line crosses its vertex circle in two places
  - Since there are three crossing lines there are six such crossing points
  - Let's call these the six crossing points
- Connect the six crossing points in sequence around the figure to create a hexagon
  - This hexagon is not *regular* unless you started out with an equilateral triangle
  - It probably looks more like a potato
  - Let's call the six segments of this hexagon the potato segments
- Make equilateral triangles from all six potato segments with the points oriented outward away from the starting triangle
  - All six equilateral triangles have two vertices on the potato hexagon and one vertex outwards from the potato hexagon
  - Let's call these the hex points
- Connect the six hex points in sequence with line segments

Does your hexagon look like a regular hexagon? One way to check it is to set your compass to the distance
between two adjacent vertices of your big hexagon. You can move your pivot pin to each of the six vertices
in turn and see if the distance to the two adjacent vertices is the same all the way around. If you got
close then congratulations, you are very precise with your construction. If like me there is some variation
in the distances between your vertices of your big hexagon: We have the consolation prize of getting to try
again if we are so inclined.

```python
# Calculate pi by adding up little squares
# Let's fill in the first quadrant with little squares; and if they are on or in the unit circle we will sum them to estimate the area. 

###########
###########
##
## Change the value of 'size_of_square' to see how it affects the approximation of pi.
## If you make this value very small the program will take longer to complete its calculation.
##
size_of_square = 0.01
##
###########
###########


# count up little squares to get pi. Those squares must be inside the unit circle, that is, 
#   the circle at the origin with a radius of 1. 

# We will loop over little boxes at location (x, y).  
# x ranges from 0 to 1 and y ranges from 0 to 1. The total area of this box is what? 
# We only count boxes in our total if they are at a distance of 1 or less from the origin.
# How do we calculate that distance (what is the name of the formula)?
# Once we are done we have looked at only one quadrant of the unit circle.
# Therefore we multiply the answer by what? 

import numpy as np


how_many_squares = int(1.0 / size_of_square)

print('There are', how_many_squares, 'little squares on one side of the big square.')

area_of_little_square = size_of_square * size_of_square

print('The little squares have an area of', area_of_little_square, 'square units.')
print('   (a number like 1e-06 is read "one times ten to the minus six" or "one millionth")')

# area is a bucket of variable that will increase when we find little squares inside the circle
area = 0.0

for x in np.arange(0.0, 1.0, size_of_square):
    for y in np.arange(0.0, 1.0, size_of_square):
        dist = np.sqrt(x*x+y*y)
        if dist < 1.0:
            area = area + area_of_little_square

# area is just in the first quadrant; so let's multiply it by four to account for all 4 quadrants
area = area * 4

print('Our estimate of pi is ', area)

# some kilroy matters
# It would be nice to create a little drawing of this
# It would be nice to do the area and volume of a sphere as well
```
