### Kongisberg Bridges Problem


The problem given above about bridges comes from a map of the city of Konigsberg, back in the 1700s. 
Euler solved it but pretended that it was not really a mathematical problem. This is because topology
and graph theory had not been invented yet... but after he was done talking they were. 


In what follows: We restate the problem, give the answer, and then give a short version of Euler's solution.

<br><br>${\Huge \infty}$

Is there a path through the town of Konigsberg (map at top of this page) that crosses each bridge only once? 

<br><br><br>
**Solution**: No.


<br><br>${\Huge \infty}$


Why is there no such path? That is: See if you can come up with a proof. 


**Euler's Solution** Notice that there are four pieces of land in the problem, each with an odd number 
of bridges connecting it with other pieces of land. Call these pieces of land $A$, $B$, $C$, and $D$. 
$A$ is the area at the lower left. $B$ is the island. $C$ is the large area at the top. 
$D$ is the area on the right. All are separated by the blue water; and then there 
are the seven bridges in red. 


A bridge crossing between two pieces of land can be described using two consecutive letters. If I cross a bridge
from $A$ to $B$ I can write $AB$. If I go from $A$ to $B$ to $C$ to $D$ and back to $A$ that can be written
in five letters as $ABCDA$. With five letters we describe crossing four bridges. 
Make sure this makes sense to you before continuing.


Suppose there is a solution. That is: Imagine that some path crosses all seven bridges once. 
This path will cross a sequence of bridges and can be described using the letters $A$, $B$, $C$ and $D$. 
Since there are seven bridges this solution would be written using 8 letters. 

Notice that land area $A$ has three bridges so the letter A must appear twice in this solution. For 
example $ABAD$ would cross all three of these bridges once as part of the solution; and $A$ appears two
times in $ABAD$.


<br><br>${\Huge \infty}$


Write in your notebook why the letter $B$ (for island $B$) must appear three times in a solution path.


Regions $A$, $C$ and $D$ must appear twice each in a solution. $B$ must appear three times in a solution.
That is: the various pieces of land must appear 2 + 3 + 2 + 2 = 9 times; so there must be 9 letters 
in the solution. However earlier we realized that a solution must be 8 letters long... so a 9-letter 
solution must have 8 letters... so there is no solution.

### Eulerian graphs

If you are not familiar with graphs please see the Notebooks on this subject. The first is **1 G Graphs** and they 
continue from there. We use the terms *vertices* and *edges* here to extend the Konigsberg bridges problem. 


A *walk* through a graph begins at some vertex and ends at another; and it spells out the edges along
the way. A *closed walk* ends back where it begins. For the Konigsberg bridges we wanted to know if 
we could cross each bridge once. This can be viewed as a graph problem where the four regions $A$, $B$, $C$ and $D$
are represented as *vertices* and the bridges are *edges* connecting them. 


We can present any graph and ask the same problem: Is there a walk that crosses all the edges just once?
We can make it more difficult by requiring that we wind up back where we started.
