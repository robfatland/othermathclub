To be honest this claim seems wobbly... let's see what happens when I try to prove it. 


We know $n^2$ is even. This means that $2$ is one of the 
prime factors of $n^2$. Also we know that $n$ is either a prime or the
product of some primes; so $n=p_1\cdot p_2\cdot p_3 \cdot p_4 \cdots p_i$. The $p$'s are primes.


Notice two things, please: First we *imagine* a prime number $p$ but it might be one of
several so we give it a little subscript, the number $1$, so it looks like $p_1$. This is multiplied
by $p_2$, and so on. The notation allows for a whole bunch of primes multiplied together
to arrive at $n$. But $n$ might be a prime number. If that is the case then $n=p_1$
and there are no other primes involved. Writing in this way expresses the abstract
nature of $n$ (we don't know how many prime factors it has) together with the definite
nature of $n$ (it is *most definitely* prime or the product of primes).


So now to continue:


$n^2\;=\;2\cdot q=n\cdot n=p_1\cdot p_2\cdot p_3\cdot p_4\cdots p_i\cdot p_1\cdot p_2\cdot p_3\cdot p_4\cdots p_i$


In words: One of the prime factors of $n^2$ is $2$ because $n^2$ is even. 
Since $n^2$ is $n$ times $n$ we know all the prime factors of $n$ appear 
twice in the prime factorization of $n^2$.
Also $n^2$ has only *one* prime factorization. (The order of the factors can change
but the list of factors may not.) Therefore only the prime factors of 
$n$ appear in the prime factorization of $n^2$. Since $2$ is one of those prime
factors it follows that $2$ is a prime
factor of $n$ and so $n$ must be even. 

My claim is wrong. If $n^2$ is even then $n$ must also be even.


We get a bonus result from this as well: Even perfect squares must be multiples
of $4$ since an even number squared must contribute two prime factors of $2$.

### Lots of primes

Now we have arrived at our first 'serious' proof. 'Serious'

**Claim:** There is an infinite number of prime numbers.


**Short Proof**

Assume the contrary, that the number of primes is finite. List them and form the product
of all of them, then add one. The result is a number that is not divisible by any of 
the primes on the list. It is either a new prime or the product of new primes not on the 
list. This contradicts the assumption that the number of primes is finite. Therefore the
number of primes is infinite. $\boxdot$


**Long Proof**


This proof is the same as the one above with more explanation. 


Assume there is a *finite* number of prime numbers, say $n$ of them. 


The list of them is $\;p_1, p_2, \dots, p_n$.
For example $p_1=2$, $p_2=3$ and $p_{10}=29$. 


Multiply all of these prime numbers together. Call the product $A$. Then add 1 to $A$. Call this $B$.


$$A = p_1 \cdot p_2 \cdot p_3 \cdots  p_n$$


$$B = A + 1$$


Now $A$ is divisible by all of the prime numbers. By adding one to $A$ we see that $B$ is 
*not* divisible by *any* of the prime numbers. 


$B$ is either a *new* prime not in the list or it is composite (the product of two or more prime numbers).
If it is the first case, that $B$ is a new prime: Then we contradict our assumption that we had the complete
list of all of the primes. 


If it is the second case, that $B$ is the product of two or more prime
numbers then *they* must not be on the complete list of primes. In either case we have discovered new
prime numbers that were not on our list of all of the primes. 


At this moment we can go in two different directions. One is to say 'The assumption that there is
a finite list of primes produced a contradiction: A number not on the list that is also prime.
Therefore the assumption is false and the list of primes is infinite.'


The second direction is to say 'Well $B$ creates at least one new prime number so let's add that
to the list of all of the primes.' The problem with this is it puts us back in the same position
as before. We can begin again with the new list to make a new version of $B$ -- call it $B_2$ -- 
which will produce still more new prime numbers that are not on our list. And so on: 
The process can be repeated as many times as 
we like. And so again there is no limit to the number of prime numbers. 


That is a proof that there is an infinitude of primes; so we generally end with a little symbol
indicating the proof is done. $\boxdot$

This by the way is called a proof-by-contradiction. We assume that the opposite of our claim is true and 
we show that this assumption leads to an impossible situation: A logical contradiction that tells us our
assumption was wrong. Therefore we go back and say that the *opposite* of our assumption must be correct.


> **Claim**: The number of primes is infinite<br>
> **Assumption**: The number of primes is finite<br>
> **Contradiction**: The assumption leads via logic to an impossible situation (discovery of new primes)<br>
> **Result**: The assumption is false<br>
> **Conclusion**: The opposite of the assumption must be true: The **claim** is proven
