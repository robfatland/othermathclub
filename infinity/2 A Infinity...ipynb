### Friendly introduction


On this page we will take two new approaches to the idea of infinity. 
The first idea has to do with counting... and it really comes down to the idea that infinity is
countable (to infinity) but pieces of infinity are also countable (to infinity). 


Wait... what?


Suppose that you and a friend each started counting numbers up to infinity. Your idea is to find
out what infinity divided by two is equal to. You hatch a plan where *you* count the *even* 
numbers and *she* counts the *odd* ones. 


By the time you get up to $100$ you have counted $50$ numbers and so has she.
Clearly by the time you get to $N$ you each have $N/2$ but since $N$ increases without limit...
neither one of you ever runs out of numbers to count, even or odd. 


You each have an unbounded number of evens and odds to count... so you both decide that 
infinity divided by two is infinity. You write the following equation in your notebook:


$${\large \frac{\infty}{2}\;=\;\infty}$$


You know a little algebra so you multiply both sides by two and write down this result as well:


$${\large \infty \; \times \;2\;=\;\infty}$$


Oh dear. Ok so we get it: Infinity does not behave like a number. 
It is definitely one level stranger than numbers.

This means that infinity as an idea is ok to work with. But infinity is not a number in the sense
that we are used to. Numbers when you divide them by two produce new numbers, not themselves.


**Therefore we will avoid putting the infinity symbol $\infty$ in equations with equal-signs**


Let's plan to keep using our *idea* of infinity without trying to make it be a number. 
We shall be careful about comparing *numbers* to *infinity*.
We can say $7 > 3$ but not $\infty > 7$ because infinity is
not a number. Is this clear? Muddy? A good idea? A bad one?


Mathematical thinking requires us to be very precise in how we use and understand language.


What else? Well I said there are two ideas on this page. The first is the counting idea...
and the second idea has to do with randomness and information; and we will borrow the number $\pi$ for this. 


To get started on thinking about $\pi$: You may have heard that the digits of $\pi$ begin $3.1415\dots$ and
continue for quite a long ways, in fact forever. This is entirely true and those digits seem to occur at
random... even though nobody is rolling dice to determine what they are. This leads to a paradox of sorts 
and an interesting thought: If the digits of $\pi$ are random then would it be possible to find 400 nines
in a row somewhere???


### Terms

- **infinity**: A set of ideas connected with being *unbounded* or *without limit*.
- **inverse**: The number needed to multiply by to arrive at $1$. The inverse of $2$ is $1/2$.
- **reciprocal**: same as inverse; the reciprocal of $x$ is $1/x$
- countable: A set whose elements can be assigned matching integers $1,\;2,\;3,\;\dots$

### Finding a sequence of digits in ${\large \pi}$


$\pi$ is irrational. Does this mean you can not argue with $\pi$? No, matters are much worse than this. 


*Irrational* means that $\pi$ can not be written as a fraction of integers: Integer numerator, integer denominator. 
This in turn means that the digits of $\pi$ never do what the digits of rational fractions do: They repeat forever. 
As an example calculate $1/7$ as a decimal to detect the repeating digits. 
$\pi$ continues forever... its digits are countable and infinite... and so there is a paradox. 
(A paradox is two views of something that both seem true but that seem to disagree at the same time.)


  - There is only one correct version of $\pi$
  - The digits of $\pi$ have nothing to do with one another... they appear to be random


It is as if the universe rolled an infinite number of dice to determine the digits of $\pi$... and 
then those digits were frozen forever.


You can never *predict* the next digit of $\pi$ by looking at the digits that come before it. 
You can *calculate* the next digit of $\pi$ but you can not *guess* what it might be. If you guess
then your chances of getting it right are one in ten, no better.

```python
my_digits = '14159'


# ******

nDigits = 1000000

def iter_file(filename, chunksize=100000):
    with open(filename) as f:
        while True:
            substring = f.read(chunksize)
            if not substring:
                return
            # for Python 3
            # yield from substring
            # for Python 2
            for char in substring:
                yield char

from toolz import sliding_window

def look_for_sequence(filename, sequence):
    s = iter_file(filename)
    for i, group in enumerate(sliding_window(len(sequence), s)):
        if ''.join(group) == sequence:
            print ('found your sequence "{0}" at location {1} in pi'.format(sequence, i))
            return True
    return False      

if not look_for_sequence('/home/nbuser/pi1000000.txt', my_digits):
    print ('Sorry but your sequence "{0}" is not exist in the first million digits of pi.'.format(my_digits))
```

### How many times does it appear


The program below tells you how many times a sequence appears in the first one million digits of $\pi$. 

```python
my_digits = '747'


def iter_file(filename, chunksize=100000):
    with open(filename) as f:
        while True:
            substring = f.read(chunksize)
            if not substring:
                return
            # for Python 3
            # yield from substring
            # for Python 2
            for char in substring:
                yield char

from toolz import sliding_window

def look_for_sequence(filename, sequence):
    nHits = 0
    s = iter_file(filename)
    for i, group in enumerate(sliding_window(len(sequence), s)):
        if ''.join(group) == sequence:
            nHits += 1
    return nHits

count = look_for_sequence('/home/nbuser/pi1000000.txt', my_digits)
print ('Found "{0}" {1} times in the first million digits of pi.'.format(my_digits, count))
```
