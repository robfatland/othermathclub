## *And Then...* 
### A Kid Pipkin story for young readers


The time had come for Kid Pipkin's birthday party; and seeing as 
how she was turning 18 it was a momentous event; a Long Expected 
Party of great import; for one does not turn 18 every day you should 
know. Everyone had taken special care to travel to the port town
(which we have modeled on Sitka; but it is fictional)
and everyone had taken extra special care to arrive at the party
on time. People who like to arrive fashionably late miss the 
best part of the party, which is the shared excitement and happiness
to be ***at a party!!!*** I suspect that them what is fashionably 
late suffer a combination of self-conscious anxiety and a preoccupation 
with looking just so. Anyway they will get here eventually and we will
welcome them with open arms. 

But back to our story... some party guests were faces familiar, as 
in small towns we see one another almost every day; but
other guests Kid Pipkin knew were scattered far and wide and she 
had not heard from or seen them in some while. 
Of course the celebration was to be held at Fyz Byzkyn's farmhouse; 
and Kid Pipkin (being a considerate soul) had arrived well in 
advance of the start time (2PM) to help with preparations. 


'Oh I'm so glad you came early' exclaimed Fyz Byzkyn as she walked 
in the door. 'The horses have not been watered yet and today they 
happen to need precisely seven gallons. Would you mind terribly?' 


Kid Pipkin nodded, thankful for an important task to do, and stepped 
outside and across to the water trough where she found only unmarked
nine and five gallon containers. She turned on the spigot but then what?


Coming back into the house she remarked 'Your horses seemed in very good spirits, Kitty Dawn particularly.' (Kitty Dawn has had only minor roles
in the stories but she is the only named horse. When you ride Kittie Dawn
at a gallop she finds it amusing to veer under lower tree limbs to try and
knock you off; so you are obliged to hold onto the pommel as you
lean over and down to one side. Someone should have a talk with her 
about this as it is really quite dangerous.)


'Yes I'm not surprised they are feeling good, the horses' said Fyz Byzkyn.
'I got in a load of fresh hay from the fields yesterday... and I think 
Kittie has been teaching the other horses and the cow to nibble blackberries 
along the fence now those have come in. July has just flown by!'


There was a knock at the door, and Kid Pipkin went to welcome... 
who would it be? In fact it proved to be Alebasi, recently escaped and 
on the lam from the law once again. She was sporting a lovely diamond 
necklace that contrasted fabulously with her t-shirt and torn jeans.  
Further up the drive walked Abcdef Ghijklmno Pqrstuvwxyz together with
Zyxwvutsr Qponmlkji Hgfedcba, hand in hand. They had stepped away from 
their easels at the beach and flown up to celebrate the occasion.


Soon they were followed by Curious George walking up with the Man with 
the Yellow Hat. The Man With The Yellow hat does not speak in this story, 
for as was his custom he was present primarily to protect Curious George 
from harm, and Curious George will not be at any risk except
perhaps of a sugar overdose. But the Man with the Yellow Hat does smile 
a great deal and he enjoys being a part of everything.

Not far behind them came Horst Borskin deep in conversation with George 
Dyson. Horst Borskin seemed very unhappy about something as he gesticulated
angrily, flailing his arms about; so Kid Pipkin was pleased that he was 
having a good day. Her eyes swept across the 
gathering throng -- by now it was a smallish throng -- and Kid Pipkin 
found herself moved by these familiar and beloved people, all of them
coming here to celebrate the day with her. 
Years had passed by but this seemed not to matter in the slightest: 
Everyone was chattering away and laughing as if they were picking up 
from conversations begun this morning or just the day before. 


The doorbell rang again and this time it was the girl with the skirt of 
many different colors, the skirt in variegated shades of orange, and as she
made her entrance finally behind her it was Put Bunkin (whose real name 
was Bisaella, and whose *real* real name was Isabella) carrying a rather
large box. She was still Mayor at this time and I am afraid we do not find 
out what was in the box.


Soon they sat down to supper (for it was as I said early afternoon), 
and this proved to be a capital meal prepared by Fyz Byzkyn in a flurry 
of smoke and odors mysterious, slightly burnt-smelling smells that
wafted out from the kitchen.


The order of events of this supper was peculiar by ordinary standards, 
to put it mildly: First Fyz Byzkyn brought out a birthday cake with 18 
candles burning away, and George Dyson waved his hand and suddenly the
little flames connected together with sparkling beams of light, to 
produce a planar graph. Then according to custom they sang Happy Birthday 
(Zyxwvutsr sang it backwards) before Kid Pipkin made a wish
and blew the candles out. The little light beam edges of the graph,
their candle flame vertices blown out, tumbled upward 
to strike the ceiling and puff out in smoke and sparks. 'Thank you, 
George Dyson' said Fyz Byzkyn. 


The cake proved to be an Inverse Cake with three very thick layers of 
frosting separated by thin slivers of cake, and with more cake 
formed in a cylinder around about the outside and a disc of cake 
capping the top. Fyz Byzkyn seemed so pleased with 
his invention that nobody offered the observation that a cake made 
mostly of frosting was a bit on the rich side. 
And Curious George didn't mind anyway. 
After the cake was finished it was time for the main course 
of crab-stuffed mushrooms and salmon-stuffed brussels sprouts; 
and then a salad was served; and at last a series of appetizers 
appeared which all seemed to involve pesto.


"This is very much like the Backwards Suppers they would feed us
in the dreadful Army" griped Horst Borskyn with a sour expression 
and a disapproving shake of his head. 
As usual nobody had any idea what Horst Borskyn was talking about, 
particularly as he had never served in any Army... so they merely nodded 
and looked thoughtful. But then unexpectedly, once a minute or so 
of silence had passed following this peculiar remark, 
a gleam of triumph appeared in Fyz Byzkyn's eyes. Here it comes at last.


"Oh ho" he began, his eyes crinkling up as he enjoyed the moment 
beyond measure, "this meal is not *like* a Backwards Supper, Horst Borskyn,
it ***is*** a Backwards Supper", and he burst into triumphant laughter. 
Once he had wiped some tears of mirth away he 
continued "...and now I should like to know, from the birthday girl: 
How did you enjoy the Backwards Supper Kid Pipkin??"


Overcome, Kid Pipkin tried to think of a suitably grandiose response...
but what she managed to say was "Why yes of course it was ever so much the 
most splendid birthday supper I have ever had, and the best you
have ever cooked!'" with a great smile beaming. (A little awkward
perhaps but she meant every word.) Then suddenly her countenance 
faded and fell to something like sadness.  Everyone grew quiet and 
she continued in a small voice '...I only wish we 
could begin with this backwards supper and keep going backwards in time
ever so quickly to the days when I first met you all, so many many
years ago, 
that we might go out upon our adventures all over once again...'
And her eyes began shedding great tears. 
Nobody spoke but they all stood up and gathered about her and 
arms reached out to create a group hug that was 
the biggest group hug any of them could remember, 
and there was a great deal of sniffling and 
wiping of cheeks and noses.

After a time everyone stood up and some little smiles flickered about the
room. 'What shall we do now, then?' asked George Dyson. 


'Well we can't eat cake...' answered Put Bunkin (whose real name was of 
course But Punkyn)(and at this there was some polite chuckling) 
'...but perhaps we should tell stories, and remind ourselves of 
some of those adventures. I will go first...' and she began by recalling 
the vast caverns within the high vaulted peaks of the Sierra Nevada 
mountains, and unfortunately (she reminded them) she 
had been unable to go on this particular adventure when it began,
owing to her obligations as Mayor. 
(This was after her job as the Harbor Master.) 
'But something, some little voice told me, some
whisper of intuition said that things were not going well for you, 
and as it happened the breeze was blowing
in just the right direction...'


'Oh, I remember this!' interrupted Fyz Byzkyn 'We had solved the clock puzzles... but then
we were surrounded by those terrible creatures,
and it was earthquakes every few minutes and somehow lava was pouring...
flowing towards us... what were those creatures anyway? Like spiders!
And we felt doomed!
And well anyway Put Bunkin you just appeared
out of nowhere in your hot air balloon! I've never been so relieved 
ever ever in my life!'


'Yes' laughed Put Bunkyn 'that was my best and only chance to arrive on the scene to save the day!'


'Bunkin the machine?' asked George Dyson quietly... but nobody seemed 
to follow his meaning except Kid Pipkin who smiled.


'...and remind me: Was this not part of the great chase, as we
pursued the two fugitive wizards?' 
asked Horst Borskin. The very two wizards in question 
looked down at the floor in slight embarrassment. 
They had been the quietest guests and were feeling a 
little self-conscious, having been villains in the earliest adventures,
at a certain moment long ago. You may not remember this though of course 
you know their names, but it began with some shenanigans involving a 
rather unpleasant dinosaur.


'Yes' said George Dyson, and sensing the wizards' discomfort he 
added '...and I do wish that all of our stories could end as that one 
did, making new friends that once were our enemies.' 


There was a pause as everyone had a little think, and then the girl with 
the multicolored skirt (who had also been fairly quiet to this point) 
spoke up. 'I seem to recall there was a time 
when we had to negotiate with a terrible sea monster who was grabbing 
passing ships... but 
the memory is so distant now... I remember this creature had built a 
castle of stone on the sea floor...' 


'Oh indeed' said George Dyson 'and that was when we discovered that you 
lived with and spoke with and even rode about
on whales in your spare time. I always found it curious that you 
didn't really mention this earlier.'


'Megaptera novaeangliae as I recall' said Put Bunkin, by which she meant
they were particularly humpback whales, and she was rather pleased that
she knew the Latin.


'Well, one doesn't like to make a fuss' replied the girl with the orange
skirt. Strange that we have never learned her name don't you think? 'And 
besides,' she continued, 'don't we many of us reside in rather
unusual habitats? I mean, consider for example that George lives in 
a miraculous treehouse in a forest on Vancouver,
high up in a Douglas fir... and Alebasi's usual domicile is...' her 
voice trailed off as she considered whether it was polite to continue.

'...*prison*!!!' Alebasi blurted out with a big smile. 
'That's true and I count myself
among former foes now in the friends category. 
I am proud to say Kid Pipkin that you were responsible for
my arrest and incarceration at least three times, which I do 
respect. As I was breaking out of prison yesterday (to attend 
this party) I found myself wishing it a little more difficult
these days. With all my past practice a jail break has become 
about as difficult as shopping for groceries. Plus now come to 
think of it: I'm out of prison again! I will need
some lodging and I must start paying rent again... so... 
let me see... Fyz Byzkyn you wouldn't happen to have a 
spare room I might rent? I can pay in cash or diamonds, 
however you like.'


'Yes' said Fyz Byzkyn (though it sounded more like 'Yus') 
'Of course you may stay in the guest room for 
as long as you like; and if anyone else would like to 
sleep over you are all of you more than welcome. Always.'


The room lapsed into general chit chat for some time until Put 
spoke up again (she seemed to have a sense of how these things 
should proceed) by asking at an above-medium volume 'Kid Pipkin 
what have you been up to lately?' 


There was a quiet pause as Kid Pipkin seemed to consider, and then she 
spoke. 'Well as you know
perhaps, I have a small sailboat now and I am considering taking 
her out on a voyage of sorts.
In fact I suppose now I had better tell you that this will be 
good-bye... it is part of why i came 
over so rather sad earlier.'


Everyone looked stunned. 'Goodbye?? Where are you going?' asked Fyz Byzkyn in the resulting quiet.


'I am... I think I have worked out how to sail to Tomorrow...'


'What do you mean? You say that as if Tomorrow is a place.'


'Well obviously we just go to sleep tonight and we will all 
wake up tomorrow... but have you
ever noticed that tomorrow always becomes today? So even though 
we agree on what tomorrow 
means as a real thing, we are never able to get there?'


Everyone listened quietly, though Horst Borskyn made a little, 
uncomfortable cough.


'So I have done a little mathematics and I found a way to navigate to 
what I believe we call Tomorrow;
and since this is seemingly impossible I had to solve an 
impossible navigation problem for it to all work out. 
Specifically I will have to sail to the moon. And the way to do that 
is: You know how a full moon sets on the ocean and sometimes 
there is a kind of a 
mirage where the bottom of the moon's circumference flips over and the effect is a kind of 
an Omega shape?' (Those who knew what a capital omega looks like 
now pictured this.) 
'Well one day I was thinking about this but not in terms of
planets moving about and orbital mechanics and atmospheric lensing
of moonlight... 
but instead I found myself thinking in an abstract way about 
calculus on manifolds, 
and I proved to myself (I *think*) that for a couple of moments 
in the course of the full moon setting on the ocean 
I could navigate across the mirage in my sailboat...
but only if I made a branch cut across time... 
and this would bring me to the moon but, because of the branch cut:
not the moon *Today*. And then I sorted out that if I go about the branch 
cut *counterclockwise* I get to Yesterday.
But this is less interesting as well as also impossible,
as we have aready been there... we know Yesterday... 
and so that leaves me with going *clockwise* which takes me to...
Tomorrow.
That is where I shall go.' 
(Kid Pipkin's eyes lit up just a bit in describing all of this.)


And as the room tried to make sense of what she had said, 
Fyz Byzkyn (ever a practical soul), asked
'...and when can we expect you to return?'

Kid Pipkin's eyes fell, and it seemed she might cry again. 
'I don't know, honestly...
perhaps not for a very long time.'

Curious George took a step towards Kid Pipkin and grabbed the 
hem of her shirt, looking 
up to ask without words if he might come along. George Dyson 
said 'I'm not working on 
anything at the moment, even perhaps for a very long time...' 
Put Bunkin spoke next 'You 
know I have space for us on the Diesel Duck if you do not mind 
a bigger vessel. It could
be dangerous... might be handy to have some 
dangerous-looking friends along with you.' 

And at this Kid Pipkin did burst into tears once again. Before she could say anything the girl with the multicolored skirt spoke. 'I have been to Tomorrow, and as much as I would like to go again with you Kid Pipkin, I know that you have to go by yourself. I don't understand how your navigation plan works... everyone finds Tomorrow in their own way I think. 
I got there with the help of some sea turtles... but anyway I 
understand that we can't come with you.'


'Thankyou' was all that Kid Pipkin (whose real name was Kyd Pypkyn) 
could manage to say. 


'If you vanish right at this
instant I shall not be amused' said Horst Borskyn.


'Soon... but not just this instant' said Kid Pipkin.


To restore a mood of a celebration they all agreed it was time 
to play a game of baseball. It was a strange sort of baseball and 
they played their hearts out, out on the open field behind the 
barn. The tall tree with the enscription carved high up on the 
bark by the girl who had lost her cow acted as home base, and 
the horses joined in, and there was a baseball, two tennis balls, 
a basket ball and a soccer ball all in play. And there were not 
two but three teams, and no score, and everyone had to run a great 
deal back and forth, and they were obliged to jump across the furrow 
in the ground that had never been filled in, the furrow made by a 
certain sled that one evening long ago had crashed down from the sky
leaving a pair of gold-rimmed spectacles. Then each of them, having 
run and jumped according to the complicated rules, 
would huff and blow with their hands on their knees to get their 
breath back... and nibble a few blackberries. 


'The game has ended!' shouted Fyz Byzkyn after two and a half
hours, and Put Bunkin shepherded them all back inside, and Kid Pipkin
made sandwiches and coffee, and for awhile it was munching 
and sipping.


Kid Pipkin surveying everyone intent on dinner said, rather to herself, 
'Oh how I shall miss you all.' And at this Put Bunkin started
and exclaimed 'Ok enough already! No more sad! Consider won't you, 
first there is more out there than you know, more than anyone 
can ever know, and you are simply stepping forward, further in, 
further on, into the great wide world. And what is more, we are, here
in this room, none of us: *real*. We are ideas, we are archetypes, 
we are emotions and emblems and stories and symbols; and you have 
but to think of us and voila we are with you, for the rest of your life. 
We are not real in the way you imagine what is *real*,
but at the same instant we are more real than every *real* 
person on earth whom you have never met! Come, you look baffled now.
Don't look baffled; I shall prove this to you.'
Kid Pipkin still blinked about the room, not comprehending. 


'Stand up, come here, look at me' said Put Bunkin. Kid Pipkin 
did as she was directed. 'I am the spirit of your imagination mixed with 
your spirit of responsibility.' continued Put Bunkin. 'A part of you 
needs stability, safety, a plan to get out of trouble when things 
go wrong in your adventures. That is me: I am you
taking care of you, and so I *am* you. Now give me a hug!' 
After hesitating at what might happen, Kid Pipkin took a step 
forward and wrapped her arms about Put Bunkin for 
what she intended to be the greatest forever hug of the world. 
But to her surprise after Put Bunkin
had hugged her back she felt Put Bunkin change and compress and... 
fold?... there was no better term for it; and fold again and 
again until Put had folded down into a small paper card the size 
of a birthday card. On this card was a beautiful abstract painting, 
and the painting *was* Put Bunkin. Kid Pipkin placed the card 
in her shirt pocket, above her heart, where she could feel it alive 
and glowing. And then one by one each of them came up and gave 
her a hug. 


George Dyson was next; and Kid Pipkin said 'George you can do magic... 
but I can not...so don't say you are me.' George Dyson asked 
'Am I magic? Yes. I am the change you can bring to the world.
Discipline, reason, observation, experiment... hard work...' and then 
in a whisper he added 'in other words: *fun*...' and Kid Pipkin gave 
him a hug. And suddenly, instead of George Dyson
there was a beautiful painting on a card, and into her pocket it went.


Curious George raised his eyebrows, and Kid Pipkin saw his willingness to 
set aside caution and explore and improvise, to be on friendly terms with
the world, to assume the best in everyone as a starting point. She was 
surprised at how much he represented, insofar as he was just a little 
monkey from a children's book.
She folded him up, and she folded up the Man with the Yellow Hat 
as well, whom she supposed represented cleaning up afterwards.


Alebasi before her hug stated in no uncertain terms that she was confidence 
and humor... and most importantly outrage. 'I am your emotions telling 
you what is important. Listen to me. Together you and I shall shout at the unfairness of the world' she said. 


The girl with the skirt of many colors said, 'To you I am the knowledge 
and the wisdom of other people, the people you hope to find...'


Horst Borskyn explained 'I may seem grumpy to you. Well... I suppose so. 
I *am* grumpy. I *am* complaining. It is part of making sense of things that make
no sense to me. My gift to you... to be honest I find it easier to come back to
being kind once I give myself permission to get the grumpy out of my ears' 
and with this somewhat peculiar remark, up he folded with a hug.


And last came Fyz Byzkyn. Just as had the others he began to tell
Kid Pipkin just who he was. 'Can I tell you a secret?' he began.
'I am' he said, and then paused a moment, unable to speak,
and eventually in a wavery voice he continued '...Did you know
that I am just the 
ability to fail... and to try,
and to fail, and to not understand, and to try, and fail, 
and care about trying, and then fail, and still not understand, 
and try, and fail again... and so on... I believe that I am right 
in what I think... while I know for certain that I am most likely quite 
*wrong*... And I am not terribly bright, I know... but I try to 
never let this discourage me. I am the one who tries.'
And there was so much love in his eyes. And before reaching out 
to give her his hug he touched first her left temple, and then 
her right temple, saying as he did so: 'Sirius... Rigel...'
And as he folded into a beautiful painting, a painting like 
no other, and as she placed his card in her shirt pocket, 
she became aware of a gentle rocking feeling, 
and she found that she was on a little sailboat,
on a calm ocean, and there on a horizon that seemed somehow close 
enough to reach over and touch there hung
an impossibly large orangy moon, about to dip its edge
into the sea.
