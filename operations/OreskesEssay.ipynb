### Whether tis nobler


Recently a friend regailed me about their encounter with a person who expressed
an ill-considered opinion. 'Ill-conidered' may be putting things too lightly; the
opinion expressed was shocking and outrageous and so I was suitably shocked and 
outraged; but my friend, the witness, did not have time, inclination, patience 
or energy to offer a reply, not even the time of day let alone a calm 
rebuttal. Rather they wandered off to quieter and more interesting places. 


Now my question: Under what circumstances might one exchange a word or two when
faced with such ill-considered opinions? I feel the answer concerns *time*... for
example having the necessary time to follow through should this be necessary. 
Suppose there is time for such a word... time not deferred to tomorrow, but in the moment.
How then to engage? And not in the manner of 'How do I enlighten a fool?' but rather, 
'Given a difference in views, how shall I attempt a conversation?'


My first thought in reply is that English can be a very expressive, nuanced language; 
so that is a good place to start. Suppose we place questions an exploratory question
inside of 'wrapper' language that might defuse antagonism: "Help me understand, why do 
you say that X is Y?" Here "X is Y" is as close as possible a quote of the troubling 
opinion; and the wrapper "Help me understand" casts the other person in a helpful and 
enlightening role. 


The underlying goal of starting with defusing language does not *bury* my agenda 
(which is to express a supported counter-opinion) but it does acknowledge that the 
only *constructive* path available is based upon opening doors to new ideas. So
my plan involves dual participation. The person with the troubling opinion
is granted equal status from the outset. In fact we are acting as one player in
a two player game. Plus there is a hidden facet: We are also in the role of 
game Arbiter. So let's take a swing at defining the game as an optimization process.


* Player 1 states an opinion S. Player 1 may have an underlying rationale P. 
* Player 2 responds with an inclusive question Q. 
* The Arbiter (who is also Player 2) listens to the ensuing exchange.
* Player 2 is attempting to reach a resolution R in minimal time.
    * Player 2 has two kinds of remark at their disposal
        * Expository questions
        * Statements on the topic (factual or normative)
    * Player 2 can say things that are any combination of these remarks
* The Arbiter may at any point disqualify Player 1
    * Grounds for disqualification include
        * Rudeness
        * Refusal to respond in genuine good faith
        * Sophistry: The broad category of false argument
        * Resorting to unverifiable assertions
        * Resorting to invocation of authority


When the Arbiter disqualifies Player 1 we say the game is a draw but the Arbiter gets a personal win.


When the game reaches R we say Player 2 is the winner. 


The only circumstances under which Player 1 wins is if the discussion arrives at P.


The short version is this: Prepare; and by being prepared: Keep the 
conversation moving on a non-threatening ('safe') axis. Facts and reasoning come into play
so I digress somewhat into -- or borrow from -- how science works. A premise
is that *because science is successful* it is worth understanding it as 
a means to this business of 'staying on a safe axis'.


Another premise of this essay is that self-awareness (or meta-perception) 
(or preparation) is a means of self empowerment; and this is a good thing. It
may help with cold feet when it comes to leaping into the fray. Who among us
with icy feet relishes a duel of wits? Not I, say I!

#### 1.1 Necessary conditions


These days: Identity is raveled up with confrontation and escalation.
Pointed remarks, taking offense, hurling insults, breaking windows and worse: 
The underlying principle is *'question me: question my existence'* in this our age of outrage. 
So let's agree that aggravating another person along the identity axis is an option
but not likely conducive to an exchange of views. We need that alternative
orthogonal axis of conversation... 


As we sidle on up to thoughts on *how* to be orthogonal I will venture a side remark: 
What if the topic is very 
*important* to me. Perhaps the offending remark is overt racism.
Perhaps in that case I throw reserve to the wind and make some pointed comment:
***'THE REAL PROBLEM HERE IS THAT YOU ARE A ADJECTIVE NOUN.'***
Very well; I said my piece. Let us distinguish between 
*engagement* and *outburst*: In an *outburst* I deal with a misguided view by responding
back in the heat of the moment. There is nothing wrong with that (so long as 
everyone present is safe). 
There are times when a good shout out is called for. Contrast *engagement*: 
I have the time; I come in prepared; with a plan; and a calm mind: 
My hope is that here is an opportunity broaden someone's view and 
in addition broaden my own. The point: Knowing about the 
distinction between outburst and engagement is useful
meta-perspective.

### Professor Oreskes takes on the Royal Society


This is about the balance between
the value of authoritative voices on the one hand, and inflexibility
that is sometimes bound up with authority.


Around 1660 a new club sprang up in London calling itself the Royal 
Society. The members came up with a nice motto: *'Nullius in verba'* 
which translates 'Don't take anybody's word for it'. 
To illustrate: Judge Samuel Dexter published a communication offering to
resolve a mystery: Where do swallows disappear to in winter? He maintained 
that they swim down to the bottoms of ponds and hibernate
in the mud until spring. Let's give the Royal Society their say: 
*Nullius in verba!*.
If you are unable to supply evidence for your claims that can be reliably 
reproduced: Swallows at the bottom of lakes is just words. 
Emphasis on reproducibility was a major advance in how science gets 
done. It also highlights a recurring 
social aspect: How are we to treat subject matter experts?


Today scientists will occasionally make an ironic remark about how science advances by 
death, meaning: The passing of an intransigent elder scientist makes 
room for new ideas to flourish.
However this is only part of the picture. Scientific opinions are 
developed over years of concentrated effort. Often progress is gradual; 
sometimes progress proceeds in jumps; but by whatever
degrees progress is made there is (usually) a self-examination process
in its wake. 'Does this make sense? Is it self-consistent? Does it fit together with
other ideas? What are the remaining questions? What new questions arise? 
Are we missing something?' Also 99% of the time this discussion is polite
and includes a lot of 'maybe' language.



In the interest of representing major revisions of established theory: 
I can't resist an example.



At some point back around 1900 it was an
accepted factual mystery that the orbit of mercury about the sun -- the *elliptical* orbit --
did not perfectly close back on itself. Draw this orbit as many ellipses over time:
The long axis of the ellipse slowly turns or *precess* over centuries.
The little problem here was that this precession could not be accounted for
using Newton's laws of universal gravitation. As a result someone said 'The
discrepancy can be accounted for by the gravitational influence of
a planet we have not detected yet. Let's call it *Vulcan*.' And like that: The search 
for planet Vulcan was on. But they never found it. What they did find, however, 
was that a major revision of the theory of gravity from Newton was possible. 
A German physicist had worked out this new theory of gravity; and it fit the 
data for the precession of Mercury's orbit perfectly. Furthermore it made 
predictions about the bending of starlight by massive objects which could be tested.
So the tests went ahead and the new theory was again successful. As a result 
it gradually became part of the picture; and every time it made a successful new
prediction it took another step towards being considered correct. 'Newton says
X and Einstein says Y so we measured it and it turned out to be Y... another win
for Einstein's theory.'


I seem to have gotten away from the original topic of the ill-considered opinion
so let me try and bring that back here. First of all from the science picture 
there are generally two types of communication in play: The internal discussion 
between scientists concerning scientific
understanding; and communicating and understanding science in the broader sphere 
of human culture. How do discussions in these two forums connect with one another? 
Beyond interest they connect when scientific results impact culture of course, 
the acute example being climate change. So now a few words on how these are
related.


Naomi Oreskes is a historian of science. Last I checked she was at Harvard. 
In 2004 she published an analysis of papers on climate change in the scientific literature.
Her paper, 'Beyond the Ivory Tower: The Scientific Consensus on Climate Change' (*Science*)
immediately produced a great wave of relief in the geosciences because it 
controverted the disinformation campaign that 'the science of climate change isn't settled yet'. 
The consensus in the literature, Dr. Oreskes reported, was overwhelming. So this looks like
an appeal to authority: 'Everyone agrees so it must be true.' 


She later wrote a book on authority in science called 
[Why trust science?](https://press.princeton.edu/books/hardcover/9780691179001/why-trust-science).
Her point in this book is that trusting designated experts -- scientists -- is not only acceptable
practice, it is necessary to human survival. In a 2019 interview in the wake of the publication
of this book she states: "We have a number of issues that are truly life and death that hinge 
upon our understanding and taking to heart scientific evidence. The two obvious ones on the global 
scale? Climate change and vaccinations." (See https://undark.org/2019/11/01/five-questions-naomi-oreskes.)


Here are a few *at random* observations and assertions I can make about how science works. 
These are useful to have in hand when talking to a non-scientist because they may go against
common perception... and I present these as 'subject to debate'. 


- 'The scientific method' as a coherent procedure is a complete fabrication, an illusion
    - Science is done by whatever means seem appropriate
    - There are underlying ideas of consistency and reproducibility; but nothing resembling a recipe
    - For want of a clear recipe: Scientific research is often improvisational, chaotic and biased
    - Evaluating scientific progress in terms of 'how' can be a real conundrum
        - Middle school science fairs often bring students into contact with this conundrum
    - This begs the question: Is there more to science than chaotic improvisation?
- There is a balanced approach to understanding scientific consensus
    - Premise: A scientist is legitimately looking to better understand an objective reality
    - Premise: Every scientist brings their own bias to this process
    - Aggregated literature represents an expression of consensus
        - This is backed up by time (years invested) and careful thought
        - Consensus is always regarded as 'subject to future amendment or modification'
        - Challenging or controverting consensus is an open possibility
            - But the circumstances for this are usually difficult to achieve
                - A new explanation must account for the existing evidence
                - A new explanation must in particular not dispense with fact
                - New evidence is necessary to support a new explanation
                - There is no utility in this process for unsupported statements of *belief*
        - Much more common: New work builds upon and sharpens existing consensus
- This raises the question 'Why is scientific consensus disparaged in popular culture?'
    - One reasonable explanation: 'Cynical actors undermine the truth.' -Naomi Oreskes
        - She has spent several years studying how tobacco companies approached lung cancer
    - Anyone with a microphone and an agenda can use scientific consensus for their own ends
        - The idea here is that *attention* is the only currency that matters
            - Graffiti on a boxcar gets very little attention
            - Graffiti on a wall in the Louvre gets headlines
            
            
Our story so far: A hypothetical individual has expressed views whose veracity we question.
The case is made for engaging in an *off-axis* discussion. By *off-axis* I mean a rhetorical
direction that does not threaten the hypothetical person's sense of identity. So far two framing 
concepts have been suggested as a *means* of such an off-axis discussion. First is engagement: Develop 
a conversation as respectful and *participatory*. Second is consensus: Understand and work with science 
as a nuanced human endeavor; admitting the possibility that scientists are designated experts for 
good reason. Preparation in both these regards is -- I assert -- essential to success; 
metaphorical open a door to new possibilities.


Ok. So far so good. Let's add in a sequence of supporting ideas.

### Part 3: Rules of engagement


All other things being equal: Engage when prepared; therefore prepare!


#### 1. What the point is **not**

We see engagement on satirical TV shows like The Daily Show. These are calculated to get a laugh.
I certainly have a good laugh. The press and social media goes further: They delight in amplifying
situations where person A belittles person B or idea C. In 2015 Oklahoma Senator James Inhofe 
belittled climate change on the senate floor. There happened to be a snowstorm; so
he took the podium in the senate chamber holding a snowball he had scooped up outside.
In the moment he owned the message: Climate change is obviously a hoax. 
Nobody was in that place with a microphone to point out Inhofe's failure to understand 'climate'.
Nobody had a microphone to point out that February 2015 was the warmest February on record.
But the press was there to make Inhofe with his snowball the star of the news cycle.


The point is not to belittle, nor is it to conflate conversation with entertainment. 

#### 2. Crossing motivations, the Alan Hamlet lesson


Alan Hamlet is a climate scientist. In about 2011 he attended a public forum in Skagit County
in person to explain the impacts of climate change on Skagit Valley. He was met with skepticism
as Skagit residents took turns at the microphone 
reading from disinformation flyers they had printed off the internet. 
They (the incensed residents of Skagit county) were looking for someone who personified taking
away their personal freedom. Alan Hamlet was their candidate for this role. However he 
was simply present to explain how rising sea level was creating up-river pressure from the ocean; 
which pushed salt water up the Skagit River, closer and closer to the freshwater intake that provides 
drinking and irrigation water to Skagit county residents. Alan Hamlet was there to explain how 
lower summer river flow and higher temperatures were killing salmon eggs in the upper Skagit River.
Alan Hamlet did not drive for two hours to attend a town hall meeting so he could talk to farmers
about handing over their guns. He was not intent on making them feel guilty over driving cars that 
produce carbon dioxide. He was interested in promoting public awareness of facts, specifically
the physical consequences of climate change. 

#### 3. The nature of disagreement: The Tad Pfeffer remark


Some time around 2007 a glaciologist named Tad Pfeffer 
hosted a climate change forum in a lecture hall on the campus of the University of Colorado
Boulder. The audience was mostly scientists; and the discussion at one point turned to 
'winning the argument' over the reality of climate change. A participant expressed the view
that this argument *must* be won simply because the facts are the facts, end of story. 


Tad took the microphone and he took a deep breath. He began by agreeing that 
yes, the point of science is to get at the truth, to acknowledge the facts, to arrive at a correct 
explanation no matter how uncomfortable that may be. But in jurisprudence, in a court of law,
the reality is different. There the point is to win the case, irrespective of what happens to be 
true. If you go into a court of law expecting to establish what is true: You are in the wrong place.


This remark was born out by Waleed Abdalati in a different gathering of scientists. 
The remarks he had to offer were from his experiences testifying before Congress on climate change. 
He had encountered legislators who were virulently, vocally opposed to the very concept of 
human-induced climate changed. These people would make a point of being completely adversarial and dismissive
in open-door hearings, those that are open to the news media. And yet: He was surprised to find that in 
closed-door hearings -- no news cameras -- these same legislators were interested in understanding 
human-induced climate change in perfectly congenial and reasonable discussions.

#### 4. Shall Brock teach himself?


A co-worker of mine, one Brock Johnson, was interested in discussing and understanding
climate change. He had read that water vapor was a much more impactful greenhouse
gas than carbon dioxide by perhaps a factor of ten. 
Yet all the focus of global warming centered on carbon dioxide, not water vapor. 
It seemed to him that scientists were focused on the wrong gas. 


I made the mistake of looking up the answer and explaining it to him. 
He was unsatisfied with my explanation and asked additional questions. 
After a few iterations of explanation and more questions I realized: 
Brock was not interested in understanding the science. Brock was 
interested in having someone else do the intellectual work for him. 


Alright so without casting aspersions on anyone ('Brock' is a fictional name)
let's deem this *intellectual laziness*. We see it whenever someone objects to 
something (or doesn't *believe* in something) that they are at a loss to 
accurately describe. Senator Inhofe gave us the perfect example with his
snowball mentioned above.


My conclusion from the back-and-forth with Brock was that intellectual laziness is 
a real possibility. (I am extremely naive in this regard; it had not occurred to me 
up to that point.) Having decided that such laziness is not an actionable 
state of mind: What is left to be done here? Perhaps we can have a conversation 
that subtly *addresses* intellectual laziness, again taking care to avoid using 
pointed language. 'Insults rarely strengthen an argument' as the old saying goes. 
What if we say to Brock, 'Yeah, water vapor *is* a greenhouse gas, and you 
are correct when you infer that it is *not* the critical factor in human-driven 
warming of the earth. Carbon dioxide is the primary culprit. So I have a suggestion.
You look up why, do the reading, and then you explain it to me. I'll do the reading 
also; and I will be prepared to ask you perplexing questions.' From this point
Brock has a choice: Does he back down from his questions about water vapor or does
he step up and learn about the situation for himself? If the former: Such Brocks 
are not worth the time of engagement. If the latter: Let the conversation resume.


How did we do, going for a non-loaded process and an open door? 

#### 5. Normative and positive statements


A normative statement is about the way things should be. 
A positive statement is about the way things are. 
'There are no swallows hibernating at the bottom of this lake' sounds like a negative; but as a 
factual remark it is a *positive statement*.
'We should convince the swallows to hibernate on the moon rather than in lakes' is a weird normative statement. 


How things should be is really a matter of understanding the way things are and then
prioritizing future outcomes; and making a plan for getting to those outcomes. So 
positive statements precede normative statements.


In my observation people like to jump to normative statements without first agreeing on how things
actually *are*. 


Be on the lookout for a pivot from positive to normative. It is ok to call out 
*normatives* as *not positives*. Perhaps: Don't argue normatives; at least,
not until the positives are sorted out.

#### 6. The rhetorical fulcrum


A fulcrum supports a board that can tip to one side or the other. The metaphor is meant to 
invoke how small changes in language -- particularly in agreeing on meaning -- can have big consequences. 


Let's suppose that language matters. Terminology matters. Some terms are pretty innocent looking... 
and yet they can have very inflammatory connotations. A big part of preparation is knowing what 
that lit fuse looks like. In a chat about climate change the term *climate* is frequently 
misconstrued. Climate refers to a running average view taken over several decades. 
Year-to-year we find that environmental observations such as temperature are noisy; but
a 35 year running average reduces this noise and shows the actual trend.


Practically speaking we have to be listening for signs of a misunderstanding. 
The moment we hear "the past ten years have shown
a marked downward trend in global temperature" we have the classic 
misunderstanding of climate. What to do? Simply ask for the microphone and
point out climate is an average over 35 years. Now celebrate that the conversation 
course-correction, talking about the same thing!


Let's look at two additional examples of oft-misunderstood language: 


- climate change consensus is based on observation
    - the consensus on human-driven climate change is based on physics
        - ...and not on computer models
        - the moment we hear "climate models disagree with one another"...
            - ...we have arrived at a classic misunderstanding
            - Climate models are predictive *only* in a rudimentary sense
            - Consider the earth as a complicated system (and don't forget to include the sun)
                - Imagine modeling the room one is in on a scale of one-centimeter cubes
                - Now ask 'How big is the computer needed to model the earth?'
            - Make a computer model: Is it correct? Answering this is a starting point for modeling
                - From here we go to a very exploratory, speculative process of improving models
                - Modeling is not the basis of *understanding* the nature of climate change
                - Modeling is a hopeful attempt to understand the potential severity of climate change in coming decades
            
- evolution is both a fact and a theory
    - evolution theory is supported by multiple independent lines of factual evidence
    - commonly objections to evolution invoke *belief*: 'I don't *believe* in evolution.'
    - Scientific theory is a matter of explanation supported by reason, observation and experiment
    - An alternative approach to understanding based in belief begs absurd questions:
        - Are other scientific theories subject to a person's state of belief?
            - If I choose to stop believing in...
                - ...electromagnetism: does my car fail to start?
                - ...quantum mechanics: does my television burst into flames?
                - ...gravity: can I fly?
                - ...biology: can I cure cancer?
        - Ok no... but evolution is *different* because my TV does not use it
            - ...or really it is different as a threat to my sense of identity
            - Very well, let us set up and ask a helpful question about *intent*
                - Science tells us that evolution is...
                    - ...simple at a high level, based on two simple observations
                    - ...complicated in how it gets the job done (molecular biology)
                - Consider the tens of thousands of individuals responsible for evolution theory
                    - They devoted years of their lives on their particular analysis...
                    - They did experiments, they built new equipment, they wrote papers, ...
                    - They persevered and gradually this increasingly complicated picture emerged...
                    - ...that explains the fundamental nature of life on earth
                - So speculate: Which explanation of this process seems more likely?
                    - Those scientists solved a vast, intricate, utterly fascinating puzzle
                    - Those scientists concocted a story so they could threaten your self-esteem

#### 7. Rhetorical flourishes: boom or bust?


'Let's have you play the role of the scientist here' might work well if scientists are
regarded in a positive light. On the other hand if scientists are only in it for the 
money, are part of a grand conspiracy, etcetera, then obviously suggesting someone take
on that role will fall directly on the identity threat axis. 


To generalize: I have found that using rhetorical devices can be dangerous. The risk
is a loss of sincerity. Playing to the crowd or filming an interview that will belittle 
someone on the Daily Show: It can be done but it probably departs from our stated purpose. 



#### 8. Logical fallacies


Look them up! Read all about them! Three of my favorites are *ad hominem argument*, *arguing from my limitations*,
and *whataboutism*. The latter is very prevalent in political discourse. But the one that really irks me the 
most is the second of these three: 'This topic is too difficult for me to understand so I conclude thus.'


Having a good understanding of logical fallacies in arguments is useful preparation. The trick is to
recognize them when they are played; but then work towards a rebuttal that relies on reason rather than
citation. In other words, it may not work out to shout 'Aha! That is proof by appeal to authority! 
Logical fallacy! Face it, you are going to have to do better!'


And speaking of appeal to authority: As mentioned above there is room for invoking authority in these
discussions. Argument is never simple.


#### 9. Oh yes, one more thing


A most pernicious logical fallacy slash rhetorical swindle is posing a question 
which is (obviously or subtly) based on a falsehood. 'Have you stopped beating your dog?'
Take a deep breath... 'Your question presumes I beat my dog when I do not, nor have I.' 



#### 10. Ambiguity invites Participation


The temptation for me is to attempt to prove an assertion with finality. Maybe it is
from an over-developed fondness for mathematics. But there ought to be an entire
new essay on making ambiguity, absence of finality, the goal. I would write it but 
I have already written too much... so I leave it to you. 



#### 11. The miracle of objective reality


Up near the top of this essay I cited the premise that meta-perception is a good thing. 
(Notice the normative nature of this statement.) There is another premise lurking here
that is worth a quick mention: The world operates on consistent rules, the 'laws of 
nature', independent of our degree of or lack of understanding of those rules. This is
sometimes called *objective reality*. A good understanding of these rules
is what science is trying to approximate, again with emphasis on 'because it is fun'.


There is a bit of a circular argument here: Why does nature follow rules that seem to
be the same everywhere? Why is objective reality so self-consistent and so complicated? 
The reason is that there *are* other realities that are not consistent... 
but because of that they are unable to support life as we understand it; so that is why
we do not find ourselves *there*. The one we landed in is nice and consistent and does not
are what we think of it. Those are the cards (the reality) that we have been dealt. 



#### 12. The burden of proof


Consider the challenge to prove there are no extraterrestrials currently visiting earth. 
Very difficult...  hard to disprove a negative. On the other hand the proof of the alternative
has met with zero successes to date. On whom does the burden of proof rest? A reasonable
rule of thumb is 'Extraordinary claims require extraordinary proof.' So let us be prepared
to ask 'Whose problem is this, anyway?'



### Part 4. Coda


Through all of this (particularly in my theme of *preparedness*) there strikes me there
is a combative tone. There are a lot of insults to distill and hurl in
the heat of an argument. 'You obviously couldn't be bothered to understand climate change.'
'That's what we call what-about-ism.' 'You have presumed your conclusion as your premise.' 
These aren't necessarily directly insulting; but they are conversational dead ends. They 
leave the other person with nowhere to go. So this is the challenge and the peril of trying 
to navigate a conversation off of the axis of threat to identity: It can be very easy to
slip right back there anyway. 


So here is a final word: I use the word ambiguity to mean not reaching a conclusion. 
It might suffice, we hope it might be enough, to bring in new ideas and to leave open air 
around the disussion. For a person such as myself who needs questions to have closed-form
answers, who needs stories to have satisfying endings: This idea of leaving open air can 
be very unpleasant. At the end of the day what we are doing, being willing to leave off
early, is that we are betting on the other person to think about it. I claim this is an
even bet. 


Alright, so much for a philosophy of argumentation. As for me: No more of this blubbering 
now, we are going a-whaling, and there is plenty of that yet to come. Let us scrape the 
ice from our frosted feet, and see what sort of a place this world may be.

Cargo Cult Science
 

by RICHARD P. FEYNMAN

 
Some remarks on science, pseudoscience, and learning how to not fool yourself. Caltech’s 1974 commencement address.


During the Middle Ages there were all kinds of crazy ideas, such as that a piece of rhinoceros horn would increase potency.  (Another crazy idea of the Middle Ages is these hats we have on today—which is too loose in my case.)  Then a method was discovered for separating the ideas—which was to try one to see if it worked, and if it didn’t work, to eliminate it.  This method became organized, of course, into science.  And it developed very well, so that we are now in the scientific age.  It is such a scientific age, in fact, that we have difficulty in understanding how­ witch doctors could ever have existed, when nothing that they proposed ever really worked—or very little of it did.

 

But even today I meet lots of people who sooner or later get me into a conversation about UFO’s, or astrology, or some form of mysticism, expanded consciousness, new types of awareness, ESP, and so forth.  And I’ve concluded that it’s not a scientific world.

 

Most people believe so many wonderful things that I decided to investigate why they did.  And what has been referred to as my curiosity for investigation has landed me in a difficulty where I found so much junk to talk about that I can’t do it in this talk.  I’m overwhelmed.  First I started out by investigating various ideas of mysticism, and mystic experiences.  I went into isolation tanks (they’re dark and quiet and you float in Epsom salts) and got many hours of hallucinations, so I know something about that.  Then I went to Esalen, which is a hotbed of this kind of thought (it’s a wonderful place; you should go visit there). Then I became overwhelmed. I didn’t realize how much there was.

 

I was sitting, for example, in a hot bath and there’s another guy and a girl in the bath.  He says to the girl, “I’m learning massage and I wonder if I could practice on you?”  She says OK, so she gets up on a table and he starts off on her foot—working on her big toe and pushing it around.  Then he turns to what is apparently his instructor, and says, “I feel a kind of dent.  Is that the pituitary?”  And she says, “No, that’s not the way it feels.” I say, “You’re a hell of a long way from the pituitary, man.”  And they both looked at me—I had blown my cover, you see—and she said, “It’s reflexology.”  So I closed my eyes and appeared to be meditating.

 

That’s just an example of the kind of things that overwhelm me.  I also looked into extrasensory perception and PSI phenomena, and the latest craze there was Uri Geller, a man who is supposed to be able to bend keys by rubbing them with his finger.  So I went to his hotel room, on his invitation, to see a demonstration of both mind reading and bending keys.  He didn’t do any mind reading that succeeded; nobody can read my mind, I guess.  And my boy held a key and Geller rubbed it, and nothing happened.  Then he told us it works better under water, and so you can picture all of us standing in the bathroom with the water turned on and the key under it, and him rubbing the key with his finger.  Nothing happened.  So I was unable to investigate that phenomenon.

 

But then I began to think, what else is there that we believe?  (And I thought then about the witch doctors, and how easy it would have been to check on them by noticing that nothing really worked.)  So I found things that even more people believe, such as that we have some knowledge of how to educate.  There are big schools of reading methods and mathematics methods, and so forth, but if you notice, you’ll see the reading scores keep going down—or hardly going up—in spite of the fact that we continually use these same people to improve the methods.  There’s a witch doctor remedy that doesn’t work.  It ought to be looked into: how do they know that their method should work?  Another example is how to treat criminals.  We obviously have made no progress—lots of theory, but no progress—in decreasing the amount of crime by the method that we use to handle criminals.

 

Yet these things are said to be scientific.  We study them.  And I think ordinary people with commonsense ideas are intimidated by this pseudoscience.  A teacher who has some good idea of how to teach her children to read is forced by the school system to do it some other way—or is even fooled by the school system into thinking that her method is not necessarily a good one.  Or a parent of bad boys, after disciplining them in one way or another, feels guilty for the rest of her life because she didn’t do “the right thing,” according to the experts.

 

So we really ought to look into theories that don’t work, and science that isn’t science.

 

I tried to find a principle for discovering more of these kinds of things, and came up with the following system.  Any time you find yourself in a conversation at a cocktail party—in which you do not feel uncomfortable that the hostess might come around and say, “Why are you fellows talking shop?’’ or that your wife will come around and say, “Why are you flirting again?”—then you can be sure you are talking about something about which nobody knows anything.

 

Using this method, I discovered a few more topics that I had forgotten—among them the efficacy of various forms of psychotherapy.  So I began to investigate through the library, and so on, and I have so much to tell you that I can’t do it at all.  I will have to limit myself to just a few little things.  I’ll concentrate on the things more people believe in.  Maybe I will give a series of speeches next year on all these subjects.  It will take a long time.

 

I think the educational and psychological studies I mentioned are examples of what I would like to call Cargo Cult Science.  In the South Seas there is a Cargo Cult of people.  During the war they saw airplanes land with lots of good materials, and they want the same thing to happen now.  So they’ve arranged to make things like runways, to put fires along the sides of the runways, to make a wooden hut for a man to sit in, with two wooden pieces on his head like headphones and bars of bamboo sticking out like antennas—he’s the controller—and they wait for the airplanes to land.  They’re doing everything right.  The form is perfect.  It looks exactly the way it looked before.  But it doesn’t work.  No airplanes land.  So I call these things Cargo Cult Science, because they follow all the apparent precepts and forms of scientific investigation, but they’re missing something essential, because the planes don’t land.

 

Now it behooves me, of course, to tell you what they’re missing.  But it would he just about as difficult to explain to the South Sea Islanders how they have to arrange things so that they get some wealth in their system.  It is not something simple like telling them how to improve the shapes of the earphones.  But there is one feature I notice that is generally missing in Cargo Cult Science.  That is the idea that we all hope you have learned in studying science in school—we never explicitly say what this is, but just hope that you catch on by all the examples of scientific investigation.  It is interesting, therefore, to bring it out now and speak of it explicitly.  It’s a kind of scientific integrity, a principle of scientific thought that corresponds to a kind of utter honesty—a kind of leaning over backwards.  For example, if you’re doing an experiment, you should report everything that you think might make it invalid—not only what you think is right about it: other causes that could possibly explain your results; and things you thought of that you’ve eliminated by some other experiment, and how they worked—to make sure the other fellow can tell they have been eliminated.

 

Details that could throw doubt on your interpretation must be given, if you know them.  You must do the best you can—if you know anything at all wrong, or possibly wrong—to explain it.  If you make a theory, for example, and advertise it, or put it out, then you must also put down all the facts that disagree with it, as well as those that agree with it.  There is also a more subtle problem.  When you have put a lot of ideas together to make an elaborate theory, you want to make sure, when explaining what it fits, that those things it fits are not just the things that gave you the idea for the theory; but that the finished theory makes something else come out right, in addition.

 

In summary, the idea is to try to give all of the information to help others to judge the value of your contribution; not just the information that leads to judgment in one particular direction or another.

 

The easiest way to explain this idea is to contrast it, for example, with advertising.  Last night I heard that Wesson Oil doesn’t soak through food.  Well, that’s true.  It’s not dishonest; but the thing I’m talking about is not just a matter of not being dishonest, it’s a matter of scientific integrity, which is another level.  The fact that should be added to that advertising statement is that no oils soak through food, if operated at a certain temperature.  If operated at another temperature, they all will—including Wesson Oil.  So it’s the implication which has been conveyed, not the fact, which is true, and the difference is what we have to deal with.

 

We’ve learned from experience that the truth will out.  Other experimenters will repeat your experiment and find out whether you were wrong or right.  Nature’s phenomena will agree or they’ll disagree with your theory.  And, although you may gain some temporary fame and excitement, you will not gain a good reputation as a scientist if you haven’t tried to be very careful in this kind of work.  And it’s this type of integrity, this kind of care not to fool yourself, that is missing to a large extent in much of the research in Cargo Cult Science.

 

A great deal of their difficulty is, of course, the difficulty of the subject and the inapplicability of the scientific method to the subject.  Nevertheless, it should be remarked that this is not the only difficulty.  That’s why the planes don’t land—but they don’t land.

 

We have learned a lot from experience about how to handle some of the ways we fool ourselves.  One example: Millikan measured the charge on an electron by an experiment with falling oil drops and got an answer which we now know not to be quite right.  It’s a little bit off, because he had the incorrect value for the viscosity of air.  It’s interesting to look at the history of measurements of the charge of the electron, after Millikan.  If you plot them as a function of time, you find that one is a little bigger than Millikan’s, and the next one’s a little bit bigger than that, and the next one’s a little bit bigger than that, until finally they settle down to a number which is higher.

 

Why didn’t they discover that the new number was higher right away?  It’s a thing that scientists are ashamed of—this history—because it’s apparent that people did things like this: When they got a number that was too high above Millikan’s, they thought something must be wrong—and they would look for and find a reason why something might be wrong.  When they got a number closer to Millikan’s value they didn’t look so hard.  And so they eliminated the numbers that were too far off, and did other things like that.  We’ve learned those tricks nowadays, and now we don’t have that kind of a disease.

 

But this long history of learning how to not fool ourselves—of having utter scientific integrity—is, I’m sorry to say, something that we haven’t specifically included in any particular course that I know of.  We just hope you’ve caught on by osmosis.

 

The first principle is that you must not fool yourself—and you are the easiest person to fool.  So you have to be very careful about that.  After you’ve not fooled yourself, it’s easy not to fool other scientists.  You just have to be honest in a conventional way after that. 

 

I would like to add something that’s not essential to the science, but something I kind of believe, which is that you should not fool the layman when you’re talking as a scientist. I’m not trying to tell you what to do about cheating on your wife, or fooling your girlfriend, or something like that, when you’re not trying to be a scientist, but just trying to be an ordinary human being.  We’ll leave those problems up to you and your rabbi.  I’m talking about a specific, extra type of integrity that is not lying, but bending over backwards to show how you’re maybe wrong, that you ought to do when acting as a scientist.  And this is our responsibility as scientists, certainly to other scientists, and I think to laymen.

 

For example, I was a little surprised when I was talking to a friend who was going to go on the radio.  He does work on cosmology and astronomy, and he wondered how he would explain what the applications of this work were.  “Well,” I said, “there aren’t any.”  He said, “Yes, but then we won’t get support for more research of this kind.”  I think that’s kind of dishonest.  If you’re representing yourself as a scientist, then you should explain to the layman what you’re doing—and if they don’t want to support you under those circumstances, then that’s their decision.

 

One example of the principle is this: If you’ve made up your mind to test a theory, or you want to explain some idea, you should always decide to publish it whichever way it comes out.  If we only publish results of a certain kind, we can make the argument look good.  We must publish both kinds of result.  For example—let’s take advertising again—suppose some particular cigarette has some particular property, like low nicotine.  It’s published widely by the company that this means it is good for you—they don’t say, for instance, that the tars are a different proportion, or that something else is the matter with the cigarette.  In other words, publication probability depends upon the answer.  That should not be done.

 

I say that’s also important in giving certain types of government advice. Supposing a senator asked you for advice about whether drilling a hole should be done in his state; and you decide it would he better in some other state.  If you don’t publish such a result, it seems to me you’re not giving scientific advice.  You’re being used.  If your answer happens to come out in the direction the government or the politicians like, they can use it as an argument in their favor; if it comes out the other way, they don’t publish it at all.  That’s not giving scientific advice.

 

Other kinds of errors are more characteristic of poor science.  When I was at Cornell.  I often talked to the people in the psychology department.  One of the students told me she wanted to do an experiment that went something like this—I don’t remember it in detail, but it had been found by others that under certain circumstances, X, rats did something, A.  She was curious as to whether, if she changed the circumstances to Y, they would still do, A.  So her proposal was to do the experiment under circumstances Y and see if they still did A.

 

I explained to her that it was necessary first to repeat in her laboratory the experiment of the other person—to do it under condition X to see if she could also get result A—and then change to Y and see if A changed.  Then she would know that the real difference was the thing she thought she had under control.

 

She was very delighted with this new idea, and went to her professor.  And his reply was, no, you cannot do that, because the experiment has already been done and you would be wasting time.  This was in about 1935 or so, and it seems to have been the general policy then to not try to repeat psychological experiments, but only to change the conditions and see what happens.

 

Nowadays there’s a certain danger of the same thing happening, even in the famous field of physics.  I was shocked to hear of an experiment done at the big accelerator at the National Accelerator Laboratory, where a person used deuterium.  In order to compare his heavy hydrogen results to what might happen to light hydrogen he had to use data from someone else’s experiment on light hydrogen, which was done on different apparatus.  When asked he said it was because he couldn’t get time on the program (because there’s so little time and it’s such expensive apparatus) to do the experiment with light hydrogen on this apparatus because there wouldn’t be any new result.  And so the men in charge of programs at NAL are so anxious for new results, in order to get more money to keep the thing going for public relations purposes, they are destroying—possibly—the value of the experiments themselves, which is the whole purpose of the thing.  It is often hard for the experimenters there to complete their work as their scientific integrity demands.

 

All experiments in psychology are not of this type, however.  For example, there have been many experiments running rats through all kinds of mazes, and so on—with little clear result.  But in 1937 a man named Young did a very interesting one.  He had a long corridor with doors all along one side where the rats came in, and doors along the other side where the food was.  He wanted to see if he could train the rats to go in at the third door down from wherever he started them off.  No.  The rats went immediately to the door where the food had been the time before.

 

The question was, how did the rats know, because the corridor was so beautifully built and so uniform, that this was the same door as before?  Obviously there was something about the door that was different from the other doors.  So he painted the doors very carefully, arranging the textures on the faces of the doors exactly the same.  Still the rats could tell.  Then he thought maybe the rats were smelling the food, so he used chemicals to change the smell after each run.  Still the rats could tell.  Then he realized the rats might be able to tell by seeing the lights and the arrangement in the laboratory like any commonsense person.  So he covered the corridor, and, still the rats could tell.

 

He finally found that they could tell by the way the floor sounded when they ran over it.  And he could only fix that by putting his corridor in sand.  So he covered one after another of all possible clues and finally was able to fool the rats so that they had to learn to go in the third door.  If he relaxed any of his conditions, the rats could tell.

 

Now, from a scientific standpoint, that is an A‑Number‑l experiment. That is the experiment that makes rat‑running experiments sensible, because it uncovers the clues that the rat is really using—not what you think it’s using.  And that is the experiment that tells exactly what conditions you have to use in order to be careful and control everything in an experiment with rat‑running.

 

I looked into the subsequent history of this research.  The subsequent experiment, and the one after that, never referred to Mr. Young.  They never used any of his criteria of putting the corridor on sand, or being very careful.  They just went right on running rats in the same old way, and paid no attention to the great discoveries of Mr. Young, and his papers are not referred to, because he didn’t discover anything about the rats.  In fact, he discovered all the things you have to do to discover something about rats.  But not paying attention to experiments like that is a characteristic of Cargo Cult Science.

 

Another example is the ESP experiments of Mr. Rhine, and other people.  As various people have made criticisms—and they themselves have made criticisms of their own experiments—they improve the techniques so that the effects are smaller, and smaller, and smaller until they gradually disappear.  All the parapsychologists are looking for some experiment that can be repeated—that you can do again and get the same effect—statistically, even.  They run a million rats—no, it’s people this time—they do a lot of things and get a certain statistical effect.  Next time they try it they don’t get it any more.  And now you find a man saying that it is an irrelevant demand to expect a repeatable experiment.  This is science?

 

This man also speaks about a new institution, in a talk in which he was resigning as Director of the Institute of Parapsychology.  And, in telling people what to do next, he says that one of the things they have to do is be sure they only train students who have shown their ability to get PSI results to an acceptable extent—not to waste their time on those ambitious and interested students who get only chance results.  It is very dangerous to have such a policy in teaching—to teach students only how to get certain results, rather than how to do an experiment with scientific integrity.

 

So I wish to you—I have no more time, so I have just one wish for you—the good luck to be somewhere where you are free to maintain the kind of integrity I have described, and where you do not feel forced by a need to maintain your position in the organization, or financial support, or so on, to lose your integrity.  May you have that freedom.  May I also give you one last bit of advice: Never say that you’ll give a talk unless you know clearly what you’re going to talk about and more or less what you’re going to say.
