#### Elaborations


- Fibonacci Ascent: Using the F-Series to take first coding steps
    - Let's guess our way from 1, 1 to 1, 1, 2, 3, 5, 8, 13
    - Now let's print this on our screen: print(1) etcetera
    - The Ascent is where we get progressively more sophisticated
        - a = 1; print(a) and so on
        - a = 1; b = 1; print(a + b)
        - c = a + b; print(c)
        - Now let's use a, b, and c in a ladder (still no looping)
            - ||: a = b; b = c; c = a + b; print(c) :||
        - Would it not be pleasant to create a loop?
            - for d in \[1, 2, 3, 4, 5, 6, 7, 8\]: and so on
        - Digression: Here is a peculiar thing about squares
            - Draw the Fibo spiral
    - To wrap up:
        - We used three variables a, b, c for bookkeeping
        - When we assigned them we did no clobber
            - a = b; b = a is a clobber
            - A swap requires a third variable c
            - A Fibonacci requires a third variable c
        - After we got tired of copying the code we wrote a loop
            - This used a fourth (counting) variable d
            - This used a for-loop: Very useful
            - The square brackets thing is called a List in Python
                - The length of the list is how many times the for-loop ran
    - There is another counting mechanism called range()
        - This allows us to easily repeat a loop 100 times
    - There is an extension where we can input the first two numbers
    - There is an extension idea called Fibonacci ratios
    - Where next? We have a project called Ms. Halfway that is similar to Fibonacci

- Ms. Halfway: Where does she arrive and when?
    - We need five markers: Home, Post Office, Was, Is, Goal
    - At the outset: Was = Is = Home and Goal = Post Office
    - The three rules are (1) Go halfway to Goal (2) Goal > Was (3) Was > Is
    - For an interesting connection: How to see whether a number is divisible by 11.


<br>



- If: An introduction to logic and information by means of four Prisoners
    - 4 Prisoners peel potatoes to make hash
    - One day they are given the opportunity to earn their freedom
    - h h h h facing > numbered 1 2 3 4
    - There is a pool of 3 red hats and 2 black hats
        - When you wear a hat you are unable to see its color
    - When the warden says "begin" you have 30 seconds to raise your hand and declare your hat color
    - First round is R R B B so 1 can say "R"
    - Second round: B B R R
        - 1 says nothing: Their hat could be B or R
        - 2 thinks "If my hat is R: 1 will say B... but 1 is silent"
            -  2 can say "B"
    - Third round: B R R B > 2 can say "R"
    - What about 3 R 3 B?
    - Introduce Python keywords True and False
        - Practical example: if True: print("too True"); else
    - Extension: s = int(input('Enter a number: ')); if s % 2: print('Odd') etcetera
