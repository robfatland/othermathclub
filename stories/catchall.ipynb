```
# First part: The Rule18() function
def Rule18(c0, c1, c2):
    if (c0 == 1 and c1 == 0 and c2 == 0) or (c0 == 0 and c1 == 0 and c2 == 1):
        return 1
    return 0


# Second part: Prepare!
world_size = 20
t = [0]*world_size
t[world_size//2] = 1
print(type(t), 'is the type of t')
print(type(t[0]), 'is the type of the first element of t')
length_of_t = len(t)
print("world size is", world_size, "and length of t is", length_of_t)
print("(they should be the same)")
counter_limit = world_size // 2
counter = 0


# Third part: While-loop with five tasks
while counter < counter_limit:
    # Task 1: Increment counter by 1
    counter = counter + 1
    
    # Task 2: Print the t[] list on the screen
    t_string = ['o' if i == 1 else ' ' for i in t]    
    print(' '.join(t_string))
    
    # Task 3: Create the empty list called "n"
    n = []
    
    # Task 4: Apply Rule 18 across all of list t to fill out list n
    for i in range(length_of_t):
        i0 = i - 1
        i1 = i
        i2 = (i + 1) % length_of_t
        n.append(Rule18(t[i0], t[i1], t[i2]))
        
    # Task 5: Copy the next list n back into the today list t
    t = n[:]
```

## Task

Suppose we want to know how many times each unique letter or character appears in a string `s`. Let's write a function to 
do this called `UniqueCharCounts`. We will pass it a string `s`. 

```
def UniqueCharCounts(s): return [(u, list(s).count(u)) for u in sorted(set(s))]

s = 'xxxabb'

print(UniqueCharCounts(s))
```


Isn't that nice? Here is how it works.


- Notice that `UniqueCharCounts()` is a **function** that returns something. So that *something* should be a list of counts of unique characters, right? 
- Let's imagine as in the example above that we sent the string `xxxabb`. The result should be a list of characters and counts, like this, alphabetized:

```
[ ('a', 1), ('b', 2), ('x', 3) ]
```

- Notice that what is being returned by the function is in square brackets. This is how we declare a list in Python. 
- So whatever is inside the square brackets should generate the list above, again: `[ ('a', 1), ('b', 2), ('x', 3) ]`
    - The good news is: This code takes advantage of the grammar rules of Python (the syntax) to do just this.
- Notice inside the square brackets there are two things: `(u, list(s).count(u))` followed by a for loop `for u in sorted(set(s))`. 
    - This combination of two things does what we hope for: 
        - It runs the for loop and for each run through that loop...
        - ...it generates a tuple like `('a', 1)`. 
- Let's look at what the for loop is doing. The for-loop variable is a dummy variable `u`. Inside the parens is `set(s)`. 
    - This breaks the string `s` apart into individual characters and makes a set out of them. 
    - A set is like a list with one important difference: A set may not have more than one of any of its elements. 
    - So this means that `set(s)` is the set of unique characters in the string `s`. It will be `('x', 'a', 'b')` in our example.
    - What about the `sorted()` function applied to `set(s)`? As you might guess it sorts the elements of the set alphabetically. 
- So all told the for-loop is saying the following: 
    - Take the string `s`. 
    - Break it down as a set of unique characters (which in our example will be `('x', 'a', 'b')` ). 
    - Alphabetize those characters so now the set is `('a', 'b', 'x')`. 
    - Make `u` be a dummy variable that takes on each of the values in that set. 
        - So that means `u` will be first `'a'`, then `'b'`, then `'x'`. 
    - Use this dummy variable `u` to make the tuples `(u, list(s).count(u))`
    - Once you have made all these tuples, return all of them bundled up in a list


Now all we have to do is figure out what that tuple is.


The first part of each tuple is `u`. And we just figured out what that is: It is going to be a sequence of unique characters 
in the string presented in alphabetical order. So the first part of the tuples generated by the for-loop is going to be 
first `'a'`, then `'b'`, then `'x'`. That's what we wanted, so so far so good.



The second part of each tuple is `list(s).count(u)`. Every time this little bit of code runs: It will first generate a list 
of characters from the string `s`. Then it will use the list built-in function `.count()` to count the number of times that 
`u` shows up in the list (which is built from the string `s`).


So for the first run-through of the for loop we have `u` taking on the value `'a'` and the tuple will be `('a', 1)` because 
`list(s).count('a')` will give us `1`: The `'a'` character appears one time in the list made of characters built from the string `s`. 



What is the point? The approach I used here takes advantage of the fact that Python data structures, specifically lists, 
sets and strings, each have their own characteristics and associated methods. So we can often make code that is very 
economical... which may not seem that important I know, but it is, so I'm bothering telling you so. The reason it is 
important is because simple economical code like this is much less likely to be buggy. This saves you time both 
in writing the code and more time in not having to debug it.

# Electrons


What is an electron?
Four answers... first for impatient people, then some historical narrative, then an art-based narrative, and then finally back to the answer for impatient people.

For those impatient for a working understanding
Look around you: Light falls into your eyes and your brain interprets the pattern as a representation of objects or matter. What is this matter? A computer, a chair, maybe some lunch? Back in 1680 or so with Isaac Newton poking around we (the human race) began to get very methodical and systematic about asking 'What is this matter made of really?' The natural course of inquiry was to try and break things down into smaller and smaller pieces; and then ask 'if I understand the pieces do I understand the whole chunk of matter?' This question occupied the intellectual curiosity and stubborn persistence of the human race: From then until now; where a large portion of the puzzles was figured out by the 1920s. This section summarizes the results: Very approximately.

If you pick up a book and a styrofoam box you notice a difference in mass and volume. Those are basic properties of objects. When you let go of the book it falls to the floor; so we say 'there is a force called gravity that operates between objects with mass.' So far so good. Is this everything we need to understand laws of force? No. We need to add another item to our short list of properties of matter: Charge.

When you run a comb through someone's hair a bunch of times and hold the comb up to a balloon there is some stickiness going on apparently not related to gravity. This demonstrates a property of some matter -- but not all matter -- called charge. Also sometimes called electrical charge; so you can see the word electron sneaking in here.

At very small scales matter is made of aggregates of atoms called molecules. Atoms are the building blocks. They in turn are assembled from three smaller building blocks: Protons, neutrons and electrons. The neutrons do not have an electrical charge; the electrons and the protons do. All three have mass. And now here is the curious thing: The mass of the electron is about 1/1836 the mass of the proton; but the charge of the electron is equal to the charge of the proton except that it is of opposite sign. By convention we say the electron has charge -1 and the proton has charge +1. The rule of charge-based force is: Opposite charges attract and like charges repel. So electrons and protons are drawn together.

There is no good reason I know of why electrons are so light in comparison to protons and yet have precisely the same magnitude of charge.

Let's consider the implication of a force of attraction based on electrical charges, where opposites attract. Perhaps when an electron is near a proton they just collapse together to form a neutron that has a total charge of zero. This is a great idea but it does not work out in practice. This paragraph exists to tell you that some of the rules of how things work on the scale of electrons and protons are very different from the rules of how things work on the scale of you and me. There is not much we can do about this; but the rules for small things are quite interesting and fun to learn about so on balance I think it is a win for us that they are different. Anyway one of the small things rules says that electrons and protons will stick together by electrical force of attraction but only so close to one another and no closer. As a result we have a model of atoms where protons gather together in the nucleus or center of the atom creating a point of positive charge and electrons gather about this center (but not at the center) as smears or clouds of negative charge. When seen from a distance an atom with the same number of electrons as protons appears to have no charge: It is electrically neutral.

To make this picture complete we need to add neutrons -- recall they have no charge -- that are also bound up in the nucleus of the atom with the protons. Neutrons have no charge and weigh about the same as protons, maybe 1% heavier. You might reasonably ask: If there are a bunch of protons stuck together at the center of the atom... and they all have positive charge... and like charges repel or push one another away... and if the neutrons are there also but have no charge to speak of... then doesn't this repelling force cause the nucleus to disintegrate or even fly apart?? It is a good question because, yes, protons do push one another away. But atoms do not disintegrate and this is due to another one of the small things rules. This new rule states that there is yet another force called the nuclear force that is much stronger than the electrical force and it pulls the positively charged protons and the un-charged neutrons together. The nuclear force binds protons and neutrons together in the nucleus of the atom. This makes the atom stable and we are now arriving at our model of how atoms work in chemistry. There is one more idea needed.

The final idea concerns the electrons that are not (by the first rule of small things) permitted to exist inside the nucleus of the atom. Like the protons (with positive charge) the electrons (with negative charge) tend to push one another away. The electrons exist as little smears or clouds of negative charge that are attempting to be close to the positively charged nucleus by attraction but far from one another by repulsion at the same time. This leads to some very interesting mathematics where we can derive the shapes of the electron clouds. These electron cloud shapes are given the rather unfortunate (because it is misleading) name atomic orbitals. For the moment I will stick with the descriptive language: Electrons exist in differently shaped cloudy smears around the nucleus of the atom. The cloud smears stick out in different directions. In summary we have negative charge electron cloud smears and positive charge protons in the nucleus. These charges interact to create a kind of lumpy bumpy atomic charge geometry; and as one lumpy bumpy charge smear atom happens to pass near another one: They might just stick together. This creates a very complicated world of atomic interactions that on our large human being scale we call chemistry. So chemistry is in a sense the connection between the world of small things and the world of large things; and it depends entirely upon the finer details of how negatively charged electrons interact between atoms.

# More Electrons


A second look at 'what are electrons?'
The story of electrons is closely bound up with the story of light. Let us return therefore to Isaac Newton's thoughts on the nature of light, around the year 1660. He noticed that a traditional telescope magnifies by bending light through glass lenses; and this caused a problem because the different colors of light would each bend differently. This creates little rainbows around white dots of light (say like stars) as in the image below. This blurriness could use a descriptive name; and since the rainbow-smears are not what we want we can say "It is a mess." And since color is involved we can say "In fact it is a color mess." However people are not happy with short simple terms like this so instead they say "It is a chromatic aberration."

drawing

Newton noticed that reflected light bounced the same no matter its color... the reflection angle did not change from red to yellow to green to cyan to blue. So... he went to some pains to invent and build a new type of telescope from curved mirrors rather than lenses. Today such telescopes are called Newtonian reflector telescopes. This little parenthesis is to emphasize two things: First that Newton saw the question of how light works as a challenge, and second that he was prepared to take on that challenge in every way he could imagine. He persevered by devious plans and schemes to better understand light. He thought about what it means to polish glass, for example, and he carried out a famous experiment on sunlight using prisms.

Now before going further with Sir Isaac: What is the plan here? The plan is to tell three stories spanning from 1670 to 1905. The first concerns Isaac Newton's disagreements with certain people on the nature of light. He did an experiment with prisms and suggested using the word spectrum for the rainbows this produces. The idea of a spectrum leads into the second story, about how spectrums differ one from another; and what can one figure out from careful observation of spectra? (Which do you prefer: spectrums or spectra?) Part two arrives at the predictive nature of good science: 'If my understanding is correct then it must be true that...'

The third story jumps to 1905 to connect electrons and light once more. It involves a fellow employed in the city of Bern in Switzerland as a patent office clerk. His job was to evaluate new inventions... did they deserve a patent? But in his spare time he was able to indulge his curiosity on the nature of light and electricity. Since Newton's time he had the benefit of 250 years of additional thought, observation and experiment to build upon. So those will be the three stories.

Part One: Prism sentences
Newton put some very opaque curtains at his window blocking the light; except where he put a little prism. On a sunny day this prism would break the light of the sun into a rainbow of color on the far wall of his room. This he called a spectrum. Now sunlight is white whereas the spectrum cast upon the wall was red and orange and yellow and green and cyan and blue and violet. How to explain this?

Well prisms and rainbows were well known; so there were already two ideas about the nature of light popular at this time. Notice the word popular: Does this not imply that science is a parade of fads and fashions? Yes! Absolutely true, but with one little asterisk at the end, perhaps like this:

Science is a progression of fads and fashions; of opinions, biases, beliefs, even prejudices*.

And so what about this asterisk? There must be the accompanying asterisk at the bottom of the page, and just what is written after that? Well in 1660 (independent of Newton) a group of people in London decided to form a club with the motto Nullius in verba which is Latin for "Don't take anyone's word for it." The purpose of the club was to start explaining how things worked; and their motto was important as a milestone in the shift to repeatable demonstration as the basis for belief in a particular explanation.

*but remember: Nullius in verba!

Ok so what were these two related fashionable ideas about light? First white is a color. And second light is a phenomenon of waves. Newton through his experiments arrived at different conclusions: That light is composed of particles (corpuscles he said) and that white light is in fact a superposition of our familiar colors and not a color itself.

There is a very nice demonstration of Newton's experiments with prisms here. I highly recommend it; it runs about five minutes. It includes a good example of how to not over-simplify when it discusses recombining a spectrum into white light.

It is easy to miss a small point here: In Newton's experiment can you still explain the results using the old idea that 'white light is a color'? Certainly you can but Newton's explanation has the advantage of being simple and self-consistent. Similarly his conclusion that light is made of particles led to further ongoing debate. The history is fun to read but to conclude our first story let's imprint in our memory this image of Newton as a young man (in his twenties) setting up a prism to cast a solar spectrum on the wall of his darkened dormitory room.

### Overview

Here is a text adventure game as Python code:


```
print('You are outside a castle full of monsters. The entrance is just wide enough for you to squeeze in.')
print('The wind whistles and cold rain is soaking you. Inside you see a fire... but you hear scary sounds.')

action = input('What do you do? ')
if action == 'run away':
    print ('You win! Congratulations on your great fountain of wisdom!')
else:
    print ('The monsters get you... sorry but at least you made a tasty meal.')
````

### How to make this more interesting?

First we can think of `while True:` infinite loops as a way to keep asking the player "What do you do?"
We can use the `break` keyword to get out of this infinite while loop when the player guesses 
the correct action. Our program needs to interpret what the player says they do in relation to
the that correct plan.


One way to interpret the player response is to convert it into
a Python **list** of words. Then we search for key words in that list to decide if the player had
the right idea.


This makes use of two useful things in Python: The **split()** method of strings and the **in** 
function for **lists**. In the example below the player should light a fire in the fireplace because 
this will scare a cat who is sleeping there; and this cat has a clue on her collar. 

```
while True: 
    print('you are in the library. There is a dark fireplace here and a desk with locked drawers.')
    action = input('What do you do?')
    action_list = action.split()
    if 'light' in action_list and 'fire' in action_list: 
        print('Finding matches and wood nearby you light a fire in the fireplace.')
        print('As you do so a terrified cat leaps out of the fireplace and runs across the room.')
        break
    else: 
        print('That doesn't seem to help much.')
```

Notice that the `break` takes us out of `while True:` and on to the next phase of the game. 

### What am I carrying?

Often these text adventures require you to pick up useful items. You can keep these in a separate list called **inventory**.


[Club page](https://github.com/robfatland/pythonbytes/blob/master/README.md)


# PythonBytes Project Ideas

## Curriculum work


This is work we do together to brush up on Python: Both concepts and literacy, knowing "how to speak".


:star: [Falling out of the sky for beginners](https://github.com/robfatland/pythonbytes/tree/master/projects/falling)<BR>
:star: [Rule 18](https://github.com/robfatland/pythonbytes/blob/master/homework/README.md)<BR> 
:star: [Computer Games Of Ancient Lore](https://github.com/coding-horror/basic-computer-games)<BR>
:star: **Turtles** are creatures we can use to draw on the computer screen; very useful<BR>
:star: **Steps** is a simple online game. To play: Check the pinned note on the slack #level-2 channel<BR>
:star: **Mocean** is a more complicated online game that reminds me of ***Among Us***. To play you must write your own app<BR>
:star: **Recursion** (The dictionary defines this **recursion: See *recursion***<BR>
:star: **Solving problems about bugs, ducks, wolves, snails, kids and fish**<BR>
:star: **Creativity from paintings to optical illusions to music**<BR>
:star: **Ocean science: exploring the ocean with robots**<BR>


## Why does this list not mention Python?


Because Python is the tool chest for exploring ideas, specifically ideas we find interesting.
It's about the *ideas*. 


## What about projects?


If you have a project idea: Our job is to help you figure out how to work on it in a meaningful way. 
Let's not make it too easy. Let's not make it too hard.


If you are wondering what project to do: Our job is to help you find it. Below are more ideas.



## Project ideas


Our hope: You already have a sense of the world as mysterious and interesting. We live here; so welcome to this world!


Order of these projects: From easier to advanced. If you are wondering "which ones should I do?"
my answer is "Do them all, of course!" 


:star: Print the first 20 Fibonacci numbers
<BR>
:star: Is it prime?
<BR>
:star: Spiral!
<BR>
:star: Spirograph!
<BR>
:star: Calculate [**pi**](https://github.com/robfatland/pythonbytes/tree/master/projects/pi#pythonbytes-project-in-depth)
<BR>
:star: Draw a tree
<BR>
<BR>
<img src="https://github.com/robfatland/pythonbytes/blob/master/projects/bees/baobab.png" alt="drawing" width="280"/>
<BR>
<BR>
:star: Print nicely the first ten rows of the Pascal triangle together with the sum of each row
<BR>
:star: [**fortune teller**](https://github.com/robfatland/pythonbytes/tree/master/projects/fortune#pythonbytes-project-in-depth)
<BR>
:star: Investigate the number of [consecutive partitions of positive integers](https://github.com/robfatland/pythonbytes/tree/master/projects/consecutive%20partitions)
<BR>
<BR><BR>
 Play Guess A Number From One To One Hundred
<BR>
 Pebble pile game [Nim](https://nrich.maths.org/1209)
<BR>
 Text adventure
<BR>
<BR><BR>
 Draw an optical illusion
<BR>
<BR>
<img src="https://github.com/robfatland/pythonbytes/blob/master/projects/art/opticalillusion.png" alt="o-illusion" width="300"/>
<BR>
<BR>
 Become the next Piet Mondrian (a renowned [**artist**](https://github.com/robfatland/pythonbytes/tree/master/projects/art#pythonbytes-project-in-depth))
<BR>
<BR>
<img src="https://github.com/robfatland/pythonbytes/blob/master/projects/art/mondrian.png" alt="drawing" width="300"/>
<BR>
<BR>
 Use drones to locate [**beehives**](https://github.com/robfatland/pythonbytes/tree/master/projects/bees#pythonbytes-project-in-depth)
<BR>
 Solve the [4 Bugs On The Table](https://github.com/robfatland/pythonbytes/tree/master/projects/bugs#pythonbytes-project-in-depth) problem
<BR>
 Create something [**pointless**](https://github.com/robfatland/pythonbytes/tree/master/projects/pointless#pythonbytes-project-in-depth)
<BR>
 Text [**adventure**](https://github.com/robfatland/pythonbytes/tree/master/projects/adventure#pythonbytes-project-in-depth)
<BR>
 Tic Tac Toe player
<BR>
<BR><BR>
:atom: Play Conway's Life
<BR>
:atom: [**Sudoku solver**](https://github.com/robfatland/pythonbytes/tree/master/projects/sudoku#pythonbytes-project-in-depth)
<BR>
<BR>
<img src="https://github.com/robfatland/pythonbytes/blob/master/projects/sudoku/sudoku.png" alt="drawing" width="280"/>
<BR>
<BR>
:atom: [**online game client**](https://github.com/robfatland/mocean): Explore a new planet
<BR>
:atom: [**knight's tour**](https://github.com/robfatland/pythonbytes/tree/master/projects/knight#pythonbytes-project-in-depth):
<BR>
:atom: [**video game**](https://github.com/robfatland/pythonbytes/tree/master/projects/videogame#pythonbytes-project-in-depth)
<BR>
:atom: [**Chaos Game**](https://github.com/robfatland/pythonbytes/tree/master/projects/fractals-I#pythonbytes-project-in-depth)
<BR>
:atom: [**maze escape**](https://github.com/robfatland/pythonbytes/tree/master/projects/maze#pythonbytes-project-in-depth)
<BR>
:atom: [**fractal art**](https://github.com/robfatland/pythonbytes/tree/master/projects/fractals-II#pythonbytes-project-in-depth)
<BR>
<BR>
<img src="https://github.com/robfatland/pythonbytes/blob/master/projects/fractals-II/juliaset.png" alt="drawing" width="280"/>
<BR>
<BR>
:atom: oceanography data science project I: Basics
<BR>
:atom: oceanography data science project II: Undersea volcano
<BR>
:atom: oceanography data science project III: Humpback whale vocalization
<BR>
<BR>
<img src="https://github.com/robfatland/pythonbytes/blob/master/projects/whales/humpback.png" alt="drawing" width="300"/>
<BR>
<BR>
<BR>
<img src="https://github.com/robfatland/pythonbytes/blob/master/projects/whales/humpback_spectrogram.png" alt="drawing" width="300"/>
<BR>
<BR>
:atom: [Minecraft construction](https://github.com/robfatland/pythonbytes/tree/master/projects/minecraft#pythonbytes-project-in-depth)
<BR>
:atom: Gravity slingshot courtesy of [**Stargirl**](https://github.com/robfatland/pythonbytes/tree/master/projects/stargirl#pythonbytes-project-in-depth)
<BR>
<BR>
<img src="https://github.com/robfatland/pythonbytes/blob/master/projects/stargirl/stargirl.png" alt="drawing" width="340"/>
<BR>
<BR>
:atom: Build structures in [minecraft](https://www.amazon.com/Learn-Program-Minecraft-Transform-Python/dp/1593276702)
<BR><BR>
   
   
## Projects at these Other Websites

   
* Use a search engine to find problem pages like [this one](https://adriann.github.io/programming_problems.html)
* Read about the Euler programming project
* Read about the Rosalind Bio-informatics programming project
* Look at replit.com, python.org, Anaconda, YouTube, ... 



## For Students
   
This section gives you, the student, some background and context for the club. 

   

### Coding Lecture

If you really want to learn Python: Try to create a *lecture* or *talk* about how Python works. 
You can present this lecture to yourself, your family, your friends, ... anyone interested in listening.
Here are some possible lecture ideas.

* The act of *coding* translates *real world* thoughts and problems into the *computer world*
* Coding is a process of repetition: We write code, we test it, we fix mistakes, we repeat... until it works
* How to code:
    * Code on paper first (no computer)
    * Write and test small bits of code at a time (do not write huge programs all at once)
    * Program using functions that begin (in Python) with `def` which is short for `definition`
        * When the programs run, these functions are like instruction manuals
    * Program using parameters rather than constants (hard-coding is less flexible than parameters)
    * Use `print()`, `time`, `# comments` and `assert` to make your program transparent and reliable
    * Computer programs have something called **flow of control**... what is that?
    * What are three parts of a **variable**?
    * What does the strange word `bool` mean?
    * Why and how does Python use indenting?
    * How does one escape from `while True`? 
        * Give at least four ways
            * Way 1...
            * Way 2...
            * Way 3...
            * Way 4...
            * Way 5: Throw the computer into a volcano
   
   
   
### Why Projects? 
   
   
Our goal is to help you connect to the world of science, engineering and mathematics via Python. 
Once you know the basics of Python coding you are ready to begin thinking about ideas and
problems; and thinking about science and math. This is Project-based work. A Python project
could end up being a program that is only ten lines of code: What matters is that it does 
something that you think is meaningful.
   
   
Here are some of the specific tools and methods that enable you to do Project work:


* Coding vocabulary and grammer: To teach computers to solve problems 
* Draw figures like the optical illusion below
* Use data structures...
    * This gets us closer to data science and artificial intelligence
* Set up your computer to talk to other computers
    * This is the basis for server-based games like Minecraft and Among Us
* Control physical devices
* See how biology connects to computing
* See how computing enables us to create art and music
* Learn debugging skills: When code does not work the way it should you can fix it.


<img src="https://github.com/robfatland/pythonbytes/blob/master/projects/art/opticalillusion.png" alt="o-illusion" width="400"/>


### The Three Asks of you
   

If you would like to give the club a try there are three "asks" of you, the student. You should talk these over
with your parent or guardian. Can you commit to these? 


#### 1


First ask: You will have to set up some tools that will enable
you to do productive work. At the beginning there are two.
First sign up for a free student account on a 
[website](https://replit.com) and learn to use it. 
Second sign in to the club Slack channel.
Depending how things go there may be some further steps later on. 


#### 2


Build for yourself a **participatory** and **initiative-based** frame of mind. 
Consider a student named *Esmerelda* 
who was in the club a couple of years ago. 
Whenever we were talking over Zoom or in person she would constantly 
interrupt to ask questions. Guess whether this was a good thing or a bad thing.

   
Answer: It was not bad, it was not good, it was ***Awesome***. 
The percentage of times she was made to feel *bad* about asking
questions was Zero Percent. Other students benefitted from her questions.
She was often saying
out loud what they were thinking. 
   
Be ready to interrupt and say what is confusing. Be like Esmerelda. 
Let's be ready to work with and communicate with our guides and our fellow students. 


#### 3
 
The club has no grades and no penalties and no deadlines except for Demo Nights.
   

As a result, day to day, it is easy to just skip it; not do any Python. 
Plus there are hundreds of things in the world that are
always trying to get our attention and our time. Some of these things are great, like learning 
to play the viola. Let's do those cool things! But some of those things are 
watching YouTube videos about throwing watermelons off of bridges. Those 
things are not worth it. Ask #3 is **make time to do Python and do Python**.


Please think 
carefully before you sign up for Python Bytes! If all three of these asks are 
really what you can commit to: Sign up!


How much time? You should agree with
your parent or guardian that you can devote 20 minutes per day for five days every week on 
average. That's close to 2 hours of coding per week, a minimum necessary to make progress. 
Much better for you is to have at least one longer session per week where you don't have 
interruptions. 


In addition to *Slack* and *Replit* you need access to Zoom. It is also nice to have
a reference book such as [Python For Kids](https://nostarch.com/pythonforkids).


### Some coding ideas


* The best place to start work on a computer program is *away* from your computer. Why?
    * Answer: Computer programming requires you to organize ideas. The computer interface can distract you from this. 


* Program for 20 or 30 minutes each day. Why?
    * Otherwise you will forget what you are doing and have to repeat. You will get frustrated.


* Use **`replit`** to start with. Why?
    * It saves your work for you
    * You do not need to install special software


* Later in the year: Use PyCharm or Anaconda on your own computer. Why? 
    * Your Python code will run faster, do cool things, and be easier to debug when it runs on your laptop
 

* What are we going to do when we start?
    * We will do some simple projects together and get to know one another. 




# PythonBytes Project In-Depth


## Art


[Here is a link to this page.](https://github.com/robfatland/pythonbytes/tree/master/projects/art#pythonbytes-project-in-depth)


### Overview


Interested in art and design via computer?  What sort of art might you create? One approach is to look for inspiration in the 
paintings of the 20th and 21st century. You might for example use a search engine to find sources of **abstract art**. 
I found Piet Mondrian...


<img src="./mondrian.png" alt="drawing" width="350"/>


and  


<img src="./abstract.png" alt="drawing" width="400"/>


Optical illusions can be artistic... 


<img src="./opticalillusion.png" alt="o-illusion" width="400"/>


Could you write a Python program to produce something like this?


<img src="./lines.png" alt="drawing" width="300"/>


The Stargirl project produced this image...


<img src="./trajectories.png" alt="drawing" width="400"/>


...while some whimsical engineering produced this...


<img src="./stargirl.png" alt="drawing" width="400"/>


The fractals-I and fractals-II projects will also produce some interesting images for you to enjoy.


<img src="./julia.png" alt="drawing" width="300"/>





```
from turtle import Turtle

square, grid = 120, 139
halfsquare = square/2
nSubSquares = 3
nHalfSubSquares = (nSubSquares - 1) / 2

t = Turtle()
t.speed(1000)
t.setpos(0,0)
t.up()

for i in range(nSubSquares):
  for j in range(nSubSquares):
    cornerX = (i-nHalfSubSquares)*grid - halfsquare
    cornerY = (j-nHalfSubSquares)*grid + halfsquare
    t.setpos(cornerX, cornerY)
    t.setheading(0)
    t.down()
    t.forward(square)
    t.right(90)
    t.forward(square)
    t.right(90)
    t.forward(square)
    t.right(90)
    t.forward(square)
    t.up()

radius = 300
ntheta = 66
dtheta = 180. / float(ntheta)

for i in range(ntheta):
  theta = float(i)*dtheta
  t.setpos(0., 0.)
  t.setheading(theta)
  t.down()
  t.forward(radius)
  t.up()
  t.setpos(0., 0.)
  t.down()
  t.forward(-radius)
  t.up()

t.hideturtle()
```
