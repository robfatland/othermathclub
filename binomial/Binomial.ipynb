# Egon and the Binomial Theorem


## An Egon Adventure


### "Egon is stumbling home in bluing and rags..."


Egon is a fictional painter with extraordinary perseverence: Once he gets an idea in his head 
he will keep after it, for much longer than most people.


Most mornings around 9:30 or so Egon has a coffee in the piazza, the public square 
at the center of the town of Monte Carlo on the Mediterranean coast. 


'A television said "Binomial Theorem" at me last night", observes Egon. "Do you have any notion of what 
that means?"


"Certainly!" responded his friend Kyd (who was a mathematician). She wrote down on a fresh sheet of paper ${( \alpha  +  \beta )}^n$.


Egon stares at this. "What is the $\alpha$?"


"Oh that's the beauty of the binomial theorem: $\alpha$ can be anything you like."


Egon blinks... and stares for a few more moments. "And $\beta$? I suppose $\beta$ can be anything I like as well?"


"Precisely!" responded Kyd. "Now you're getting the hang of it!"


Egon stared further, becoming both nonplussed and dogged. He continued: 'Is $n$ also not important
and also anything I like as well?' 


'Oh no no no no' Kyd replied. $n$ is *very* important. Today, in this conversation, 
$n$ is a positive integer... or it can be zero. So $n$ is zero or 1 or 
2 or 3 or 4... so on like that.'


'So $\alpha$ is anything, $\beta$ is anything, they don't matter anyway, and $n$ is 0, 1, 2, 3, ...'


'You got it.'


Why is there no *equal* sign? Isn't this math? I recall there was supposed to 
be an *equal* sign somewhere when it is math.'


'Yes there is an equals sign. I didn't draw it yet.'


'Well so what does it equal?' asks Egon. 


Kyd says,
'Suppose $n$ is 2. Then you have $\alpha \ + \ \beta$ raised to the second power, that is $\alpha + \beta$ squared.'


She wrote $(\alpha + \beta)^2$.


This is...


She completed the equation: $(\alpha + \beta)^2 = \alpha^2 \ + \ 2 \cdot \alpha \cdot \beta \ + \ \beta^2$.'


If we make $n$ larger, like 3 or 4, the result is messier.
