### The Twelth Parolee

One day a prison warden decides to parole some prisoners. She devises a system and follows it very carefully.
There are $n$ numbered prison cells $1, 2, 3, \dots, n$: All locked. She goes to cell $1$ and flips the lock: 
It is now unlocked. She proceeds down the row flipping the lock of every one-th cell. In other words she
does not skip any cells... and now they are all *unlocked*. We can say she flipped the lock of cell $1$ and 
every cell that is a multiple of $1$... which is all of the $n$ cells since all integers are multiples of $1$. 


She returns to the start and flips the lock on cell $2$ and all the cells that are multiples of $2$. 
(If you are keeping track these even-numbered cells are now all *locked* again.)
She goes back to the beginning and flips the lock of cell 3 and the locks of all cells that are 
multiples of 3. (Notice cell $6$ is now *unlocked*.) 
She repeats this procedure with 4, with 5, with 6, ... all the way up to finally $n$. 


When she is done some of the cells are locked and some are unlocked. 
All the prisoners in the unlocked cells are given parole. 
For example the door of cell $1$ was flipped only once; so the prisoner in that cell is the first parolee. 
What is the cell number of the 12th parolee?


### The abstract nature of $n$


We were not told what $n$ is in the above problem. This is *abstract*
thinking. 


Some problems will introduce an abstract $n$ and then ask
what $n$ actually is. Perhaps $n=23$. In such cases we might think of $n$ as a fellow 
wearing a t-shirt with the number $23$ printed on it; but he wears a jacket over it. 
We must use clues and logic to determine: Ah, $n$'s t-shirt reads $23$. 


In other cases we never find out what $n$ is. $n$ can be *anything*. In the 12th parolee problem
it is fair to say $n$ must be a positive integer. It could be $1$ or $2$ or $500$ but it is abstract,
undetermined, ambiguous, open-ended... could be anything. Of course there must be at least as many 
cells as the cell number of the 12th parolee since there *is* a twelth parolee. 


It is important (in mathematics) to develop a sense of when we are being abstract and when things are
becoming definite. A square of side 2 meters has an area of 4 square meters. A square of side $a$ has an 
area of $a^2$ and never mind the units whether they are meters or miles. 



### Professor Smullyan

Professor Smullyan was a mathematician who lived a long and interesting life. He left us a legacy of 
books filled with logic problems (although warning: Some are a bit dated). 
Eight of Professor Smullyan's logic problems are reproduced here for your consideration. 
All are from the book _What is the Name of this Book?_. 


On the island of knights and knaves live persons who are one of two types: Either *knight* or *knave*. 
Knights will always tell the truth and knaves will always lie.
Suppose you arrive on this island and encounter a person
who says 'I am a knight.' What are they? Answer: You do not know because both a knight and a knave would say this. 
However suppose they say 'I am a knight or I am a knave.' Can you determine what they are now? 
What if they say 'I am a knight *and* I am a knave'? 
What question could you ask to determine if a person is a knight or a knave?
