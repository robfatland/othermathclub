## Challenging Logic Problems

Certain students have asked for challenging problems involving logic. 
Here are a few I'm aware of... 

### The *perfect logician* premise

These problems often involve persons participating. I use the term *mathematician* to indicate someone
who is instantly able to reach correct logical conclusions from information given to them. This person
is sometimes called a *perfect logician*. This premise allows you to avoid thinking about trick answers.
These problems are about application of logic, not finding trick answers. 

### Two numbers and two mathematicians

Two mathematicians $S$ and $P$ are told that integers $x$ and $y$ have been chosen such that 
$1 < x < y$ and $x+y < 100$.

$S$ is given the value $x+y$.
$P$ is given the value $xy$.  

They have the following conversation (all true statements):

**P**:  I cannot determine the two numbers.

**S**:  I knew that.

**P**:  Oh really? Well now I know what they are.

**S**:  Oh really? Well now so do I.

What are $x$ and $y$?

### Conway's Wizards


This problem is similar to the one above.


Last night I sat behind two wizards on a bus, and overheard the following: 


A: “I have a positive integral number of children, whose ages are positive integers, 
the sum of which is the number of this bus, while the product is my own age.” 


B: “How interesting! Perhaps if you told me your age and the number of your children, 
I could work out their individual ages?” 


A: “No.” 


B: “Aha! AT LAST I know how old you are!”


Now: What was the number of the bus?

### Prisoners and hats


This problem concerns prisoners who must determine the color of hats placed on their heads that they
can not see. The hats are red or black. While each prisoner is unable to see the color of his or her 
own hat some prisoners can see the hats of other prisoners. 


Prisoners are not allowed to speak except to declare the color of the hat on their head. 


The important point is that the prisoner may say 'my hat is red' or 'my hat is black' but they must otherwise 
remain silent. In some cases a prisoner is forbidden from speaking at all regardless of what they know. 
This is explained further below. If a prisoner correctly identifies their hat color all of the prisoners
are released. We are not interested in guessing here; only in cases where prisoners can be certain of 
their hat color.


Let's define a particular prisoner-hat problem using three (or four) numbers: $(p, b, r \left[,q\right])$.


- $p$ is the number of prisoners
  - they are placed in a line of chairs facing to the left
  - they are numbered $1, 2, 3, \dots, p$ from left to right
  - as they face to to the left: $2$ can see the hat of $1$ but not of $3$. $3$ can see the hats of $1$ and $2$, and so on
  - ...therefore prisoner $p$ can see the other $p-1$ prisoners' hats


- $b$ is the number of *black* hats in the hat bin


- $r$ is the number of *red* hats in the hat bin
  - hats are placed on prisoner's heads from the bin
  - some hats may be left over and not placed on anyone's head
  - all of the prisoners know how many hats of each color are in the bin
  
  
- $q$ is optional: The number of prisoners starting at the far right who may not speak
  - prisoners who do not speak may realize their own hat color but everyone knows they are prevented from speaking
  - if a number is not given for $q$ then $q=0$ and anyone may speak

Now a problem can be phrased 'Given a particular (p, b, r, q) will the prisoners go free?' 


Examples


For (1, 1, 0) the prisoner immediately says 'my hat is black' since there is only one (black) hat. He goes free.


For (1, 1, 0, 1) the prisoner may not speak, alas. 


For (1, 1, 1) the prisoner's hat is either black or red so he has no way of being certain. He can only guess but
this is not allowed, alas.


For (2, 1, 1) we have two prisoners. Prisoner 1 sees nothing and is silent. Prisoner 2 sees a hat of one color (red or black)
and immediately says he is wearing a hat of the opposite color (black or red) since there are only two hats. 


For (2, 2, 0) either prisoner can announce that he is wearing a black hat. Remember that they both know how many
hats there are of each color in the hat bin.


Now we make the problems more complicated. Problems are solvable by the prisoners only if they are 
solvable for any and every possible sequence of hats. 


What about (3, 2, 1)? This is easily solved by $3$. What about (3, 2, 2) where one hat is left over? 


(5, 3, 2) is easily solved by prisoner $5$... but what about (5,3,2,1)? In this case prisoner $5$ may not speak.


Is (6, 10, 1) solvable? 


If you set yourself some additional problems and solve them you may start to develop an idea of which
hat problems are solvable. 
The abstract and difficult problem here is actually this: If you have $p$ prisoners and $b=p$ black hats: 
What is the largest number of red hats you can have such that the problem is still solvable?  If you
solve this problem then all prisoner-hat problems become instantly decideable. 


### Eye color problem as given by Randall Munroe, paraphrased

A certain fellow named Randall Munroe has carefully presented the problem given below. 
I have done a little paraphrasing; he gets 100% of the credit for popularizing this problem.
I transcribe the problem here to decouple it from other web content.


*A group of people with assorted eye colors live on an island.*
*No one knows the color of their eyes.* 
*Every night at midnight a ferry stops at the island.*
*Any islanders who have figured out the color of their own eyes then leave the island, and the rest stay.*
*Everyone can see everyone else at all times.*
*Everyone keeps a count of the number of people they see with each eye color (excluding themselves).*
*However they cannot otherwise communicate in any way concerning eye color.* 
*Everyone on the island knows all the rules in this paragraph.*

*On this island there are 100 blue-eyed people, 100 brown-eyed people, and the Guru.*
*The Guru happens to have green eyes.* 
*So any blue-eyed person sees 100 people with brown eyes and 99 people with blue eyes (and one with green).*
*This does not tell that person his or her own eye color.*
*As far as he or she knows the totals could be 101 brown and 99 blue.*
*Or 100 brown, 99 blue, and he/she could have red eyes.*


*The Guru is allowed to speak once, let's say at noon on one day in all their endless years on the island.*
*Standing before the islanders she says:*

***"I can see someone who has blue eyes."***

*Question: Who leaves the island, and on what night?*

*There are no mirrors or other tricks. The answer is based on logic.*
*The Guru is not making eye contact with anyone in particular.*
*She is simply saying "I count at least one blue-eyed person on this island who isn't me."*
*The answer is not "no one leaves."*

### Zebra puzzle 

- There are five houses.
- The Englishman lives in the red house.
- The Spaniard owns the dog.
- Coffee is drunk in the green house.
- The Ukrainian drinks tea.
- The green house is immediately to the right of the ivory house.
- The Old Gold smoker owns snails.
- Kools are smoked in the yellow house.
- Milk is drunk in the middle house.
- The Norwegian lives in the first house.
- The man who smokes Chesterfields lives in the house next to the man with the fox.
- Kools are smoked in the house next to the house where the horse is kept.
- The Lucky Strike smoker drinks orange juice.
- The Japanese smokes Parliaments.
- The Norwegian lives next to the blue house.

Now, who drinks water? Who owns the zebra?
In the interest of clarity, it must be added that each of the five houses is painted a different color, 
and their inhabitants are of different national extractions, own different pets, drink different beverages 
and smoke different brands of American cigarettes. 

One other thing: in statement 6, right means your right.
