### First infinity problem

Let's start with a math problem, first by reminding you about squares and square roots. 

You know that the square of $7$ is $49$: A seven-by-seven square
has an area of $49$. We write $7^2=7\cdot7=49$. We can also go in the other direction, the *inverse*
direction: $49$ has a square root and that is $7$. We write $\sqrt{49}=7$.
Notice $49$ has another square root, $\sqrt{49}=-7$ since $(-7)^2=49$.


The square root of $1$ is $1$. The square root of $4$ is $2$. Therefore the square root of $2$ 
is somewhere between $1$ and $2$. In fact it is approximately $7/5$ because $\left(\frac{7}{5}\right)^2=\frac{49}{25}$
which is close to but a bit less than $2$. The square root of $2$ must be a bit more than $7/5$.  


Now what if a square root is placed inside another square root? For example what about $\sqrt{4+\sqrt{21+\sqrt{16}}}$?
If we work from the right to the left it works out quite sensibly. At the right we have $\sqrt{16}$ and we 
know this is $4$; and so we just keep replacing things as we figure them out. Follow these steps: 


$$\sqrt{4+\sqrt{21+\sqrt{16}}}=\sqrt{4+\sqrt{21+4}}=\sqrt{4+\sqrt{25}}=\sqrt{4+5}=\sqrt{9}=3$$



So it is no problem to put square roots inside other square roots, particularly if they cooperate 
like this. Now we are all set for the problem.


${\Huge \infty}$


Supposing that we define $x$ as follows:


$${\Large x\;=\;}\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\cdots}}}}$$



...then what actual number is $x$ equal to?


The three dots at the end (right side) mean that this pattern of square roots and twos continues like 
this forever. Let's write it again with a few more square root signs and a few more twos (and notice
the three dots are still there meaning '...and so on forever...')


$${\Large x\;=\;}\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\cdots}}}}}}}}}}$$

So it is an infinity of twos and square roots. This seems to me an impossible problem at first glance; 
for what could $x$ be equal to if the square roots of $2$ never stop coming? Wouldn't it take infinity
time just to calculate this?


Also if we keep adding square roots of $2$ would it not keep growing forever? Perhaps $x$ is infinity... 



Think about this for a bit, write your thoughts in your notebook... and then if you are still puzzled as to
what $x$ could be: Please allow me to try and convince you that $x$ is not infinite at all; and in fact
$x$ is a very sensible number.

### A Solution To The First Infinity Problem

I notice that a *copy* of $x$ is built into the *definition* of $x$. A *copy* in the sense
of the symbols... and this *copy* of $x$ includes the dots that mean 'and so on forever'. 
Here is what I mean as a story in equations:


$${\Large x\;=\;}\sqrt{\;\;2\;+\;\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\cdots}}}}}}}}$$


Now again with parentheses added for emphasis...


$${\Large x\;=\;}\sqrt{\;\;2\;\;\;+\;\;\;\left( \sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\cdots}}}}}}}\right)}$$


Here is the point: What is inside the parentheses is actually $x$... so this becomes


$${\Large x\;=\;}\sqrt{{\Large \;2\;\;+\;\;(x)}}$$


I use *symbols* to convince you that $x$ appears inside its own definition: $x=\sqrt{2+x}$.
Now the infinite square roots and twos and the three dots have vanished into the symbol $x$.

Now let's make a couple guesses as to what $x$ might be... perhaps $x$ is $0$ or $1$ or $2$ or $3$. 

- $x = 0$ does not work because $0\ne\sqrt{2 + 0}$. 
- $x = 1$ does not work because $1\ne\sqrt{2 + 1}$. 
- $x = 2$ works great: $2 = \sqrt{2+2}$. 
- $x = 3$ does not work because $3\ne\sqrt{2 + 3}$. 


You can look for other solutions if you like. There *is* another. $\boxdot$


<br><br><br>${\Huge \infty}$

### Solution

The first amino acid is a choice of one in $20$. The second one is also a choice of 1 in 20 so there are
$20 \cdot 20 = 400$ ways of picking the first two amino acids. Each additional amino acid
is a choice of one among 20; so this multiplies the number of possibilities by another factor
of $20$. Therefore...


$$Possible\;peptides = 20\cdot20\cdot20\cdot20\cdot20\cdot20\cdot20\cdot20\cdot20\cdot20=20^{10}=10,240,000,000,000$$


That is, with only ***ten*** building blocks you can make 10 trillion possible peptides.

### Powers of two in the denominator


Find the sum $S = \frac{1}{2} + \frac{1}{4} + \frac{1}{8} + \frac{1}{16} + \frac{1}{32} + \cdots$


#### Solution


- Draw a square one unit on a side.
- Shade in the left half: This is $1/2$.
- Shade in half of the right half: This is $1/4$.
- Shade in half of what remains: This is $1/8$.


We see two things: First the total shaded area gets
closer and closer to $1$, the area of the square.
Second: Each time we shade in half of the remaining area
there is a little unshaded piece remaining.


Now we ask: If this shading process goes on, never-ending, always smaller: 
How much remains un-shaded? The unshaded area diminishes forever, getting
closer and closer -- as close as you can imagine -- to $0$
so we can say $S=2$. $\boxdot$

### Check the infinite square roots of two solution


Let's calculate the value of the [surd](https://www.mathsisfun.com/surds.html) $x$ if ...


$${\Large x\;=\;}\sqrt{\;\;2\;+\;\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\cdots}}}}}}}}$$


If we ignore the 'dot-dot-dot' forever we can check our solution '$x=2$' *approximately*. 
The cell below lets you do this using a slider. To try this out: Click on the cell, hold down **Shift** 
and press **Enter**. Then drag the slider. Please write down what you find.

````python
from ipywidgets import *
from traitlets import dlink
import numpy as np

def surd_sqrt2(nRootTwos):
    sum = 0.0
    for i in range(nRootTwos): sum = np.sqrt(2.0 + sum)
    if nRootTwos == 1:
        print 'The square root of 2 by itself is', '%.15f' % sum
        return
    print 'Using', nRootTwos, 'terms x is approximately', '%.15f' % sum

interact(surd_sqrt2, nRootTwos=widgets.IntSlider(min=1,max=24,step=1,value=1))
````

### Check the reciprocal powers of two solution

The program below lets you pick how many fractions you add up from $S = \frac{1}{2} + \frac{1}{4} + \frac{1}{8} + \frac{1}{16} + \frac{1}{32} + \cdots$. Again: Run the cell and then move the slider.

````python
from ipywidgets import *
from traitlets import dlink
def reciprocal_powers_of_2(nFractions):
    sum = 0.0
    for i in range(1,nFractions+1): sum += 1.0/float(2**i)
    if nFractions == 1: 
        print 'With just one fraction (being 1/2) the sum of fractions is of course 1/2'
        return
    print 'With', nFractions, 'fractions the sum of fractions is', '%.20f' % sum

interact(reciprocal_powers_of_2, nFractions=widgets.IntSlider(min=1,max=55,step=1,value=1))
````

The program below allows you to calculate the sum of the reciprocals of squares:

````python
from ipywidgets import *
from traitlets import dlink
import numpy as np

def reciprocal_squares(nFracs):
    sum = 0.0
    for i in range(1,nFracs+1): sum += 1.0/float(i*i)
    if nFracs == 1: 
        print 'With just one fraction the sum is of course 1'
        return
    print 'With', nFracs, 'fractions the sum is', '%.20f' % sum
    print '( Professor Euler predicts this eventually reaches', np.pi*np.pi/6., ')'

interact(reciprocal_squares, nFracs=widgets.IntSlider(min=1,max=100000,step=1,value=1))
````

### Friendly introduction


On this page we will take two new approaches to the idea of infinity. 
The first idea has to do with counting... and it really comes down to the idea that infinity is
countable (to infinity) but pieces of infinity are also countable (to infinity). 


Wait... what?


Suppose that you and a friend each started counting numbers up to infinity. Your idea is to find
out what infinity divided by two is equal to. You hatch a plan where *you* count the *even* 
numbers and *she* counts the *odd* ones. 


By the time you get up to $100$ you have counted $50$ numbers and so has she.
Clearly by the time you get to $N$ you each have $N/2$ but since $N$ increases without limit...
neither one of you ever runs out of numbers to count, even or odd. 


You each have an unbounded number of evens and odds to count... so you both decide that 
infinity divided by two is infinity. You write the following equation in your notebook:


$${\large \frac{\infty}{2}\;=\;\infty}$$


You know a little algebra so you multiply both sides by two and write down this result as well:


$${\large \infty \; \times \;2\;=\;\infty}$$


Oh dear. Ok so we get it: Infinity does not behave like a number. 
It is definitely one level stranger than numbers.

This means that infinity as an idea is ok to work with. But infinity is not a number in the sense
that we are used to. Numbers when you divide them by two produce new numbers, not themselves.


**Therefore we will avoid putting the infinity symbol $\infty$ in equations with equal-signs**


Let's plan to keep using our *idea* of infinity without trying to make it be a number. 
We shall be careful about comparing *numbers* to *infinity*.
We can say $7 > 3$ but not $\infty > 7$ because infinity is
not a number. Is this clear? Muddy? A good idea? A bad one?


Mathematical thinking requires us to be very precise in how we use and understand language.


What else? Well I said there are two ideas on this page. The first is the counting idea...
and the second idea has to do with randomness and information; and we will borrow the number $\pi$ for this. 


To get started on thinking about $\pi$: You may have heard that the digits of $\pi$ begin $3.1415\dots$ and
continue for quite a long ways, in fact forever. This is entirely true and those digits seem to occur at
random... even though nobody is rolling dice to determine what they are. This leads to a paradox of sorts 
and an interesting thought: If the digits of $\pi$ are random then would it be possible to find 400 nines
in a row somewhere???


### Terms

- **infinity**: A set of ideas connected with being *unbounded* or *without limit*.
- **inverse**: The number needed to multiply by to arrive at $1$. The inverse of $2$ is $1/2$.
- **reciprocal**: same as inverse; the reciprocal of $x$ is $1/x$
- countable: A set whose elements can be assigned matching integers $1,\;2,\;3,\;\dots$


### Counting infinity


Ok, so here we go with the first idea about counting. 


Suppose rather than the counting numbers you and your friend arrive in the mythical city of ***Cynosure***.
Like other large cities Cynosure goes on *forever*... only *more* so. Big cities on earth like Los Angeles 
and Sao Paolo and Tokyo and Mumbai only *seem* to go on forever; but Cynosure really *does* go on forever. 


Like the integers: While Cynosure goes on forever it does *start* at a definite place. A street corner, in fact.
Cynosure starts at the corner of **1st Street** and **1st Avenue**.

### Finding a sequence of digits in ${\large \pi}$


$\pi$ is irrational. Does this mean you can not argue with $\pi$? No, matters are much worse than this. 


*Irrational* means that $\pi$ can not be written as a fraction of integers: Integer numerator, integer denominator. 
This in turn means that the digits of $\pi$ never do what the digits of rational fractions do: They repeat forever. 
As an example calculate $1/7$ as a decimal to detect the repeating digits. 
$\pi$ continues forever... its digits are countable and infinite... and so there is a paradox. 
(A paradox is two views of something that both seem true but that seem to disagree at the same time.)


  - There is only one correct version of $\pi$
  - The digits of $\pi$ have nothing to do with one another... they appear to be random


It is as if the universe rolled an infinite number of dice to determine the digits of $\pi$... and 
then those digits were frozen forever.


You can never *predict* the next digit of $\pi$ by looking at the digits that come before it. 
You can *calculate* the next digit of $\pi$ but you can not *guess* what it might be. If you guess
then your chances of getting it right are one in ten, no better.

````python
my_digits = '14159'


# ******

nDigits = 1000000

def iter_file(filename, chunksize=100000):
    with open(filename) as f:
        while True:
            substring = f.read(chunksize)
            if not substring:
                return
            # for Python 3
            # yield from substring
            # for Python 2
            for char in substring:
                yield char

from toolz import sliding_window

def look_for_sequence(filename, sequence):
    s = iter_file(filename)
    for i, group in enumerate(sliding_window(len(sequence), s)):
        if ''.join(group) == sequence:
            print ('found your sequence "{0}" at location {1} in pi'.format(sequence, i))
            return True
    return False      

if not look_for_sequence('/home/nbuser/pi1000000.txt', my_digits):
    print ('Sorry but your sequence "{0}" is not exist in the first million digits of pi.'.format(my_digits))
````

### How many times does it appear


The program below tells you how many times a sequence appears in the first one million digits of $\pi$. 

````python
my_digits = '747'


def iter_file(filename, chunksize=100000):
    with open(filename) as f:
        while True:
            substring = f.read(chunksize)
            if not substring:
                return
            # for Python 3
            # yield from substring
            # for Python 2
            for char in substring:
                yield char

from toolz import sliding_window

def look_for_sequence(filename, sequence):
    nHits = 0
    s = iter_file(filename)
    for i, group in enumerate(sliding_window(len(sequence), s)):
        if ''.join(group) == sequence:
            nHits += 1
    return nHits

count = look_for_sequence('/home/nbuser/pi1000000.txt', my_digits)
print ('Found "{0}" {1} times in the first million digits of pi.'.format(my_digits, count))
````
