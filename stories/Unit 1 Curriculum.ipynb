## Ms. Halfway


Instructors note: I intentionally omit the use of loops here. We can make that a
later lesson in how programs manage to do repetitious tasks. 


### Premise

Set up a student to be Ms. Halfway.
Set up a student to be at the post office.
Set up a student to be at Ms. Halfway's house at 0.0.
These students initially do not have much to do.


Enact the following narrative: 


Miss Halfaway lives in a house at location Zero.
Her local post office is 1 kilometer away at location 1.
She sets out to send a letter at the post office but halfway there she stops. 
'It's too far' she thinks.
Her location is now 0.5.
She decides to return home. 
Halfway there she stops again. 
Her location is now 0.25. 
She turns around again and decides to go back to 0.5... 
And so on. 
Each little journey on her adventure is half as far as the previous journey. 
Eventually her next journey will become microscopic. 
When this happens: Where is she located?


***If you know the answer or have a guess: Write it down! (shhhh!)***


- Our goal is to write a computer program to solve this problem.
- Before we get to writing code: Let's work on calulation.
- We begin by introducing three variables that we will use to keep track of where we are.


#### The first variable `x1`

The location: Where Ms. Halfway is currently. This variable starts out with a value of 0.0 for her house. 

#### The second variable `x2`

The location: Where Ms. Halfway intends to go. 

#### The third variable `x3`

This is what we will calculate: Where Ms. Halfway winds up after attempting to get to `x2`.

#### Let's set these variables writing Python code


```
x1 = 0.
x2 = 1.
x3 = 9999.
```

The last value `9999.0` means "we do not know what this is".


#### To notice


Before Ms. Halfway starts walking from her house towards the post office: 
We do not know where she will wind up. Actually we do but we pretend we do
not. We just know she will get half way there and come to a 
halt. This describes what she is going to do in English... but we are being
tricky Python programmers. We are making a variable that will hold the result 
of her first journey. This will be a calculation... eventually. 


This may sound strange but computer programming is a bit *strange*.


#### Let us enact the situation: Now including variables 


- Variable values are slips of paper in variable cups
- Put Ms. Halfway back at 0.
- Give the variable `x3` to a student. They stand off to the side.
- Make sure variable `x1` has value 0.0 and is held by the student at 0.0
- Likewise `x2 = 1.0` is held by the student at 1.0


***Action!*** Ms. Halfway walks from home to 0.5. Now we can calculate $x3$.


We can write $x3 = \frac{x1 + x2}{2}$. That is an *average*.


Instead for reasons we will get to: We are going to write the math for `x3` from
a sentence in English: 


*Ms. Halfway's new location is her current location ... plus ... half the difference 
between her intended destination and her current location.*


Let's translate this to a calculation of `x3`, her new location. 


$\begin{align}
x3 = x2 + \frac{1}{2} \cdot (x1 - x2) 
\end{align}$


Now that we have this equation to calculate `x3`: Let's do the first two examples.

#### Now we update the variables 


- Her new destination `x2` is where she just left from on the previous journey: `x1`.
    - Have the `x2` student get the `x1` value and then notice: They are there!
- Her new current location `x1` has been calculated: It is `x3`.
    - Have the `x1` student get the `x3` calculated value and then go stand by Ms. Halfway.
- Her new 'where she winds up` has not been calculated yet. We do that later.
    - In the `x3` bucket put 9999.


Let's now write this as Python code:


```
x2 = x1
x1 = x3
x3 = 9999.
```

Why can't we change the order and write...

```
x3 = 9999.
x1 = x3
x2 = x1
```

...???


This is an example of Python programmer thinking: We must be able to
logically follow the steps without clobbering information.


And now here is our resulting program, adding a `print()` so we can see the results. 

````python
x1 = 0.
x2 = 1.
x3 = 9999.

x3 = x1 + (1/2)*(x2 - x1)

x2 = x1
x1 = x3
x3 = 9999.

print(x1)
````

And now let us expand our program to run three journeys in a row...

````python
x1 = 0.
x2 = 1.
x3 = 9999.

x3 = x1 + (1/2)*(x2 - x1)

x2 = x1
x1 = x3
x3 = 9999.

print(x1)

x3 = x1 + (1/2)*(x2 - x1)

x2 = x1
x1 = x3
x3 = 9999.

print(x1)

x3 = x1 + (1/2)*(x2 - x1)

x2 = x1
x1 = x3
x3 = 9999.

print(x1)
````

## Prisoners and hats


### Premise


This exercise does not turn into Python code. It is an introduction to 
information and reasoning. The primary point to make is that under the
right conditions: Silence is itself a form of information. 



There are four prisoners in a row from left to right numbered 1 2 3 4. Prisoner 1 can see 
prisoners 2, 3 and 4. 2 can only see 3 and 4. 3 can see 4. 4 sees nothing. Prisoners are
not allowed to stand up or look around. They are obliged to be still. Let's indicate 
the direction they are facing with little arrows: 


```
1  ->  2  ->  3  ->  4
```


There are five hats: 3 red, 2 black. 


The warden places four of these five hats on the prisoner's heads in such a way that
the prisoners do not know the color of the hat on their head. 


The warden says "begin".


If a prisoner correctly says the color of his or her own hat: All four prisoners go free. 


None of the prisoners is allowed to make any noise to share information about what they see.
They can either call out their hat color or remain silent. 


### Round 1


Make the hats in the above prisoner order be `R R B B`. Now both prisoners 1 and 2 
can confidently say their own hats are both red.


### Round 2


Shuffle the prisoners. Make the hats `R B R B`. Now prisoner 1 can confidently say 'Red' 
but prisoner 2 can not.


### Round 3


Shuffle the prisoners. Make the hats `B R B R`.


Prisoner 1 sees 2 Red and 1 Black. So his or her hat could be either red or black. 


Prisoner 2 upon hearing nothing from Prisoner 1 reasons as follows: 


"If Prisoner 1 saw 2 black hats, he or she would immediately say 'Red'. This has 
not happened. Therefore Prisoner 1 sees 3 hats of which 2 must be red. But I see
only one red hat. Therefore my hat must be red."


Prisoner 2 then says "Red".

## The road to pi-e

Here is an equation for $\Large{\pi}$. See if it works!


$\begin{align}
\pi = \sqrt{6 \cdot (\frac{1}{1} + \frac{1}{4} + \frac{1}{9} + \frac{1}{16} + \frac{1}{25} + \dots)}
\end{align}$


Here is another way to calculate $\pi$ that uses a random number generator. 
For this we will also use a `while` loop. This is one way to make a Python program do something
over and over one million times. 

````python
from random import random
from math import sqrt

counter = 0
in_circle = 0

while counter < 1000000:
    counter = counter + 1
    x = random()
    y = random()
    distance_to_x_y = sqrt(x*x + y*y)
    if distance_to_x_y <= 1.: in_circle = in_circle + 1

my_pi = 4. * in_circle / counter

print(my_pi)
````
