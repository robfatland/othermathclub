### Content

The website is organized into three *sections*: **Curriculum** (**subject tracks**), 
**Problems** and **Appendices**. (Plus the welcome.) We go through these in this 
order and add a bit more on **Extras** and **Notebooks**. The organization and content
proceed from this chain of logic:

- Students need time to learn new math
- Club members are hoping to make the time to learn extra math
- We only have 40 minutes of club time... per week... every week or every other week
  - Therefore students must be willing to spend additional time outside of club meetings
    - Therefore they must have some way of finding new ideas and challenges
      - Therefore we have curriculum pages with *themes*...
      - ...and we have problem pages also organized by themes...
      - ...and we require students to operate their own paper notebooks...
      - ...and we have optional check in and correspondence to help students find ideas and challenges


The bottom line is: The Other Math Club is a resource for learning about mathematics that goes deeper 
than traditional paths (classes, puzzle books, competition club). We want to emphasize that this is 
only going to work for students who are willing to set aside other distractions and focus on those
deep themes that we introduce. We introduce them through club meetings and through these online notebooks.

## Background
### ...on how we think about mathematics


> Mathematics is the exploration of patterns using curiosity, imagination and reason.


The way to have fun with mathematics is to clear away distractions, find interesting
paths to explore, and be ready to make many mistakes along the way. That is: Mathematics is 
participatory in addition to being exploratory, puzzling, imaginative and founded upon reason. 


> Mathematics is a process.


A *process* means not having to answer a question quickly. This is a different emphasis
than what students are used to: Math class, math homework problems, math competitions are all
about getting the *right* answer and getting it *quickly*. Fast accurate math serves many purposes:
It makes methods automatic, it emphasizes that there *are* correct answers, it reinforces
the idea of working with care and double-checking results. However it can also miss
the historical exploratory nature of actual mathematics. 


As a human endeavor mathematics is not about lists of pre-written problems. 
Mathematics means going on extended journeys without knowing where they lead. 
It often involves hiking along a path; guessing the wrong branch and following that 
for a time only to discover that it is a dead end; then backtracking and looking for 
new paths. Progress invariably leads to more questions and more exploration.


To illustrate by example: 


Sue discovers that $x - 1 = 0$ has a *solution* for $x$, namely $x = 1$. Later she 
encounters $x^2 - 1 = 0$. Again $x = 1$ is a solution. But Sue knows that 
$-1 \cdot -1 = 1$. A second solution is available, $x = -1$. 
Now a thought occurs: If $x^1 - 1 = 0$ has one solution and $x^2 - 1 = 0$ 
has two solutions... then perhaps $x^3 - 1 = 0$ has *three* solutions. 
This guess happens to be correct but it leads into quite a complex new landscape.
So there we have it, an opportunity for an extended journey. 


> What about the basics?


The most difficult part of the Other Math Club is simply time: We don't have enough of it.
This is where the student investment of time is important. That helps; but there will always be
unfamiliar details. Some examples; do these look right? 


$$a^2=a\cdot a\;\;\;\;\;\;\;\;\; 
\sqrt{\frac{3}{5}}=\frac{\sqrt{3}}{\sqrt{5}}\;\;\;\;\;\;\;\;\;
\frac{\frac{3}{4}}{5}=\frac{3}{20}\;\;\;\;\;\;\;\;\;
{\left(\frac{\sqrt{3}a}{7}\right)}^2=\frac{3\cdot a^2}{49}$$


Another example, an idea that is not automatic: The pythagorean theorem relates the legs
of a right triangle to its hypotenuse. *Squares* are in the formula but the sides 
of the triangle are *not* squares. We often determine the length 
of a side by taking a square root. 


We will build content into this website to try and help with these things as they come up.


> Finally... why are things *true*?


Math has the notion that things are true when there is a convincing argument for them. Such an argument 
is a *proof*. We talk about proofs in the Other Math Club to understand the difference
between 'I *think* this is true' and 'I *know* this is true and here is *why*'. You may have the 
intuition that there is an infinite supply of prime numbers, for example. This is a good intuition
but why is it correct?
