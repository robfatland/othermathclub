# infinite series

Infinite series are generally sums of fraction-like monomials that converge to some value. 


## $e^x$, sine, cosine, hyperbolics


There is a *magical* nature to sine and cosine: Functions expressable as infinite degree 
polynomials. The alternating signs + - + - between successive terms evidently keep the 
sums constrained on $[-1, 1]$. By setting them all to + we get series definitions for 
hyberbolic sine and hyperbolic cosine:


$$\sinh x = \sum^{\infty}_{n=0} \frac{x^{2n+1}}{(2n+1)!} = x + \frac{x^3}{3!} + \frac{x^5}{5!} + \cdots$$

$$\cosh x = \sum^{\infty}_{n=0} \frac{x^{2n}}{(2n)!} = 1 + \frac{x^2}{2!} + \frac{x^4}{4!} + \cdots$$


It would be pleasant flag to see the plots and singular values of x for these functions.


The exponential function as a series *sans* motivation...


$$\mathrm{e}^{x} = \sum^{\infty}_{n=0} \frac{x^n}{n!} = 1 + x + \frac{x^2}{2!} + \frac{x^3}{3!} + \cdots$$


A caveat: *Don't take anybody's word for it* is a good slogan here. Why on earth would
this be accurate? What is $e$?


In any event the exponential series gives the sine and cosine 
series in a decomposition from $x = i \theta$. Hyberbolic sine and cosine are 
defined:


$$\sinh x = \tfrac12\left(e^x - e^{-x}\right)$$


$$\cosh x = \tfrac12\left(e^{x} + e^{-x}\right)$$

## finite geometric series


$$\frac{1-x^{m + 1}}{1-x} = \sum^{m}_{n=0} x^n\quad\mbox{ for } x \not= 1\text{ and } m\in\mathbb{N}_0\!$$


Variants of the infinite geometric series:


$$\frac{x^m}{1-x} = \sum^{\infty}_{n=m} x^n\quad\mbox{ for }|x| < 1 \text{ and } m\in\mathbb{N}_0\!$$


$$\frac{x}{(1-x)^2} = \sum^{\infty}_{n=1}n x^n\quad\text{ for }|x| < 1\!$$


Binomial series:


$$(1+x)^\alpha = \sum_{n=0}^\infty {\alpha \choose n} x^n\quad\mbox{ for all }|x| < 1 \text{ and all complex } \alpha\!$$


with generalized binomial coefficients


$${\alpha\choose n} = \prod_{k=1}^n \frac{\alpha-k+1}k = \frac{\alpha(\alpha-1)\cdots(\alpha-n+1)}{n!}$$


Square root (A first special case of the Binomial series):
$$\sqrt{1+x} = \sum_{n=0}^\infty \frac{(-1)^n(2n)!}{(1-2n)(n!)^2(4^n)}x^n = 1 + \textstyle \frac{1}{2}x - \frac{1}{8}x^2 + \frac{1}{16} x^3 - \frac{5}{128} x^4 + \dots\text{ for }|x|<1\!$$


Infinite geometric series (special case 2 of the Binomial series):


$$\frac{1}{1-x} = \sum^{\infty}_{n=0} x^n\text{ for }|x| < 1\!$$

### From Apostol... 


What is $\;x \cdot \alpha^x$ for $\alpha \in [0, 1)$ as $x \rightarrow \infty$? 


$$ \; \equiv \ln x = \int_{1}^{x} \frac{1}{t}dt$$


Choose $$\;c > 0$$


For $$\;t \ge 1$$ we have $$\;t^c \ge 1$$ and $$\;\frac{1}{t} \le \frac{1}{t} \cdot t^c = t^{c-1}$$.


Now for any $$\;a > 0, b > 0, x > 1$$:


$$\;0 \;<\; \ln x \; = \; \int_1^x \frac{1}{t} dt \; \le \; \int_1^x t^{c-1} dt \;=\; \frac{t^c}{c}\biggr\vert_1^x \;=\; \frac{x^c-1}{c} \;<\; \frac{x^c}{c}$$


i.e. $$\;0 < \ln x \;<\; \frac{x^c}{c}$$ for $$\;c>0$$


Now raise everything to the $\;b$ power and divide by $\;x^a$ where $\;x^a > 0$.


$$\;\frac{0^b}{x^a} < \frac{(\ln x)^b}{x^a} < \frac{x^{bc}}{c^b} \cdot \frac{1}{x^a} = \frac{x^{bc-a}}{c^b}$$ 


(again:$\;c > 0$)


Take $$\;c = a/2b > 0$$:


$$\;0 < \frac{(\ln x)^b}{x^a} < \frac{x^{-a/2}}{(a/2b)^b} = \frac{1}{(a/2b)^b \; x^{a/2}}$$


$$\;\frac{1}{(a/2b)^b} \lim_{x \rightarrow\infty} \frac{1}{x^{a/2}} \rightarrow 0$$ 

so 

$$\;\lim_{x \rightarrow\infty} \frac{(\ln x)^b}{x^a} \rightarrow 0$$


Now setting 

$$\;K = \left(\frac{a}{2b}\right)^b$$ 


we have


$$\;0 < \frac{(\ln x)^b}{x^a} < \frac{x^{-a/2}}{K}$$ 


for all 


$$\;a, b > 0$$


Let $$\;y = \ln x$$ so that $$\;e^y = x$$. 


$$\;0 < \frac{y^b}{e^{ay}} < \frac{e^{-ay/2}}{K}$$


so 


$$\;\frac{y^b}{e^{ay}} \rightarrow 0$$ 


as 


$$\;y \rightarrow \infty$$



Here of course we are really combining the idea that 


$$\;x \rightarrow \infty$$ 


and

$$\;\ln x$$ 


is unbounded so it's ok to let $$\;y$$ go to infinity as $$\;x$$ goes to infinity.


We can put this into a different form involving an arbitrary number between 0 and 1 by taking a special case where 


$$\;b = 1, a = -\ln \alpha$$ 


for some $$\;0 < \alpha < 1$$ 


so as above 


$$\;a > 0$$


Then


$$\;\frac{y^b}{e^{ay}} = \frac{y}{(e^a)^y} = \frac{y}{(e^{-\ln \alpha})^y} = y\;\left(e^{\ln \alpha}\right)^y = y\;\alpha^y \rightarrow 0$$ as $$\;y \rightarrow\infty$$


In words any constant between zero and one raised to the ''y'' power goes to zero harder than y (raised to any power) goes to infinity, so their product in that limit goes to zero. In poker terminology: Exponentials beat powers which in turn beat straights.


### quantum mechanics

I found this entry in Quora and I thought it was so nice and concise that I would reproduce it here,
paraphrased to shorten it a bit. Full credit for this very pleasant set of remarks to
the author Victor T. Toth, self-described 'IT pro and part-time physicist'.

For simplicity we proceed with a single particle; but this extends to arbitrary systems of particles / fields.

Take the Hamiltonian $H$ of a particle with position $q$ momentum $p$ in a potential field $V$:

$$H=\frac{p^2}{2m}+V(q)$$

This is classical. Now consider a number necessarily on the complex unit circle:

$$\psi = e^{i(p\cdot q - Ht)/\hbar}$$

where $t$ is time. 

(my interjection: This *includes* $H$ in the definition.)

It does not matter what $\psi$ is; only that it is a unit-length complex number that is therefore never zero. 
Therefore we can multiply $H\psi$ without reducing the $H$ equation to $0=0$.

$$H\psi = \frac{p^2}{2m}\psi+V(q)\psi$$

Now consider two identities:

$$p\psi=-i\hbar\nabla\psi$$,

$$H\psi=i\hbar\frac{\partial \psi}{\partial t}$$.

Provided $p$ and $q$ are time/spatial differentiable these hold. The Hamiltonian then becomes

$$i\hbar\frac{\partial\psi}{\partial t}=-\frac{\hbar^2}{2m}\nabla^2\psi+V(q)\psi$$

Which is 'not quite yet' the Schrodinger equation (it is still classical physics). Now suppose we
attempt to *solve* this equation for $\psi$? The equation is homogeneous in $\psi$ so given two 
solutions any linear combination of them is also a solution. In classical physics the multiple solutions
describe possible states of a given particle; but linear combinations of these solutions make no 
sense in classical physics. 

Quoting directly: 

> Quantum physics begins when I make the statement that these linear combinations are, nonetheless, 
> *valid descriptions of reality*. 
> 
> To me, this was a revelation because it pinpoints exactly where classical and quantum physics part ways. 
> It also demonstrates the futility of trying to imagine, envision, or otherwise intuit quantum physics. 
> To the extent that intuition can be used here, it must be abstract intuition unrelated to any classical 
> analogies or visualizations, as they are in fundamental contradiction with the nature of quantum physics. 
> Finally, it was a revelation because I realized that this exact line of reasoning can be used to *derive* 
> the basic equations of both nonrelativistic and relativistic quantum mechanics, as well as quantum 
> field theory, almost with ease, and most importantly, without getting bogged down in the philosophical 
> quagmire of *interpretations*.


My reaction to this is two remarks.


First the more different ways one has of looking at the quantum mechanics formalism the better; because it can
be very mystifying to work from just one set of remarks. Watching some of the Susskind videos on YouTube inspires
a similar sense as the above of 'oh this isn't so difficult!' but of course it is! Quite difficult! So the more
routes up the hill we can follow the better it will come together. 


Second my initial basis for understanding quantum mechanics was the double slit experiment as described by 
Richard Feynman. That level of 'here is what happens empirically' works well for me as an interpretive 
framework; and it maps directly to the remarks above. The nulls in the interference pattern are precisely 
the linear combinations of solutions adding to zero, so to speak, and this is in turn precisely the thing that
is impossible in classical physics.

### p.5, 6 Sixth Proof


While this proof is *very* From The Book it is presented in such an economical matter that the aesthetic element is lost on me for want of understanding the economy. This is my own limit, but as an exercise in storytelling I will present here the essential pieces first, and then recreate the proof without less economy. 


Suppose a set of numbers $C$ is to be generated from two 'constructor' sets of numbers $G$ and $H$ according to the following rules:


* $G$ is a set of the first $k$ primes, i.e. all distinct numbers; and from $G$ we can draw $0, 1, 2, ...$ up to $k$ elements to build a product $a$. If we draw $0$ elements the result is $a = 1$, not $a = 0$. The total number of $a$ values we can produce is (summing a row of Pascal's triangle) $2^k$. For example if $k = 4$ we have $G = \{ 2, 3, 5, 7 \}$ and the values for $a$ are $\{ 1, 2, 3, 5, 6, 7, 10, 14, 15, 18, 21, 30, 35, 42, 70, 210 \}$, 16 possible values.  
* $H$ is a set of squares of integers $\{ 1, 4, 9, 16, 25, ..., q^2 \}$ for some $q$. Such a number in $H$ is denoted here as $b^2$ for the corresponding integer $b$. 


How many numbers can we construct from these two sets? There are $2^k$ values for $a$ and $q$ values for $b^2$> and the two sets are distinct except that both contain $1$. Hence the total number of unique numbers that can be constructed of the form of the form $a \; b^2$ is $2^k \; q$.


Whew... that was a long way to go for such a simple point. Notice that using this method of construction we can cover all the integers (provided we let $G$ and $H$ get big enough). That is, no numbers are skipped over in this construction scheme:

Giving the a and then the b values:

```
1 = 1 x 1
2 = 2 x 1
3 = 3 x 1
4 = 1 x 4
5 = 5 x 1
6 = 6 x 1
7 = 7 x 1
8 = 2 x 4
9 = 1 x 9      (b squared = 3 x 3)
10 = 10 x 1    (a = 2 x 5)
11 = 11 x 1
12 = 3 x 4
13 = 13 x 1
14 = 14 x 1    (a = 2 x 7)

...etcetera...
```


That is, all integers may be written as a product of mutually unique primes (the a factor) times a square integer (which may be 1). If the primes in $a$ occurred twice they would be subsumed into $b^2$. If the primes in $a$ occur an even number of times: The same, just so.


And so now we have a constructivists' view of building the integers in this peculiar format, so that it is complete and so that the number of constructions can be connected to the size of the constructor sets $G$ and $H$.

## Sum of powers (Faulhaber's formula)


Ok so suppose you have some integers 1, 2, 3, 4, ..., n and you raise them to a power and add them up. Say 'to the 5th':


$$S = 1^5 + 2^5 + 3^5 + 4^5 + \cdots + {(n-1)^5} + n^5$$


It doesn't take too long working up from powers 1 and 2 and 3 to realize that this is going to be $n^6$ plus some
other terms, in fact a polynomial like this:


$$S = \frac{\;n^6}{6} + \; a_5 \cdot n^5 + \; a_4 \cdot n^4 + \; a_3 \cdot n^3 + \; a_2 \cdot n^2 + \; a_1 \cdot n + a_0$$


So then what are those $a_i$ coefficients? I went into this question because it helps approximate the 
area of a triangle and then the volume of a pyramid and so on; and of course it leads in the calculus 
to the integral formula where all the smaller powers vanish in the limit. 


**Adapted from the [Wikipedia article on Bernoulli numbers](https://en.wikipedia.org/wiki/Bernoulli_number)**


Bernoulli numbers feature in the expression for the sum of the $m$th powers of the first $n$ positive integers. For $m, n \ge 0$ define


$$S_m(n) = \sum_{k=1}^n k^m = 1^m + 2^m + \cdots + n^m$$


The idea is that you want to sum some $m$th powers for numbers $1, 2, 3, \cdots$ up to $n$.
The conclusion is that this sum can always be written as $\frac{n^{m+1}}{m+1}$ plus some other 
smaller powers of $n$, each multiplied by constants. We can write this abstractly:


$$1^m+2^m+3^m+\cdots + n^m = \frac{n^{m+1}}{m+1} + \alpha_1 \cdot n + \alpha_2 \cdot n^2 + \cdots + \alpha_m \cdot n^m$$

where the challenge is to determine the values of $\alpha_i$. These will vary with $m$ of course.
Hence to add up $1 + 2 + 3 + \cdot + n$ you will square $n$ since $m=1$ and so $m+1=2$; *n-squared*. If we 
take a sum of cubes from 1 to 5 we can casually say


$S_3(5) = 1^3 + 2^3 + 3^3 + 4^3 + 5^3 = 225 = \frac{5^4}{4} + $ some smaller powers of $5$ times some constants.


In fact the formula for summing cubes has $\alpha_1=0$, $\alpha_2=1/4$ and $\alpha_3=1/2$ and so


$S_3(5)=5^4/4+5^3/2+5^2/4=225$ so that works out nicely.


The coefficients $\alpha_i$ in these polynomials are related to the **Bernoulli numbers**
according to ***Bernoulli's formula*** which is also known as ***Faulhaber's formula***


$$S_m(n) = \frac{1}{m + 1} \sum_{k=0}^m \binom{m + 1}{k} B^+_k n^{m + 1 - k}$$


where $\binom{m + 1}{k}$ is the binomial coefficient $(m+1)$ choose $k$.


Some people use a slightly different convention for Bernoulli numbers and give Bernoulli's formula in this way:


$$S_m(n) = \frac{1}{m + 1} \sum_{k=0}^m (-1)^k \binom{m + 1}{k} B^-_k n^{m + 1 - k}$$.


Now there remain only two things: Describing how to derive the Bernoulli numbers and listing some of them. 
Deriving them is not done here; for the moment we will just take them as *given*. Perhaps they fell out of 
the sky. Remember there are two versions of the Bernoulli numbers, the + and - versions. These differ only
by the sign of $B_1^\pm$:


$$B_0^+=1,\;\; B_1^+=\frac{1}{2},\;\; B_2^+=\frac{1}{6},\;\; B_3^+=0,\;\; B_4^+=\frac{-1}{30},\;\; B_5^+=0,\;\; B_6^+=\frac{1}{42},\;\; B_7^+=0,\;\; B_8^+=\frac{-1}{30},\;\; \cdots$$

$$B_0^-=1,\;\; B_1^-=\frac{-1}{2},\;\; B_2^-=\frac{1}{6},\;\; B_3^-=0,\;\; B_4^-=\frac{-1}{30},\;\; B_5^-=0,\;\; B_6^-=\frac{1}{42},\;\; B_7^-=0,\;\; B_8^-=\frac{-1}{30},\;\; \cdots$$


So we can notice that these are identical except for $B_1$ and that all of the odd ones seem to be $0$ for
$B_3$, $B_5$ and so on. (This pattern holds true: All the odd Bernoulli numbers from 3 up are zero.)

````python
# This code compares a sum of powers to the Bernoulli formula for the same
#   There are two parameters 
#     n is the end of the sum: 1, 2, 3, 4, ..., n
#     m is the power that these numbers are raised to; then we sum
# 
# The two methods (summing directly and the Bernoulli formula) agree
#   very well but the Bernoulli formula is subject to rounding errors
#   for larger values of n. 
#

m_low = 1
m_hi = 5
n_low = 1
n_hi = 30

import numpy as np
from scipy.misc import comb

B = [1., -1./2., 1./6., 0., -1./30., 0., 1/42., 0., -1./30.]

def Bernoulli(m, n):
    sum = 0.
    for i in range(n):
        sum += np.power(i+1,m)
    formula = 0.
    for k in range(m+1):
        term1 = 1./(m+1.)
        term2 = np.power(-1., k)
        term3 = comb(m+1, k)
        term4 = B[k]
        term5 = np.power(n, m+1-k)
        formula += term1 * term2 * term3 * term4 * term5
    
    # un-comment the next line to see all the results
    # print m, n, sum, formula, round(sum-formula, 12)
    return np.abs(sum - formula)

nErrors = 0
errorSum = 0.
for m in range(m_low,m_hi+1):
    for n in range(n_low,n_hi+1):
        error = Bernoulli(m,n)
        if error != 0.: 
            nErrors += 1
            errorSum += error
            
print '\nTotal not quite equal = ', nErrors, 'with total absolute sum = ', errorSum, '\n\n'
````
