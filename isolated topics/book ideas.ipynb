### The standard math curriculum

This book takes an augmenting approach to the school math curriculum. That curriculum has its 
values and its perils, chief among the latter is failing to instill students with an appreciation
of the value of the math; and worse, sometimes the opposite. The value of the math curriculum 
lies in transferring skills to the students; and to a degree by osmosis some elements of how to
think and reason. 


An easily overlooked strength of the standard mathematics education in the US is repetitions. There is no 
substitute for using math skills repeatedly until they become second nature. This is something we 
find in the classroom; it is not something we can do in this club/book.

#### Problem


Miss Halfaway never finishes what she sets out to do. One day she decides to walk from Zero to One on the real number line. But after setting out she gets as far as 1/2 and decides to go no further. Looking back she decides to return to Zero but again only makes it halfway, stopping at 1/4. She turns around again and decides to go back to 1/2. But you guessed it: She only makes it half way and stops at 3/8. By continuously reversing direction and proceeding in this manner she eventually arrives at a particular location on the real number line. Where does she end up? What does it mean to 'arrive at a specific location' by making an infinite number of progressively shorter journeys?
