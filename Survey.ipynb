#### Elaborations

- Draw a spiral: Python program structure, Turtle graphics
    - We need some "magic code" plus `t.forward()` and `t.left()`
- Ms. Halfway: Where does she arrive and when?
    - We need five markers: Home, Post Office, Was, Is, Goal
    - At the outset: Was = Is = Home and Goal = Post Office
    - The three rules are (1) Go halfway to Goal (2) Goal > Was (3) Was > Is
    - For an interesting connection: How to see whether a number is divisible by 11.
- Chaos Game
    - Vertices in data structures: 3 at the corners of an equilateral triangle
    - Random choice
    - Use Turtle methods `towards()`, `setheading()`, `distance()`, `forward()`
        - this gets us out of doing the bookkeeping / Pythagoras
    - Use Turtle `tracer(0, 0) / update()` to speed up drawing
    - Hausdorff dimension of a fractal, of a Serpinski gasket
    - Hilbert curve, Koch snowflake, Mandelbrot, Julia, Ship, ...
    - Does the Chaos game extend easily into three dimensions?
    - What about changing the 1/2 parameter? The number of vertices?
    - What about placing vertices at random locations?
- Four Bugs
    - Start with a kinesthetic simulation
    - Coding is more challenging with four turtles
    - Use modulo to occasionally draw smaller squares
- The XOR Universe
    - Advances our idea of data structures
    - Introduce binary logic with operators AND, OR, XOR
    - Determine that the AND and OR universes are uninteresting
    - This is essentially a one-dimensional cellular automata
        - Advanced work: Explore Rule N for N on [0, 255]
- Graphing an L-System
    - axioms are x, y, + and -. The rules are x > y+x+y and y > x-y-x
    - Data structure: String as immutable, indexable
    - Notice the tremendous capacity for experimentation
        - What if the turn angle is not 60 degrees?
        - What if the rules are modified?
- Game Design
    - Flappy Bird, etcetera: Using PyGame, Pranav owns this one
- Music
    - Introducing the machinery: Translate an input sentence to music
    - As a Lindenmayer System: From a 2D space of time and pitch
    - Use music21 to write MIDI files
    - Use [midiano](https://app.midiano.com) to play MIDI files back (note many timbres available)
- Requests
    - The simple start is to create an agent that talks to a cloud API (Gargotron)
    - Next levels of sophistication depends on time-to-build on the Coaches side
    - An earlier concept was a Toroidal Ocean that one navigates in search of treasure.
