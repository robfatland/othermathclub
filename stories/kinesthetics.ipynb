### Activity and Assignment Ideas

- Counting spirals: Sunflowers, daisies, pine cones, ... what else? ...
- Hat problems: Red and Black non-facecards read well to the house when the student holds the card against the back of their head. There is no way they can accidentally see what color they have; but anyone behind them sees it easily.
- A Candy Game
  - Everyone in a circle has a random but even number of wrapped pieces of candy
  - In rounds:
      - Give half your candy to the person on your left
      - Receive a supply of candy from the person your right: simultaneously
      - Now recount your candy supply
      - If you now have an odd number of pieces take an extra from the reserve 
      - Repeat
  - How many pieces of candy were there at the start? End? 
  - Instead: "Eat" the odd pieces to make your pile even again
  - Give half your candy to the person on your left and the other half to the person on your right
  - This sort of stopping problem is related to the ring problem and the Collatz conjecture
    - It also lends itself to a fairly easy programming exercise

### Graphs

- We begin by defining a thing
    - A **thing** is a thing. 
        - It can't be defined without using another word that is just a synonym for *thing*
- A **set** is a collection of things. 
    - Label: $\{v_1, v_2, v_3, ..., v_n\}$)
- A **graph** is two sets: A set of **things** $V$ and a set of pairs of things from $V$ we call $E$
- Act this out to arrive at pictures of graphs
    - One set of students is **things**
    - A non-thing student acts as a pairer of **things**: Touch the arm of both. 'Linker'
        - Now add another linker and so on


$$G=(V,E)=(\{v_1, v_2, v_3, ...\}, \{(v_1, v_2), (v_3, v_5), ...\})$$


- Notice that this *thing* here and this *thing* here are connected. 
    - So the first *thing* has a name $v_1$...
    - ...but we can also point to it and say it is that *thing* connected to this other *thing*. 
    - Just as you know my *name* but you also know me as *acquaintance of Ms.H*.

### Negative One Squared

- Kinesthetic idea is to lay out a 2D coordinate plane / unit circle 20 feet across...
    - ...but pretend it is just a number line at the outset
    - We will do the multiply game: Two people represent two numbers
        - When they are multiplied they have to agree where they go
    - Place someone at 1, someone else at 1, now multiply yourselves!
    - Now two at -1, same
    - Now two people at 1 and -1; then -1 and 1
    - Now let's do some square roots: Stand here at 1, and now go backwards to the same spot to be square roots...
    - Now what? 
- We can redefine a **number** to be **anything that follows the rules of arithmetic**
- Call pairs of regular numbers: **doublized numbers**
    - Now we can place two people on the two axes and have them meet in the plane
- How to add? How to multiply? 
    - Yes, a new number system!
- Treat real numbers as doublized numbers with the second number zero
    - Now we have a basis for -1 x 1
    - Now we have a basis for square root 1, of which there are two; and so on
