### The Fast Baseball

Suppose you are holding a baseball 2 meters off the ground and you let it fall. In the first second
we will say it falls five meters. You let go of the baseball over a 3-meter-deep hole so that after 
one second the baseball hits the bottom of the hole. 

Now you throw the baseball horizontally rather than simply dropping it. 
Regardless of how fast you throw the baseball horizontally: Gravity will still act on the 
baseball vertically. That is: In addition to moving sideways the baseball will also fall 
towards the center of the earth. 
As the baseball moves horizontally away from you it can fall further than two meters: 
The earth curves away as it flies along. 

Determine the height of a baseball above the ground after one second if it is thrown 
horizontally from a height of two meters at 8000 meters per second. 
Do not worry about air resistance or hills and buildings and trees.

The radius of the earth is 6400 kilometers, a perfect sphere. 

Suppose you repeat the process after one second:
Go to the baseball wherever it winds up after one second, about 8000 meters from where you first threw it. 
Throw it again horizontally from that point at the same speed. 
You can see that *horizontal* is a relative term, relative to the local 'up' direction. 
What is the height of the baseball off the ground after another second? How long will it take the 
baseball to reach the ground? Does this situation remind you of anything?
