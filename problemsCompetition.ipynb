1. The longest pro tennis match every played lasted 11 hours and five minutes. How many minutes was this? 

a. 605  
b. 655  
c. 665  
d. 1005  
e. 1105


Solution: 665 minutes. This requires you to know that each hour is 60 minutes long: $11 \cdot 60 + 5 = 665$. 


1. In rectangle $ABCD$, $AB=6$ and $AD=8$. Point $M$ is the midpoint of $\overline{AD}$. 
What is the area of $\triangle AMC$?

a. 12  
b. 15  
c. 18  
d. 20  
e. 24


Solution: 12. The area is $\frac{1}{2}\cdot base \cdot altitude$ and we choose the base to be $AM$ or 4. This
makes the altitude $AB$ or 6.
