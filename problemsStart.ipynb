### $\star\star\;$ Dawn to Dusk Climb

A girl stands at the bottom of a mountain (not Mount Meru) where there is just one trail 
to the summit. She begins at dawn and walks up the trail, arriving at the very top at dusk. She is tired
so she has a little food and goes to sleep. Notice we have no information about how fast or how slow she
goes, when she takes breaks, and so on. We just know she got to the top at dusk. The next morning she 
rises at dawn and takes the same trail back down the mountain, arriving where she started the day before 
at dusk. Dawn and dusk are at the same time of day on these two days by the way. 

Prove that at some point during her hike up and at some point during her hike down the following day she
was at the exact same spot on the trail at the same time of day.

**Note**: This problem is akin to many of the problems in the *Problem* pages in the following way:
The challenge of the problem is not so much about calculating something as it is about figuring out how 
to view the problem. It is like walking around a sculpture and finding the angle to look at it where it 
suddenly makes sense.
