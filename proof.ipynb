**Claim:** There is an infinite number of prime numbers.


**Short Proof**

Assume the contrary, that the number of primes is finite. List them and form the product
of all of them, then add one. The result is a number that is not divisible by any of 
the primes on the list. It is either a new prime or the product of new primes not on the 
list. This contradicts the assumption that the number of primes is finite. Therefore the
number of primes is infinite. $\boxdot$


**Long Proof**


This proof is the same as the one above with more explanation. 


Assume there is a *finite* number of prime numbers, say $n$ of them. 


The list of them is $\;p_1, p_2, \dots, p_n$.
For example $p_1=2$, $p_2=3$ and $p_{10}=29$. 


Multiply all of these prime numbers together. Call the product $A$. Then add 1 to $A$. Call this $B$.


$$A = p_1 \cdot p_2 \cdot p_3 \cdots  p_n$$


$$B = A + 1$$


Now $A$ is divisible by all of the prime numbers. By adding one to $A$ we see that $B$ is 
*not* divisible by *any* of the prime numbers. 


$B$ is either a *new* prime not in the list or it is composite (the product of two or more prime numbers).
If it is the first case, that $B$ is a new prime: Then we contradict our assumption that we had the complete
list of all of the primes. 


If it is the second case, that $B$ is the product of two or more prime
numbers then *they* must not be on the complete list of primes. In either case we have discovered new
prime numbers that were not on our list of all of the primes. 


At this moment we can go in two different directions. One is to say 'The assumption that there is
a finite list of primes produced a contradiction: A number not on the list that is also prime.
Therefore the assumption is false and the list of primes is infinite.'


The second direction is to say 'Well $B$ creates at least one new prime number so let's add that
to the list of all of the primes.' The problem with this is it puts us back in the same position
as before. We can begin again with the new list to make a new version of $B$ -- call it $B_2$ -- 
which will produce still more new prime numbers that are not on our list. And so on: 
The process can be repeated as many times as 
we like. And so again there is no limit to the number of prime numbers. 


That is a proof that there is an infinitude of primes; so we generally end with a little symbol
indicating the proof is done. $\boxdot$

This by the way is called a proof-by-contradiction. We assume that the opposite of our claim is true and 
we show that this assumption leads to an impossible situation: A logical contradiction that tells us our
assumption was wrong. Therefore we go back and say that the *opposite* of our assumption must be correct.


> **Claim**: The number of primes is infinite<br>
> **Assumption**: The number of primes is finite<br>
> **Contradiction**: The assumption leads via logic to an impossible situation (discovery of new primes)<br>
> **Result**: The assumption is false<br>
> **Conclusion**: The opposite of the assumption must be true: The **claim** is proven

### Wilson's theorem

This theorem and its proof is mentioned in one of Ross Honsberger's books. Professor Honsberger 
mentions (Gems II p.31) that credit for the idea should go to Leibniz and the credit for the proof should go to
Lagrange... but sometimes we make mistakes in remembering who did what and then it becomes difficult to fix
them. 


**A number $n$ is prime if and only if $n$ divides evenly into $(n-1)! + 1$.**


**Proof**

Suppose $n$ divides evenly into $(n-1)!+1$. Then $n$ must be a prime because if $n$ was not
a prime we would have an impossible situation. $n$ would be the product of at least two
numbers less than $n$ and greater than $1$. Let's say these are $a$ and $b$ and it does not
matter if they are prime or not. Both $a$ and $b$ appear in the list of factors of 
$(n-1)!$ as follows:


$$(n-1)! \;=\;1\;\cdot\;2\;\cdot\;3\;\cdot\;4\;\cdot\;\dots\;\cdot\;a\;\cdot\;\dots\;\cdot\;b
\;\cdot\;\dots\;\cdot\;(n-3)\;\cdot\;(n-2)\;\cdot\;(n-1)$$

Because both $a$ and $b$ divide evenly into this number (they are factors of it) they do not divide into this number
plus one. This is easy to see by example: $5$ divides into $20$ and therefore does not divide into $21$. So to repeat:
$a$ and $b$ do not divide into $(n-1)!+1$.


So here is the problem, first by example: If $2$ and $3$ multiplied together make $6$ then any number divisible
by $6$ is also divisible by $2$ and $3$. In our proof: If the number $n$ divides into $(n-1)!+1$ and $n$ 
is the product of $a$ and $b$ then both $a$ and $b$ must divide into $(n-1)!+1$ as well. 


To summarize our last two conclusions: If $n$ is not prime then 


- $a$ and $b$ *do* divide into $(n-1)!+1$
- $a$ and $b$ *do not* divide into $(n-1)!+1$


This is impossible. Therefore $n$ is not composite. Therefore $n$ is a prime. 


In short
if $n$ divides into $(n-1)!+1$ then $n$ is prime because a factor of $n$ must but cannot divide into $(n-1)!+1$


That is the first part of the
proof, that *if* $n$ divides into $(n-1)!+1$ then $n$ is prime. The next part is more difficult: We still need
to show that $n$ **must** divide into $(n-1)!+1$. 


Since there is only one even prime ($2$) let us agree that $2$ divides into $(2-1)!+1=2$. Only the odd primes
remain. We could keep working examples ($3$ divides into $2!+1$, $5 divides into $4!+1$ and so on... but this 
is a losing game knowing as we do that the number of primes is infinite.

So taking $n$ to be an odd prime: Does $n$ divide evenly into $(n-1)!+1$ every time? 

### Working through an example 

Let's consider the odd prime $13$. Rather than show that it divides evenly into $12!+1$ (which it does) let's 
take a moment to look at all the numbers from $1$ to $12$ (since $12 = 13 - 1$). I will list pairs of numbers
and their products in this 12-element set.

$1\cdot1=1=0\cdot11+1$

$2\cdot7=14=13\cdot1+1$

$3\cdot9=27=13\cdot2+1$

$4\cdot10=40=13\cdot3+1$

$5\cdot8=40=13\cdot3+1$

$6\cdot11=66=13\cdot5+1$

$12\cdot12=144=13\cdot11+1$

That shows that each member of the set {1, 2, 3, ..., 12} matches with another (in two cases the same)
to product a product congruent to $1$ *modulo* $13$. 


Now we switch to abstract terms. $13$ becomes some prime $p$. Some number in the set 
$\{ 1, ..., (p-1) \}$ becomes $x$. Then $xy \equiv 1(mod\;p)$ is the congruence we are
discussing. The claim is that there is only one way to do all the matchings of $x$ to $y$. 

So here we go. Let's consider all possible pairings $y$ as products for some $x$ in the $\{ 1, ..., (p-1) \}$ set:
$P=\{ x, 2x, 3x, ..., (p-1)x \}$. This new set $P$ is the product set for $x$ where $x$ is paired with other numbers 
of the $\{ 1, ..., (p-1) \}$ set including somewhere in there the product $x^2$ for when $y$ happens to equal $x$.


Suppose two of these numbers in this set of products are congruent mod p; say for different elements $r$ and $s$. 
Then $rx\equiv sx(mod\;p)$ and so $(r-s)x\equiv 0(mod\;p)$. This means that $p$ divides into $(r-s)x$ with no
remainder; so $p$ must divide into $(r-s)$ or $x$ since $p$ is prime. However $(r-s$) is not zero and $x$ is
not zero and so both $(r-s)$ and $x$ are too small for $p$ to divide into them.

Since none of the members of $P$ are congruent to any others modulo $p$ each must have its own modulus. 
Furthermore none of them is divisible by $p$ so their mod $p$ value may not be zero. There are, fortunately,
exactly $(p-1)$ possible remainders and $(p-1)$ elements in $P$ so these must match up. One of them has
a modulus (remainder) of $1$ for the chosen value of $x$. This is a unique matching for that remainder. 


We noted above that $x$ might be matched with itself. Considering that possibility that $x^2\equiv 1(mod\;p)$
we see that $p$ divides evenly into $x^2-1$ so both $x=1$ and $x=(p-1)$ will work. The other values of $x$
will not partner $x$ with itself to achieve congruence to $1$ mod $p$. There is an even number of $x$ values
remaining so every number both *can* and *must* have a matching number to give $xy \equiv 1(mod\;p)$. 


The difficult step is next. 


If $a\equiv 1(mod\;p)$ and $b\equiv 2(mod\;p)$ then $a$ and $b$ are both multiples of $p$ plus $1$. 
Written this way they can be multipled; giving various cross terms that are multiples of $p$ plus that
$1 \times 1 = 1$ term at the end. So their product is *also* congruent to $1$ modulo $p$. So numbering
the $x$ and $y$ matches as $1, 2, ..., (p-1)$ we have


$$x_1y_1\equiv x_2y_2 \equiv \cdots \equiv 1(mod\;p)$$

and

$$x_1y_1\cdot x_2y_2 \cdot \cdots \equiv 1(mod\;p)$$


At this point go back and clean up by throwing out x = 1 and x = (p-1) to arrive at (p-3)/2 matchings
which span the numbers 2, 3, ..., (p-2). This is a finesse that gets us *almost* the factorial we want...


In fact it gets us 2 x 3 x 4 x ... x (p-2) which is congruent to 1 mod p. 


Now we multiply both sides by (p-1) to get (p-1)! congruent to p-1 mod p and 

p - 1 mod p is congruent to -1 mod p... 

Therefore (p-1)! is congruent to -1 mod p and 

(p-1)! + 1 is congruent to 0 mod p, the desired result.
