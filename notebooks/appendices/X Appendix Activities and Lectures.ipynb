- Expert Game
  - logical fallacy
    - *proof by assertion*
    - *proof by appeal to authority*
    - *disproof by questioning character*
    - *proof by assuming the conclusion* 
  - Four participants: E, F, A and N. 
    - E is a genuine Expert who is usually right. 
    - F is a Failed Expert who holds ideas that can be shown to be wrong. 
    - A is Assertive: He claims to be right but his claims are untested. 
    - N is a normal person: She is presented by arguments from E, F and A.
  - Presentation of $C$ will involve some research and preparation by E, F and A. 
    - Each of the three presenters has both a strategy and a method. 
    - The game proceeds in three rounds where the first round serves to set everything up:
    - E is permitted to present $C$ but this is done without proof in the first round.
    - F must present an alternative to $C$ that sounds much more plausible.
    - A must either support E, support F or provide a third view.
  - 1. E, F and A present arguments in a random order (or as chosen by N). 
    - N writes down which argument she believes is correct. 
  - 2. E, F and A provide a supporting statement. 
    - Again N writes down which argument she believes. 
    - She is welcome and encouraged even to change her mind. 
  - 3. E, F and A provide a revision statement. 
    - This statement is of the form "Since my last remarks some new evidence has come to light..." 
    - Followed by the revision or clarification. 
  - To set up the game our goal is to come up with a dubious claim that is nevertheless true. Call this claim $C$.
      - E presents $C$ correctly.  Creative manipulation of $C$ makes the presentation doubtful. 
E is trying to get N to doubt the claim originally (by selecting one of the other two presenters)
and then will try to get N to change their mind in stage 2 or stage 3. 
Therefore: Saying $C$ with a straight face is a good strategy, provided $C$ itself sounds 
outrageous and impossible. Ideally N does not believe E after round 1 but chooses F and/or A. 
In round 2 E is allowed to present a very short justification of why $C$ is true. E gets
maximum score if N changes her mind after round 2. E gets fewer points if N changes her 
mind after round 3. In round 3 E is allowed to make a short supporting statement that 
must cite prior work. That is: Round 3 is an appeal to authority. The reason for this is 
that experts build on established ideas. 
      - F presents $C$ incorrectly; and should go to pains to make the claim sound very reasonable. F would like
N to choose her argument first and continue to support that argument through to the end. F can either 
create false evidence or cite authority or resort to any other possibly unscrupulous means.
      - A presents either support for F or support for N or a third idea. A gets maximum points if they 
can convince N to arrive at a wrong solution at the end of round 3; it does not matter what N
decided in rounds 1 and 2. 
  - Example: E says "Humans are built from mistaken copies of a computer program made from four instructions."

### Monte Carlo


- Two famous mathematicians -- Ulam and Von Neumann -- invented a method of exploring called **Monte Carlo**.
- The method takes advantage of the speed of computers; and it requires the computers to generate random numbers.
- We can simulate Monte Carlo by handing out some dice and recording sums. This is a case where we can also solve exactly.
- We are using this to show how an actual experiment gets progressively closer to a theoretical result.
- This leads to special attention to the words **theory** and **experiment**: What do they mean?

### Graphs

  - It is behind this gate but notice that the gate is locked. So first we unlock the gate with some key ideas. Get it?
- We begin by defining a thing
  - A **thing** is a thing. 
  - (It can't be defined without using another word that is just a synonym for *thing*.)
- A **set** is a collection of things. 
  - (We give them labels to keep them straight: $\{v_1, v_2, v_3, ..., v_n\}$)
- A **graph** is two sets: A set of things we call V and a set of pairs of things from V we call E.


$$G=(V,E)=(\{v_1, v_2, v_3, ...\}, \{(v_1, v_2), (v_3, v_5), ...\})$$


- Together V and E sound terribly abstract and hard to do anything with.

- So it is ok for Pythagoras... just not for bigger exponents, for positive integers

### Negative One Squared


- Suppose I were to ask you what is $1 \times 1$  How about $1 \times -1$? $-1 \times 1$? $-1 \times -1$? 
- We can redefine **number** to be **anything that follows the rules of arithmetic**
- In particular let us take regular numbers in pairs and call them **complicated numbers**
- If we can come up with rules of addition and multiplication then we have an entirely new number system!
- To make progress we will have to agree that there are four ways of writing these new numbers
  - This is rather like graphs that have set notation as well as picture notation
- The four ways are: $(a, b)$, $re^{i\theta}$, 'draw a dot on the plane', $a \;+\;ib$
- Each of these has its uses but we will use the first two to define addition and multiplication
- The final step is to notice that the real numbers are a part of the complex numbers.
  - Therefore we can use complex multiplication to discover that $-1 \times -1 = 1$
