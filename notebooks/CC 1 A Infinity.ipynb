### Second problem: Calculating a surd

<br>

Let's continue to develop our excellent skills at calculating mathematical expressions. This time we will do a *surd*
which includes square roots. In fact square roots that are not integers; like the square root of $11$.  
We begin therefore by reminding ourselves how squares and square roots work, in three steps.

<br>

#### Step 1: Integer squares and square roots

The square of $7$ is $49$: A seven-by-seven square
has an area of $49$. We write $7^2=7\cdot7=49$. We can also go in the other direction, the *inverse*
direction: $49$ has a square root of $7$. We write $\sqrt{49}=7$.
Notice $49$ has another square root, $\sqrt{49}=-7$ since $(-7)^2=49$.

<br>

$\large{\;\;\infty\;\;}$ **Write all the squares for 0, 1, ..., 10 starting with $0^2 = 0$. Then write the corresponding
square roots as in '$\sqrt{49}=7,\;-7$'.**

<br>

#### Step 2: Non-integer squares and square roots (surds)

Next we consider square roots that are not integers.
The square root of $1$ is $1$ (an integer) and the square root of $4$ is $2$ (also an integer). 
What about the square root of $2$? 
That number must be bigger than $1$ and less than $2$. So $\sqrt{2}$ is somewhere between $1$ and $2$. 
In fact it is approximately $7/5$. How do we know?  Because $\left(\frac{7}{5}\right)^2=\frac{49}{25}$
which is close to but a bit less than $2$. Therefore the square root of $2$ must be a bit more than $7/5$.  

<br>

$\large{\;\;\infty\;\;}$ **Make a guess as what *approximately* the square root of three is equal to. $\sqrt{3}\;=\;?$**

<br>

#### Step 3: Square roots inside of other square roots

Now our last step in preparation for the problem. 
Suppose a square root is placed inside another square root... is this possible?
Certainly it is. For example what is $\sqrt{4+\sqrt{21+\sqrt{16}}}$?
If we work from the inside out this is quite sensible. At the right we have $\sqrt{16}$ and we 
know this is $4$; and so we just substitute the value $4$ for $\sqrt{16}$.

<br>

$\large{\;\;\infty\;\;}$ **Copy down these steps: **

<br>

$$\sqrt{4+\sqrt{21+\sqrt{16}}}=\sqrt{4+\sqrt{21+4}}=\sqrt{4+\sqrt{25}}=\sqrt{4+5}=\sqrt{9}=3$$

<br>

So it is no problem to put square roots inside other square roots, particularly if they cooperate 
like this. 

<br>

**Solution** Notice that a *copy* of $x$ is built into the *definition* of $x$. 

<br>

$${\Large x\;=\;}\sqrt{\;\;2\;+\;\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\cdots}}}}}}}}$$

<br>

Now again with parentheses added for emphasis...

<br>

$${\Large x\;=\;}\sqrt{\;\;2\;\;\;+\;\;\;\left( \sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\cdots}}}}}}}\right)}$$

<br>

What is inside the parentheses is simply $x$... so this becomes

<br>

$${\Large x\;=\;}\sqrt{{\Large \;2\;\;+\;\;x}}$$

<br>

$x$ appears inside its own definition: $x=\sqrt{2+x}$.
The infinite square roots and the dots have vanished into the symbol $x$.
Let's make a couple guesses as to what $x$ might be... perhaps $x$ is $0$ or $1$ or $2$ or $3$. 

<br>

- $x = 0$ does not work because $0\ne\sqrt{2 + 0}$. 
- $x = 1$ does not work because $1\ne\sqrt{2 + 1}$. 
- $x = 2$ works great: $2 = \sqrt{2+2}$. 
- $x = 3$ does not work because $3\ne\sqrt{2 + 3}$.

### Check the infinite square roots of two solution


Let's calculate the value of the [surd](https://www.mathsisfun.com/surds.html) $x$ if ...


$${\Large x\;=\;}\sqrt{\;\;2\;+\;\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\sqrt{2+\cdots}}}}}}}}$$


If we ignore the 'dot-dot-dot' forever we can check our solution '$x=2$' *approximately*. 
The cell below lets you do this using a slider. To try this out: Click on the cell, hold down **Shift** 
and press **Enter**. Then drag the slider. Please write down what you find.

````python
from ipywidgets import *
from traitlets import dlink
import numpy as np

def surd_sqrt2(nRootTwos):
    sum = 0.0
    for i in range(nRootTwos): sum = np.sqrt(2.0 + sum)
    if nRootTwos == 1:
        print 'The square root of 2 by itself is', '%.15f' % sum
        return
    print 'Using', nRootTwos, 'terms x is approximately', '%.15f' % sum

interact(surd_sqrt2, nRootTwos=widgets.IntSlider(min=1,max=24,step=1,value=1))
````

### Check the reciprocal powers of two solution

The program below lets you pick how many fractions you add up from $S = \frac{1}{2} + \frac{1}{4} + \frac{1}{8} + \frac{1}{16} + \frac{1}{32} + \cdots$. Again: Run the cell and then move the slider.

````python
from ipywidgets import *
from traitlets import dlink
def reciprocal_powers_of_2(nFractions):
    sum = 0.0
    for i in range(1,nFractions+1): sum += 1.0/float(2**i)
    if nFractions == 1: 
        print 'With just one fraction (being 1/2) the sum of fractions is of course 1/2'
        return
    print 'With', nFractions, 'fractions the sum of fractions is', '%.20f' % sum

interact(reciprocal_powers_of_2, nFractions=widgets.IntSlider(min=1,max=55,step=1,value=1))
````
