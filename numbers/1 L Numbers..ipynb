## Friendly introduction

This track is about numbers, particularly whole numbers like 1 and 2 and 17 and 25. To get started:
$2$ and $17$ are primes and $25$ is composite: It is the product of primes. To avoid confusion we
say that $1$ is neither prime nor composite. "One is one."


Integers are part of the *real* numbers which include fractions like $7/5$ and decimals like 6.211412341. 
If you have a math problem where you need to find a number... and then someone says 'oh by the way the 
answer is an integer'... is this more information? An example. The hypotenuse of a right triangle is five
so what are the two legs? That's a problem with many possible answers. If I now tell you 
both legs of the right triangle are integers: The problem now has but one answer. 


*Number theory* is a topic of mathematics that explores integers starting with the basic idea of primes. 
Since $24$ is the product of $2 \cdot 2 \cdot 2 \cdot 3$ we have a sense that $2$ and $3$ are 
building blocks. They are primes and they can't be further broken down into pieces. Prime numbers
are the *atoms* of numbers. They appear in the list of integers in a strange way that is neither
completely regular nor completely random but a combination of both. 


### Terms

- **Numbers**: Whole numbers: 1, 2, 3, ... Can include zero... Can include negatives ..., -3, -2, -1, 0, 1, 2, 3, ...
- **one**: One is one, one is not prime, one is not composite.
- **prime number**: A positive integer evenly divisible only by $1$ and itself. 
- **composite number**: A positive integer that is the product of two or more prime numbers.

whole numbers. This leads him into a series of dilemmas, for example:

### Watering the horses

Farmer Diophantus owns many horses... and on any given morning a random number of them are standing around
the water trough (next to the water pump) waiting for water. These horses drink exactly one gallon each and 
so one morning when Farmer Diophantus walks out to the trough and finds four horses he must pour out exactly 
four gallons. Less than this and some horses will go thirsty. More than this and the leftover water will 
attract mosquitos. 

Now if Farmer Diophantus had a one-gallon container he could simply fill it four times at the pump, pouring
it each time into the trough. However he is rather careless and has only two containers on hand. One is an 
unmarked three gallon container. The other is an unmarked eight gallon container. Because they are unmarked
they can not be partially filled with any accuracy. That is, they can only be filled to the top, completely
full. Then they may be emptied. For example the eight gallon container could be filled to eight gallons.
Then three gallons could be poured from there into the three gallon container leaving five gallons of water 
in the eight gallon container. 

To be clear: It is ok to pour water on the ground if necessary.

How can Farmer Diophantus get exactly four gallons in the watering trough for the four horses?

### Three Cards Face Down


Not all number problems involve Farmer Diophantus. This one involves three girls who are good at math.


There are three cards face down on a table; let us call them $a$, $b$, and $c$. Each has written on it a unique
positive number that nobody sees. However we know that $a$ is less than $b$ and $b$ is less than $c$ and 
$a + b + c = 13$. Remember that they are all different from one another. Now three girls peek at cards $a$, 
then $c$, then $b$ in that order. (Notice the second girl looks at card $c$, not card $b$.) After each look
they make the following remarks: 


First girl upon looking at card $a$: I know the $a$ card number but not the other two. 

Second girl upon looking at card $c$: I know the $c$ card number but not the other two. 

Third girl upon looking at card $b$: I know the $b$ card number but not the other two. 

What is the number on the $b$ card?

### Abstract versus definite 

One theme of these Notebooks is the power of *abstract thinking*. 


Abstract thinking is a part of mathematical thinking. It is a way of allowing pieces of a puzzle 
to be uncertain or unknown... while still using them to reach a conclusion. This may seem like 
a paradox -- how can one reach definite conclusions using things that are uncertain? -- 
but indeed this is absolutely *central* to the ability to reason.


While you work on the above problem please notice that we have confused the names of the cards $a$ $b$ and $c$ 
with the numbers written on the cards $a$ $b$ and $c$. This is an intentional mistake. We are doing this
to practice being *abstract*. If I say card $a$ is hiding a number then you can think of $a$ as *some number*,
a number that could be anything. It is not definite the way that the number $17$ is definite. Sometimes
we use a symbol like $a$ or $\aleph$ or $\Omega$ to represent a definite number that is simply not yet known.
Then we solve a problem and we say at the end 'Ahhhh so now I know that $\Omega=17$' and that is a solution.


On the other hand it is just as likely that we might use a symbol abstractly without *ever* finding a definite
number that it is equal to. If I use the letter $p$ for the length of one side of a triangle in order to make
a statement about triangles *no matter what $p$ happens to be* then I am being abstract once again; without
needing to turn that abstract idea into some definite number. This is the power of abstract reasoning in 
mathematics: It allows us to be correct about *any* situation, not just one or two situations. *Any* triangle,
for example. Consider the difference between these two statements:


> Sarah and Jean have the right to be happy.
>
> All human beings have the right to be happy.
