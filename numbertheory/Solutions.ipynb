## Introduction to Analytic Number Theory Chapter 1


### Useful ideas:


* 'a divides b' can be written as 'multiplication by some number': 
$a | b \implies b = a \cdot q$. I substitute this implied expression freely.


* A second common substitution from the *gcd* of $a$ and $b$:
$(a, b) = c \implies \exists \; x \textrm{ and } y \backepsilon c = xa + yb$.
This means that $x$ and $y$ exist; not that *any* $x$ and/or $y$ suffice. 
As an intermezzo: There are two 
important qualifiers for any $x$ and $y$, i.e. for an
arbitrary linear combination of $a$ and $b$.
    * A linear combination $x \cdot a + y \cdot b$ can never be a positive 
common factor of $a$ and $b$ *less than* $(a, b)$. 
    * An arbitrary linear combination $xa + yb$ will 
always be a *multiple* of $(a, b)$.<BR><BR>

In a sense, $c=(a,b)$ replaces $1$ with a new atomic unit $c$ that re-scales 
the number line. This atomic nature of the gcd is apparent in un-solvable gallon 
puzzles. For example,  'You have two containers with capacities four gallons 
and six gallons. The horses need precisely five gallons of water.' We may as 
well just give the horses six gallons of water and endure their 
disapprobation.


* The linear combination for a gcd $c$ means that
$x$ and $y$ exist and thus $ax + by$ can be substituted for $c$.

* Corollary: The linear combination that yields $c$ is 
not unique. Suppose $x$ and $y$ are established definite 
values that yield $c$. There are any number of other pairs for 
$x$ and $y$ that yield this same gcd. Take for example $a=15$
and $b=24$. Then $c=3=15\cdot(-3)+24\cdot(2)$. So $x=-3$ and
$y=2$ give $c=3$. For any $n$ we have $n(8a-5b)=0$ so to generalize
we have $x=-3+8n$ and $y=2-5n$ for any choice of $n$ will give $c=3$.


For more on this gcd thought process see the remark in problem 1.2.

##### Intermezzo on divisors and greatest common divisors


Two ideas here. First we have the idea of divisors of two numbers; for example 3 is a 
divisor of both 12 and 30. Second we have the idea that there is a largest element in
the set of common divisors. For 12 and 30 that is 6. 


Theorem 1.2 states that a common divisor of $a$ and $b$ exists and can be 
expressed as a linear combination of $a$ and $b$: $gcd = ax + by$. So far so good.
This is a building block idea: Both $a$ and $b$ are constructed of
common divisor blocks in some quantity. 

It is when we compare $a$ and $b$ that we are forced into using the gcd
building blocks. We can then take integer 
multiples of both $a$ and $b$, extending out until they differ by just one 
gcd block. Can't get differences any smaller than that. Differences that
are multiples of the gcd are fine.


A mistake to make: 'Any arbitrary linear combination of $a$ and $b$
is a divisor of $a$ and $b$.' Take $(3, 7) = 1$ for example: $3\cdot3 + (-1)\cdot7=2$
which is not a common divisor of $3$ and $7$.


A correction of that mistake: 'Any arbitrary linear combination of $a$ and $b$
is a multiple of the gcd of $a$ and $b$.' 


Getting a linear combination of $a$ and $b$ to equal an inferior common divisor
can't be done because $a$
and $b$ are always separated by some multiple of the gcd.



Result: The gcd of $a$ and $b$ can be expressed as a linear combination of $a$
and $b$; or some *multiple* of the gcd can be expressed this way; but nothing
else. Let's take $a=18$ and $b=30$ with a gcd of 6. The linear combination 
$4\cdot a-2 \cdot b=72-60=12=2\cdot6$ is twice the gcd.  


A more general take on productiong *multiples* of the
gcd: Multiply the gcd-specific "x and y" 
linear combination by any factor. 


$a=2 \cdot 3 \cdot 11$ and $b=2 \cdot 5 \cdot 7$.


$(66, 70) = 2 = 66 \cdot x + 70 \cdot y$. Values for $x$ and $y$ that
produce the gcd are $x=-18$ and $y=17$.


Multiply the gcd expression by $3$ to arrive at a new linear combination of 
$a$ and $b$ equal to $6$, three times the gcd. This $6$ is not a common divisor 
of $66$ and $70$ but it *is* a linear combination that is less than both of them.


This gives me a better handle on linear combinations in relation to 
*gcd*s. 


Additional observation:

Additional observation: Any linear combination of $a$ and $b$ will be a multiple of $(a, b)$.

Additional observation: 'How do I solve for $(a, b)$? Suppose I take a linear combination 
of $a$ and $b$ and arrive at some number $q$ less than both $a$ and $b$. Then I come up with

#### 1.12 a) Prove or provide a counterexample: $a^n | b^n \implies a | b$


For $n = 1$ the assertion is true. Continuing with $n > 1$:
$a^n | b^n \implies a \cdot a^{n-1} \cdot c = b \cdot b^{n-1}$ with $c > 0$.
This means that either $a | b$ or $a | b^{n-1}$.
If $a | b$ the assertion is true. Taking $a \nmid b$
implies $a \ne 1$ and this
leads to a contradiction when we continue casting out factors of $b$ to reduce
the exponent to $0$: $a$ must divide in succession:
$b^{n-2},\; b^{n-3},\; \dots \; b^{1}, \; 1$.

$$(\mu_k)_p(x) = \sum_{n=0}^{k} \mu_k(p^n)x^n = 1 + x + \sum_{n=2}^{k} \mu_k(p^n)x^n.$$


The prime $p$ is raised to a value less than $k$ for most of this sum, i.e. $\mu_k = 1$;
and the final term $n=k$ resolves to $(-1)^1$:


$$(\mu_k)_p(x) = \sum_{n=0}^{k} \mu_k(p^n)x^n = 1 + x + \sum_{n=2}^{k-1} x^n - x^k.$$


Observing the pattern:


$$k=1: (\mu_k)_p(x) = 1-x$$


$$k=2: (\mu_k)_p(x) = 1 + x - x^2$$


$$k=3: (\mu_k)_p(x) = 1 + x + x^2 - x^3$$


$$k=4: (\mu_k)_p(x) = 1 + x + x^2 + x^3 - x^4$$


$$k:(\mu_k)_p(x) = 1 + x + x^2 + \dots + x^{k-1} - x^k$$


In the spirit of a telescoping series: Multiply the last (general) expression above by $1-x$: 


$$(1-x) \cdot (1 + x + x^2 + \dots + x^{k-1} - x^k) =  
1 + x + x^2 + \dots + x^{k-1} - x^k - x - x^2 - x^3 - \dots - x^{k-1} - x^{k} + x^{k+1} = 1 - 2x^k + x^{k+1}.$$


So the general expression can be written 
$1 + x + x^2 + \dots + x^{k-1} - x^k = \frac{1 - 2x^k + x^{k+1}}{1-x} = (\mu_k)_p(x).\;\;$ &#x2610;
