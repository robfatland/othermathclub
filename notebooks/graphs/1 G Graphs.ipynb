### Friendly introduction

The above diagram is a graph. Looks like a moose or something. Maybe a bison.  The little circles are

If you copy this bison into your notebook (do this now so you have a spare bison at all times)

To sum up, then: Probability relies on counting possibilities which is called *combinatorics*
and meanwhile *graphs* are *also* about counting possibilities; so in both cases we need to learn 
the basics of *sets* and *counting*.

### Paths and Cycles and Circuits

You can find many paths in a graph that travel along edges from one vertex to another.  A path that 
comes back to the vertex where it begins is called a cycle. Such a path, one that returns to where 
it begins, is not allowed to cross an edge more than once. Consequently we see that a cycle involves 
at least three vertices and at least three edges. 

<br><br>${\Huge \infty}$

How many cycles are there in the graph shown below? One is shown in pink.

This question is simple to ask but the answer is rather long. Before reading further
you can take a moment to double-check your answer. There are $37$ cycles.


#### A rather long answer

How many cycles are there in this graph shown above? The graph has five vertices and 
the cycles must be distinct from one another. Let's count them all... but I will use 
the notation for combinations. If you are not familiar with this you can learn about
it from the beginning on the **probability** page. 


Three vertices are necessary for a cycle. Since all possible edges are present 
in this graph the number of possible 3-vertex cycles is equal to the number of ways 
that $3$ things can be chosen from a set with $5$ elements. This is called "$5$ choose $3$"
and is written $\binom{5}{3}$. It equals $10$.  A 3-cycle has no direction so we do not 
count it twice, like *forwards and backwards*. So there are 10 3-cycles.


For a cycle visiting four of the five vertices: One vertex is always left out. Hence 
there are five such choices of four vertices. Let's call one of these $\{ A, B, C, D \}$. 
A cycle can be written as $ABCD$ since we know that $D$ connects back to $A$ to complete the 
cycle. We could also follow $ABDC$, $ACBD$, $ACDB$, $ADBC$ and $ADCB$ making a total of six. 
However notice that $ABCD$ and $ADCB$ represent the same cycle traversed in opposite directions; 
which again we take to be the same cycle. To avoid double-counting in this way we divide 
by two to get three possible cycles for a given choice of four vertices. Therefore since 
there are five such choices with three distinct cycles each: There are 15 4-cycles.

Finally we have the Hamiltonian cycles that touch all five vertices. There is only one 
choice for the vertex set, ${ A, B, C, D, E }$. How many ways can we arrange these, being 
careful not to double-count the reverse direction cycle? Let's always choose $A$ as the 
starting vertex since $A$ is sure to be in there somewhere. Then there are four choices 
for the second vertex, three for the third, and two choices for the fourth. Only one choice 
will remain for the fifth vertex; for a total of 24 possible sequences. However as 
before this double-counts. For example the sequence $ACEDB$ is the reverse of $ABDEC$.

This accounts for all possible cycles so the total is $10 + 15 + 12 = 37$. 

A graph that contains all possible edges, incidentally, is called a *complete 
graph*. This is usual written $K_n$ where $n$ is either abstract or is an actual
number. For example in this case we have found the 
number of cycles in $K_5$ to be $37$.
