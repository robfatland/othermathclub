#### Professor Hilbert's Hotel $\star\star$
This is a paradoxical problem involving the strange nature of countable infinity.
This problem can -- according to some students -- kind of drive you crazy... so beware.


Professor Hilbert has purchased a large hotel for cheap and has gone into business, 
hanging a "VACANCY" sign out front to let people know he has available rooms. The sign 
has a switch that turns on the word "NO" in front of "VACANCY" so that it reads "NO VACANCY". 
Professor Hilbert also discovers an intercom in the front office that allows him to make 
announcements to the occupants of all of the rooms at once. Guests are permitted to stay at 
the hotel under one condition: If they hear a room reassignment over the intercom they 
must relocate to the new room instantly. The problem here is to come up with a strategy 
involving the intercom such that Professor Hilbert is never obliged to change his sign to read
NO VACANCY switch.


Here is the key piece of information that I have not told you yet: 
Upon going down the main hallway for the first time Professor Hilbert discovers to his surprise that
the hotel has an infinite number of rooms, each numbered in sequence 1, 2, 3, 4, 5 and so on. 
(This is a mathematical hotel (and that means comfort!))


On the first day of business a large bus pulls up in front of the hotel. The bus driver says to 
Professor Hilbert "I have an infinite number of passengers. They all need rooms for the night." 
Professor Hilbert responds "No problem: Simply have them get off the bus and fill up the rooms 
in order. The first person off goes to room 1, the second to room 2, and so forth. Everyone gets a 
room and all is well." This works like a charm until every guest has a room. (Remember they move
about at the speed of thought, which is close enough to instantaneous that we do not worry about
the time spent getting into the rooms.) But now the first problem comes up: The bus driver returns 
to the hotel office and says 'I'm sorry I forgot to tell you: I need a room too.' What does 
Professor Hilbert do???

#### Uninteresting positive integers
Speaking of interest: Mathematicians are fond of proving that there are no uninteresting numbers, 
as follows: Suppose we take one and two as interesting numbers (I certainly think they are) and
then suppose that many numbers are in fact interesting... but perhaps there are numbers that are
*not* interesting. If this is true then there must be a *smallest* uninteresting number; which we
will all remember for this reason. But paradoxically this makes that number *interesting* so it 
is no longer *uninteresting*. Now there must be a larger smallest uninteresting number... which 
makes that number interesting. And so on. Because there can be no smallest uninteresting number
there can not be *any* uninteresting numbers. And this form of proof reminds me of a joke...
