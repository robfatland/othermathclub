### Forward: On the unreasonable effectiveness of asking for help


This essay indirectly motivates taking up a study of number theory. It 
was inspired by two people: Caltech mathematics professor Tom Apostol; 
and Scientific American columnist Martin Gardner. Both as Paul Erdos would 
say have 'left'. 


> Apocryphal story: 
A student of rock climbing was led by his instructor to a rock 
formation. 'Here is your problem: Climb
to the top' the instructor said, pointing to a vertical cliff face.
Whereupon (the instructor departing) 
the student examined the indicated cliff face, finding it to be vertical, smooth 
and featureless, with not so much as a tiny handhold. 
The student sat upon the ground, staring at the cliff. 
A year later he returned to his instructor.
'I have not solved the problem yet' he reported.
His instructor responded, 'Oh I assumed you had died.'



I had the privilege of attending Professor Apostol's 
lectures in person in 1982. In keeping with Caltech's
culture of irreverence, my fellow classmates and I referred to 
Tom Apostol as '*Tommy*'. I see him in my mind's 
eye standing before the blackboard in Noyes Laboratory, 
all business as he begins a lecture on a version
of calculus more rigorous than the high school version. 
At some point in his narrative he would make an impossible, 
outrageous mathematical claim. There would follow a pause to 
let it sink in... one might scoff quietly... 
and then Professor Apostol 
would continue with the portentous remark,
"Now nobody can stop me from *making this claim*..." and then
another pause... do we see it yet?... and the logical 
argument would follow, at the end of 
which said claim is quite reasonably true. As Freshmen 
we did not always appreciate the fun of this approach. 
Speaking for myself: My eventual enlightenment necessitated
additional staring at notes and asking upperclassmen for help.


Now in retrospect I 
appreciate Tommy's lecture style. He happily carried the
burden of justifying his results; and always one had a
sense of forging ahead. Better keep up!


The intent of this notebook is to circle back 
and start over with Tommy, on a new subject, the topic 
of number theory. 
The starting point is grade school:
integers, addition, multiplication, divisibility, primes and
composite numbers. But chapter one of 
**Introduction to Analytic Number Theory**
accelerate rapidly, forging ahead with non-intuitive building blocks
centered on the greatest common divisor. 
Chapter two introduces the Mobius 
$\mu(n)$ function and Euler's totient $\varphi(n)$
in addition to Dirichlet multiplication. And we're off, we're 
forging ahead.


Will it be worthwhile, traveling into this new world?
Well... nobody can prevent me from *claiming* that number theory is 
a great deal of fun.



Going back still further in my personal history: I have found an
interesting dichotomy in looking for clues to the big picture. 
The book **Dinosaurs and More Dinosaurs** (1971) says aloud on every 
page, 'There is a big world out there beyond your direct experience.'
Being seven and I must admit rather naive for my age I just took it 
as a given that this bigger world was real and ours to discover.
But then there were the two books about Alice and her complicated
dreams: Seeming real page by page but clearly fiction... so on the one 
hand Alice asserts a big world beyond experience *as one reads* but
once the book is finished and set down we're back to the ordinary. The 
dichotomy can be dealt with by looking for more books... but the point
is I wanted the world of dinosaurs and the Red Queen to be real. Dinosaurs
seemed established enough... but what real thing corresponded
to the red queen?


I did not arrive at a satisfactory resolution until I happened 
upon Martin Gardner in middle school (and I hasten to add: I also
happened upon computers, another story). In the pages of Scientific
American we were gifted at regular intervals with flashes of
something marvelous. Here at last was the Red Queen: To read 
Gardner's *Mathematical Games* column was to climb up on the mantle
above the fireplace, to look through into the Looking Glass room,
and as the mirror disssolved into mist, to climb through and 
become a character in the story. Here ideas mattered in proportion 
to their strangeness and unfamiliariarity.
Gardner gave us logic puzzles, paradoxes, prisoners both
clever and unfortunate, inward-exploding fractals,
planar graphs, cleverly colored maps, games about the search 
for truth, and on and on.



Having arrived at this state of being, 'there is a big world to discover', 
the question evolves into 'How then shall we explore?' The fuse is lit, 
we are off to explore. But what happens when the trail grows cold, when
we hit a wall, run out of ideas to try? This is the point of the story
of the rock climbing student. Consider an indirect method where one develops 
a skill called 'getting help'.  
Sometimes help is not available. Sometimes sources of 
help are hidden. Sometimes they walk away and we have to go chase them
down. Of course the reverse can be true as well: We can become so
habituated to help that we become fearful of the problem. So there is 
a balance to strike; but I feel that it is too easy to err on the 
side of not finding that help. 


This goes higher up.


Professor Apostol paid his students the 
highest compliment in an interview he gave in 1997 
(Donald J. Albers (1997) An Interview with Tom Apostol, The College 
Mathematics Journal, 28:4, 250-270, DOI: 10.1080/07468342.1997.11973873). 
The interview is a worthwhile read and I feel it reinforces Martin

enough persistence and an adequate supply of pencils, scratch paper

## Solution style elements


Learning this material depends on time, cash and patience. My narrative
is intended to help internalize tools and ideas; placed as separated
text between solutions.



Multiply by one example: Suppose $(a, b) = 1$ then there
exist $x$ and y$ such that $c = c \cdot (ax + by)$.


Telescope example: $\alpha^n-1$ suggests telescoping factorization
via factor $\alpha \; - \; 1$. A step further: $\alpha^n + 1$ might
be factored via alternating signs $+ \; - \; + \; - \; \dots - \; +$. 
(See problems **1.16** and **1.17**.)


On *ducking the wave* on some problems: By this I mean missing 
an elegant solution in favor of more pedestrian methods. 
I will plead no contest as I may start working a problem from
cases; which suggests *mechanical* solutions that duck the wave. 
(See problem 2.5.)


My notion of an adequate solution / proof is vague. My criteria is
that I'm convinced; where I try and remember that I am the
easiest person to fool.
