$\Delta$ problem 1: A triangle is drawn on a piece of paper with the pointy side up and the flat side down.
A student with a pencil happens along and draws three dots $\cdots$ on the midpoints of the three sides
of this triangle. Then she connects them with three straight lines. Now there are four smaller triangles,
three with pointy side up and one with pointy side down. 


The next day she happens upon the same piece of paper. This time she does the same procedure, but she
does this to the three smaller triangles in the figure that are pointy side up. The result is now
*nine* triangles, even smaller, that are pointy side up. Plus there are now additional triangles that
are pointy side down.


Of course every subsequent day she does the same thing, always dividing up all of the pointy-side-up
triangles in the figure.


What does the figure look like after five days?

$\Delta$ problem 2: A piece of graph paper has 49 x 49 empty boxes. The page is blank but for the
top row. This consists of 24 zeros, a single 1, and 24 more zeros. Our task is to apply a rule to
the first three boxes to write a number in the second row beneath the middle box. This skips the
edge box so we just make that a zero. Same with the last cell. This leaves 47 boxes to fill in.
For each one of the 47 boxes: Look at the three boxes in the row directly above and apply the rule:
Write either a 1 or a 0. 


The rule:


```
000  >  0
001  >  1
010  >  0
011  >  1
100  >  1
101  >  0
110  >  1
111  >  0
```

> **Information on poisonous cone snails**: There is a family of ocean snails called the *cone snails* 
that eat tiny fishes and other snails. They do this by firing a little dart or harpoon at their
prey, say like an unsuspecting fish who swims
too close. In the dart is a mix of various small proteins called peptides. 
When the dart hits the fish 
the proteins are injected into it. These proteins are designed to interfere
with the fish nervous system so the fish can no longer move. The word for 'nerves' is 
*neuro* and the word for 'poison' is *toxin*. The proteins injected by the cone snail's dart 
are therefore called *neurotoxins*. There are 
over 100 different types of cone snail neurotoxins.


The pattern on the shell of a particular type of cone snail -- the textile cone snail --
resembles this rule ('rule 30') applied to the graph paper problem given above.


```
000           0
001           1
010           1
011           1
100           1
101           0
110           0
111           0
```

### Graphs


A set (perhaps 'proper'?) has elements once. Like {3, 3} is not a proper set. {3L, 3R} is a proper set
although I do not know what it particularly means. 


Graphs are initially defined as two sets where the second set is defined in terms of the first. 
That is, the second set is some *additional information* concerning the first set. 


This suggests a preface about information as something quantifiable. The remark goes: Claude Shannon is one of
the founders of information theory; a branch of applied math built on probability and statistics and used for
(among other things) making mobile phones work. Mobile phones I am told are very popular; 
so learn the mathematics of information theory.


At this point it would be helpful to state the set-based definition of a graph. 



Graphs defined in the above way are rather abstract. One idea for making graphs 
come alive is to challenge a student to turn the definition into a picture. 


A second step in livening up graphs is to immediately begin converting other sorts of problems 
to graph representation. For example the problem of arriving at 4 gallons of water from a pool of water and two bottles
(with no markings on them) with capacity 3 and 5 gallons respectively can be represented as a directed graph. 
It is directed because some of the transitions are not reversible. 


What else? Can every mathematical problem be represented as a graph? This seems unlikely. But: The Konigsberg 
bridge problem is amenable... others escape me at the moment. 

With that as an opening volley on graphs there are of course many many branches leading 
off: Kozyrev-Grinberg, Euler's formula, projection of graphs onto spheres and other topological surfaces, 
derivation of platonic solids and some other degenerate forms, Hamiltonian cycles... Ross Honsberger's books are 
an excellent resources although his degree of difficulty clocks in closer to the high school / college level. 

Taking the Platonic solids for a moment: See the Wikipedia on dice to get some hints about other regular-ish solids 
that might be amenable to the same Euler formula discovery process. I think we should agree to use the singular 
dice because 'die' does not scan well for me. I mention 'other degenerate forms' because they are a fine example
of letting the math lead us to tangential solutions. Let me elaborate: The platonic solids arise from a formula
by Euler about planar graphs and a constraint about symmetry. From these we get the graph version of polyhedral
dice but we also get what amounts to a coin and a football. The coin and football are not platonic solids because
they require some non-ideal dimensional shenanigans; but we are familiar with them and they *do* fall out of
the equations. Anyway as a last fling here is that graph problems are intrinsically Diophantine; another branch. 

Another important mathematical stretch of this Platonic solid business is the question 'what does the everything
else face represent?' In my view it is the sort of thing that is self-evident and completely mysterious at the 
same time; and so for me required a number of minutes of staring and thinking with return visits before I could 
make sense of it.

A list of tractable proofs is worth developing...

- Erdos asked of Posa, answered: Given a graph of $2n$ vertices with $n^2+1$ edges contains a triangle. 
- Consider an infinite series whose $n$'th term is the fraction with numerator $1$ and denominator equal to 
the lowest common multiple of the integers $1, 2, \cdots , n$: Prove that the sum is irrational. 
- A graph with n vertices (n >= 4) and 2n-3 edges must contain a circuit with a diagonal. $\star\;\star\;\star\;\star$ 
- A graph with n vertices (>= 6) and 3n-5 edges must contain two circuits which have no vertices in common. 
- Every graph with n vertices and n+4 edges contains two circuits which have no edges in common. 
- Dirac's theorem on Hamiltonian circuits: A graph with n vertices >= 3 in which each vertex has valence at least n/2
has a Hamiltionian circuit: A path that visits each vertex exactly once and arrives again where it began. 
- Ore's theorem: If n >= 3 and, for every pair of vertices that are not doined by an edge, the sum 
of the valences is at least n, then the graph is Hamiltonian. 
- And of course K-Grin for Hamiltonian circuits on planar graphs is great fun, Gems I.


### Diophantine problems

Diophantine analysis begins with a collection of farm animals. See chapter 2 of _Wheels, Life_ by Gardner; which connects
to Fermat's Last; so of course this topic also connect very easily with the pythagorean formula since only certain right triangles give us three integral sides. Now let's throw off a branch into geometry, specifically area of triangles and
volume of triangular pyramids.  The idea would be to start with Diophantine problems and then suggest that (now that we 
have our boots on) we shall go through this gate into a field and off into probability and geometry.

The deep problem to consider is the probability that three uniformly random segments (say 0 to 1 meters each) can 
form a triangle. Coincidentally the solution I have in mind is built on understanding triangles but in a very
circuitous [sic] path. So let's throw out the rationale needed for the problem and then come back and state this 
as not **ABA** but rather as **A B C C' D E A**, no kidding!

Taking **A** as read. 

**B** We need to know the area of a triangle, regardless of its shape. This means any of its three sides can be considered
the base and it is an argument based on subtraction of areas that is necessary for the general answer to be understood
no matter how the triangle *leans over*. And notice the two in the denominator. 

**C** Now what about a triangular pyramid. Here we can suppose *area* of base times height, with a similar sort of 
process to cope with *leaning over*. But what is the denominator? Or why is it 3 instead of 2? Does this relate to 
dimension? Well yes, I claim it does. But why does dimension appear in the denominator? Incidentally there is a side
branch where we imagine a four-dimensional pyramid by borrowing time. Time is not the fourth dimension in mathematics
but Einstein borrowed that formalism of dimensio nand *used* time; so there we can make a cube melt uniformly and ask
what is its time-volume; and does that have a denominator of 4? And yes it does. (I think.) 

**C'** Now there is a whole class of intermezzo problems concerning triangular pyramids with three faces that are 
right angles; of which two of those right triangle faces have integral sides. A tenuous link back to Diophantine.
Now we are connecting spatial imagination with the formula to arrive at multi-step problems. NTS: (6, 10) and 
(20, 25) is a good one, for example: Two pyramids sharing a base. Easier is solving the volume of a single pyramid. 

**D** Now we need to establish probability spaces for first one, then two, then three random variables. One is trivial,
two can be made interesting, and three might require constructing a model. A frame? String? 

**E** What is the volume of the probability space of three uniform random variables who obey 'longest < short + short'?

And now re-state **A**

### Proof

Instructive: Prove something self-evident so as to find a solid implication in the process. Example: 
Numbers equal to the sum of their digits can have but one digit. What falls out is that the tens and hundreds 
and so on digits must be zero. This is an onramp divisibility by
9s which leads to 3s and thence to 11s. What is needed here is a set of confidence builders: Proofs that are fairly
straightforward but not quite trivial. 


Philosophical note: There is an essay in TBMW2016 about what constitutes *understanding* and it makes the point 
that the frog tends to die in the process: Belabor formal proofs and the math is not fun. On the other hand: 
Ignore this at your peril; being unable to click on that proof idea makes one credulous. We explored proofs of 
things that are false a little bit last year; the students had some trouble identifying the swindle. 

Also it is worth pointing out that the ability to reason algebraically is central to many many proofs; 
first the symbols dance and then they speak. Without the facility for this (say 6th and 7th grade) the
flow of the argument is lost. So again: Keep it on the side of simplicity. 

Proofs in graph theory are particularly interesting... see for example the simple problem of Erdos proven
by Posa under the graph theory topic. Also Erdos to Posa:
