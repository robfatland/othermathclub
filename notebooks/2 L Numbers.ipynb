### Relatively prime numbers

Here *numbers* are the *positive integers* unless we say otherwise. 

Prime numbers are the building blocks of all the numbers. Important!

Two numbers -- whether or not they are prime numbers -- can also be *relatively prime*. 
Let's give some examples and then define what is meant by *relatively prime*.

3 and 7 are relatively prime. 4 and 8 are not. 5 and 12 are relatively prime. 12 and 14 are not. 16 and 21 are relatively prime. 18 and 21 are not.

Any number can be expressed as the product of a set of prime numbers. For example 
$12=2\cdot2\cdot3$. That is, $2\cdot2\cdot3$ is the prime factorization of $12$. 
For all numbers there is only *one* prime factorization. (You can score a Brilliancy prize 
for proving this.)

We can place a numbers prime factors in a set. For example the set of prime factors for 48 is {2, 2, 2, 2, 3}.
A number with one prime factor is a prime of course. A number with more than one prime factor is a 
*composite* number. The number $1$ is neither prime nor composite. We say '1 is 1'. 

Now for the definition: Two numbers are relatively prime if their respective sets of prime factor 
share no common elements. 
Conversely two numbers are *not* relatively prime if their prime factor sets share at least one 
common element. Here are some examples.

40 and 41: 40 has { 2, 2, 2, 5 } and 41 has { 41 } so they are relatively prime.

64 and 38: 64 has { 2, 2, 2, 2, 2, 2 } and 38 has { 2, 19 } so they are *not* relatively prime.

12 and 35: 12 has { 2, 2, 3 } and 35 has { 5, 7 } so they are relatively prime.

Suppose you have a rational number $a/b$. Describe a method for reducing this to a ratio of two *relatively prime* numbers.


#### Hint

Any number divided by itself is 1.


#### Solution

Suppose two numbers $a$ and $b$ in a ratio $a/b$ do not share any prime factors. Then we are done: They are relatively prime.

Suppose on the other hand that $a$ and $b$ share some prime factors. Write $a/b$ as prime factors and cancel all of the common factors. What remains will be a ratio of two numbers that are relatively prime. 

Did this all make sense? 

By the way: Numbers that are relatively prime are also sometimes called *coprime*.

### Is $n$ prime?

You know that $7^2=49$ so $49$ is a composite number, not a prime. Now if you didn't know this about 49; and if you were curious whether $49$ is prime you could ask *Is $49$ divisible by any smaller numbers?* You check 2 and nope, you check 3 and nope... do you need to check 4? No; if a number is not divisible by $2$ then it is not divisible by $4$. Next you check whether $49$ is divisible by $5$: No. Do we check $6$? No. If $49$ is not divisible by $2$ then it is not divisible by $6$.  So next you check $7$. Sure enough $49$ is divisible by $7$ so it is not a prime. 

Now consider an arbitrary number $n$. Is it prime? How do we figure this out? 

We could try dividing $n$ by every number from $2$ to $n-1$ to see if anything divides evenly 
into $n$. If one of these numbers *does* divide into $n$ then $n$ is not prime. 
Otherwise $n$ *is* prime. However: Checking all of those 
numbers seems excessive. We already saw that $4$ is not worth checking once you check $2$. $n$ is not divisible by $n-1$ so we don't need to check that. In fact we don't need to check composites; so we really only need to check the primes up to $n/2$, if that many. We could check $2, 3, 5, 7, ..., n/2$ and if none of them divide into $n$ evenly then we can be sure $n$ is prime. 

Let's spell this out for $49$ by saying *'Let $n=49$.'* Then $n/2$ is about $25$ and we have to check whether $49$ is divisibile by all of these: $\{2, 3, 5, 7, 11, 13, 17, 19, 23\}$. As soon as we find one ($7$ in this case) that divides evenly int $n$ ($49$ in this case) then we do not need to check any of the others; we know $n$ is composite.

##### Now for the problem...
Can we do better? In other words: Can we cut off our search for prime factors of $n$ at some number less than $n/2$? If so: What is it?

To state this problem more formally: Given some positive integer $n$ what is the smallest set of numbers we must
check for divisibility into $n$ to determine if $n$ is prime?


#### Hint for the problem above

Notice that if $n$ is composite then it can be written as the product of two numbers:

* Either those numbers are identical (and $n$ is a perfect square)
* Or one of those numbers is larger than $\sqrt{n}$ and the other is smaller than $\sqrt{n}$


#### Solution

We only need to check prime numbers for divisibility into $n$. Why? Because if $n$ is divisible by a composite then it 
will be divisible by all of the prime factors of that composite. This reduces the number of numbers we 
need to check. 

We only need to check numbers $<= \sqrt{n}$ since any factors of $n$ larger than $\sqrt{n}$ will be 
multiplied by numbers smaller than $\sqrt{n}$.

Hence the set of primes less than or equal to $\sqrt{n}$ will be sufficient to test 
('do these divide evenly into $n$?') 
to determine if $n$ is prime.

For style points: We could also use the floor function in our solution: 
check the primes $p <= \big{\lfloor}\sqrt{n}\;\big{\rfloor}$.
